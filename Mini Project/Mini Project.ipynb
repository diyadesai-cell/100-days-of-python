```python
# Bank Account Simulator

account_name = "Diya Desai"
balance = 5000

transactions = [
    ("deposit", 2000),
    ("withdraw", 1500),
    ("deposit", 3000),
    ("withdraw", 800),
    ("withdraw", 6000),
    ("deposit", 500)
]

total_deposits = 0
total_withdrawals = 0

# Process transactions
for transaction_type, amount in transactions:

    if transaction_type == "deposit":
        balance += amount
        total_deposits += amount
        print(f"Deposited Rs.{amount} | Balance: Rs.{balance}")

    elif transaction_type == "withdraw":

        if balance >= amount:
            balance -= amount
            total_withdrawals += amount
            print(f"Withdrew Rs.{amount} | Balance: Rs.{balance}")

        else:
            print(f"❌ Insufficient funds for Rs.{amount} | Balance: Rs.{balance}")

# Final Statement
print("-------------------")
print(f"Account: {account_name}")
print(f"Final Balance: Rs.{balance}")
print(f"Total Deposits: Rs.{total_deposits}")
print(f"Total Withdrawals: Rs.{total_withdrawals}")
print(f"Transactions processed: {len(transactions)}")
```

```python
# Student Report Card Generator

students = [
    ("Diya", 88, 92, 76, 85),
    ("Raj", 72, 65, 80, 70),
    ("Amit", 95, 98, 92, 97),
    ("Sara", 61, 55, 70, 58),
    ("Kiran", 79, 82, 75, 88)
]

# Subject totals
maths_total = 0
science_total = 0
english_total = 0
computer_total = 0

# Class summary variables
class_percentage_sum = 0
passed_students = 0

topper_name = ""
topper_percentage = 0

# Process each student
for student in students:

    name, maths, science, english, computer = student

    total = maths + science + english + computer
    percentage = round((total / 400) * 100, 2)

    # Grade calculation
    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    else:
        grade = "F"

    # Result
    if percentage >= 60:
        result = "Pass"
        passed_students += 1
    else:
        result = "Fail"

    # Add to class totals
    class_percentage_sum += percentage

    maths_total += maths
    science_total += science
    english_total += english
    computer_total += computer

    # Find topper
    if percentage > topper_percentage:
        topper_percentage = percentage
        topper_name = name

    # Print report card
    print(f"--- {name} ---")
    print(
        f"Marks: {maths} {science} {english} {computer} | "
        f"Total: {total}/400 | {percentage}% | "
        f"Grade: {grade} | {result}"
    )

# Class average
class_average = round(class_percentage_sum / len(students), 2)

# Subject totals dictionary
subject_totals = {
    "Maths": maths_total,
    "Science": science_total,
    "English": english_total,
    "Computer": computer_total
}

# Highest and lowest subject
best_subject = max(subject_totals, key=subject_totals.get)
worst_subject = min(subject_totals, key=subject_totals.get)

# Print class summary
print("=== CLASS SUMMARY ===")
print(f"Class average: {class_average}%")
print(f"Topper: {topper_name} ({topper_percentage}%)")
print(f"Students passed: {passed_students}")
print(f"Best subject total: {best_subject} ({subject_totals[best_subject]})")
print(f"Worst subject total: {worst_subject} ({subject_totals[worst_subject]})")
```

```python
# Shopping Cart System

# Inventory: (product, price, stock)
inventory = [
    ("Rice", 60, 50),
    ("Oil", 180, 10),
    ("Sugar", 45, 30),
    ("Milk", 60, 20),
    ("Bread", 35, 15),
    ("Butter", 95, 8)
]

# Customer cart: (product, quantity)
cart = [
    ("Rice", 3),
    ("Oil", 2),
    ("Butter", 10),
    ("Sugar", 5),
    ("Biscuit", 2)
]

subtotal = 0
items_purchased = 0
items_skipped = 0

receipt = []

print("Processing cart...")

# Process cart items
for cart_product, cart_quantity in cart:

    product_found = False

    for product, price, stock in inventory:

        if cart_product == product:
            product_found = True

            # If requested quantity is more than stock
            if cart_quantity > stock:
                print(f"⚠️ {product}: only {stock} in stock, adding {stock}")

                if stock > 0:
                    item_total = price * stock
                    subtotal += item_total
                    receipt.append((product, stock, item_total))

                    print(f"✅ {product} x{stock} = Rs.{item_total}")
                    items_purchased += 1

            else:
                item_total = price * cart_quantity
                subtotal += item_total
                receipt.append((product, cart_quantity, item_total))

                print(f"✅ {product} x{cart_quantity} = Rs.{item_total}")
                items_purchased += 1

    # Product not found
    if not product_found:
        print(f"❌ {cart_product} not found")
        items_skipped += 1

# Discount calculation
if subtotal >= 1000:
    discount_percent = 15
elif subtotal >= 500:
    discount_percent = 10
else:
    discount_percent = 5

discount_amount = subtotal * discount_percent / 100
final_total = subtotal - discount_amount

# Print receipt
print("--- RECEIPT ---")

for product, quantity, total in receipt:
    print(f"{product} x{quantity} = Rs.{total}")

print("---")
print(f"Subtotal: Rs.{subtotal}")
print(f"Discount: {discount_percent}% = Rs.{discount_amount}")
print(f"Total: Rs.{final_total}")

print(f"Items purchased: {items_purchased} | Items skipped: {items_skipped}")
```
