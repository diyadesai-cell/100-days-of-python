```python
# Takes a function as argument:
def apply(func, value):
    return func(value)

print(apply(str.upper, "hello"))   # HELLO
print(apply(len, "hello"))         # 5

# Returns a function:
def multiplier(n):
    return lambda x: x * n

double = multiplier(2)
triple = multiplier(3)
print(double(5))   # 10
print(triple(5))   # 15
```

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# map — apply function to every item:
squares = list(map(lambda x: x**2, nums))
# [1, 4, 9, 16, 25]

# filter — keep items where function returns True:
evens = list(filter(lambda x: x%2==0, nums))
# [2, 4]

# reduce — fold list into single value:
total = reduce(lambda acc, x: acc + x, nums)
# 15  (1+2+3+4+5)
product = reduce(lambda acc, x: acc * x, nums)
# 120  (1×2×3×4×5)
```

```python
#wraps — preserve function metadata in decorators:
```

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)           # preserves __name__, __doc__ etc
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@my_decorator
def greet(name):
    """Greets a person."""
    return f"Hello, {name}!"

print(greet.__name__)   # greet (not 'wrapper'!)
print(greet.__doc__)    # Greets a person.
```

```python
#Function composition:
```

```python
def compose(*funcs):
    """Apply functions right to left: compose(f,g,h)(x) = f(g(h(x)))"""
    from functools import reduce
    return reduce(lambda f, g: lambda x: f(g(x)), funcs)

add1 = lambda x: x + 1
double = lambda x: x * 2
square = lambda x: x ** 2

pipeline = compose(str, add1, double, square)
print(pipeline(3))   # str(add1(double(square(3)))) = str(19) = '19'
```

```python
from functools import reduce

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
words = ["hello", "WORLD", "Python", "ROCKS", "coding"]

# 1. Square all numbers
squares = list(map(lambda x: x ** 2, nums))

# 2. Keep only numbers > 4
greater_than_4 = list(filter(lambda x: x > 4, nums))

# 3. Sum all numbers using reduce
total_sum = reduce(lambda a, b: a + b, nums)

# 4. Multiply all numbers using reduce
total_product = reduce(lambda a, b: a * b, nums)

# 5. Normalize all words to title case
title_words = list(map(lambda w: w.title(), words))

# 6. Keep only words longer than 5 chars
long_words = list(filter(lambda w: len(w) > 5, words))

# 7. Find longest word using reduce
longest_word = reduce(
    lambda a, b: a if len(a) >= len(b) else b,
    words
)

print(squares)
print(greater_than_4)
print(total_sum)
print(total_product)
print(title_words)
print(long_words)
print(longest_word)
```

```python
from functools import partial, reduce
import operator

# -----------------------------
# 1. Using partial
# -----------------------------

# add5: adds 5 to any number
add5 = partial(operator.add, 5)

# multiply_by_3: multiplies by 3
multiply_by_3 = partial(operator.mul, 3)

# is_positive: checks if number > 0
is_positive = partial(operator.lt, 0)  # 0 < x

# starts_with_p: checks if string starts with "P"
starts_with_p = partial(
    lambda prefix, s: s.startswith(prefix),
    "P"
)

print(add5(10))                # 15
print(multiply_by_3(7))        # 21
print(is_positive(10))         # True
print(starts_with_p("Python")) # True

# -----------------------------
# 2. operator module functions
# -----------------------------

# Sum 1..10
total = reduce(operator.add, range(1, 11))
print(total)

# 10!
factorial_10 = reduce(operator.mul, range(1, 11))
print(factorial_10)

# Sort tuples by second element
data = [('a', 2), ('b', 3), ('c', 1)]
sorted_data = sorted(data, key=operator.itemgetter(1))
print(sorted_data)

# -----------------------------
# 3. Power of 2 checker
# -----------------------------

power_of_2 = partial(pow, 2)

print(f"power(2,6)={power_of_2(6)}")

def is_power_of_2(n):
    exp = 0
    while power_of_2(exp) < n:
        exp += 1
    return power_of_2(exp) == n

print(f"64 is power of 2: {is_power_of_2(64)}")
print(f"100 is power of 2: {is_power_of_2(100)}")
```

```python
from functools import reduce
import string

# ---------------------------------
# Function Composition Utilities
# ---------------------------------

def compose(*funcs):
    """Right-to-left composition"""
    def inner(x):
        for func in reversed(funcs):
            x = func(x)
        return x
    return inner


def pipe(*funcs):
    """Left-to-right composition"""
    def inner(x):
        for func in funcs:
            x = func(x)
        return x
    return inner


# ---------------------------------
# Example 1: compose and pipe
# ---------------------------------

double = lambda x: x * 2
add1 = lambda x: x + 1

print("compose(str,add1,double)(5):",
      compose(str, add1, double)(5))

print("pipe(double,add1,str)(5):",
      pipe(double, add1, str)(5))


# ---------------------------------
# Example 2: Text Processing Pipeline
# ---------------------------------

stop_words = {"is", "a", "the", "and", "of", "to"}

def strip_text(text):
    return text.strip()

def lower_text(text):
    return text.lower()

def remove_punct(text):
    return text.translate(
        str.maketrans("", "", string.punctuation)
    )

def split_words(text):
    return text.split()

def remove_stops(words):
    return [word for word in words if word not in stop_words]

def join_words(words):
    return " ".join(words)

text_pipeline = pipe(
    strip_text,
    lower_text,
    remove_punct,
    split_words,
    remove_stops,
    join_words
)

text = "   Hello, World! Python Rocks.   "
print("Text pipeline:", text_pipeline(text))


# ---------------------------------
# Example 3: Number Processing Pipeline
# ---------------------------------

def filter_evens(nums):
    return list(filter(lambda x: x % 2 == 0, nums))

def square(nums):
    return list(map(lambda x: x * x, nums))

def sum_nums(nums):
    return sum(nums)

number_pipeline = pipe(
    filter_evens,
    square,
    sum_nums
)

numbers = [2, 4, 6, 8]
print("Number pipeline:", number_pipeline(numbers))


# ---------------------------------
# Example 4: Student Validation Pipeline
# ---------------------------------

def validate_age(student):
    if student["age"] < 18:
        raise ValueError("Invalid age")
    return student

def validate_score(student):
    score = student["score"]
    if not (0 <= score <= 100):
        raise ValueError("Invalid score")
    return student

def add_grade(student):
    score = student["score"]

    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"

    student["grade"] = grade
    return student

def format_output(student):
    return (
        f"{student['name']} | "
        f"Score:{student['score']} | "
        f"Grade:{student['grade']}"
    )

student_pipeline = pipe(
    validate_age,
    validate_score,
    add_grade,
    format_output
)

students = [
    {"name": "Diya", "age": 21, "score": 88},
    {"name": "Raj", "age": 20, "score": 72},
    {"name": "Amit", "age": 22, "score": 95}
]

for student in students:
    print(student_pipeline(student))
```

```python
import time
from functools import wraps

# ---------------------------------
# 1. timer decorator
# ---------------------------------

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()

        print(f"{func.__name__} took {end - start:.2f}s")
        return result

    return wrapper


# ---------------------------------
# 2. counter decorator
# ---------------------------------

def counter(func):
    count = 0

    @wraps(func)
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1

        print(f"{func.__name__} called {count} times")
        return func(*args, **kwargs)

    return wrapper


# ---------------------------------
# 3. memoize decorator
# ---------------------------------

def memoize(func):
    cache = {}

    @wraps(func)
    def wrapper(*args):
        if args in cache:
            print(f"{func.__name__}{args} from cache")
            return cache[args]

        result = func(*args)
        cache[args] = result
        return result

    return wrapper


# ---------------------------------
# 4. validate decorator
# ---------------------------------

def validate(*validators):
    def decorator(func):

        @wraps(func)
        def wrapper(*args, **kwargs):

            for validator in validators:
                valid, message = validator(*args, **kwargs)

                if not valid:
                    print(f"Validation failed: {message}")
                    return None

            print("Validation passed")
            return func(*args, **kwargs)

        return wrapper

    return decorator


# ---------------------------------
# 5. retry decorator
# ---------------------------------

def retry(times=3, exceptions=(Exception,)):
    def decorator(func):

        @wraps(func)
        def wrapper(*args, **kwargs):

            for attempt in range(1, times + 1):
                try:
                    result = func(*args, **kwargs)

                    if attempt > 1:
                        print(f"Success on attempt {attempt}")

                    return result

                except exceptions:

                    if attempt < times:
                        print(
                            f"Attempt {attempt} failed, retrying..."
                        )
                    else:
                        raise

        return wrapper

    return decorator


# =================================
# TEST 1: timer
# =================================

@timer
def slowfunc():
    time.sleep(0.1)

slowfunc()


# =================================
# TEST 2: counter
# =================================

@counter
def greet():
    print("Hello!")

greet()
greet()
greet()


# =================================
# TEST 3: timer + memoize fibonacci
# =================================

@timer
@memoize
def cached_fib(n):
    if n <= 1:
        return n

    return cached_fib(n - 1) + cached_fib(n - 2)

print(f"fib(30) = {cached_fib(30)}")
cached_fib(30)      # from cache


# =================================
# TEST 4: validate
# =================================

def age_validator(name, age):
    return age >= 0, "age must be >= 0"


@validate(age_validator)
def register(name, age):
    print(f"Registered {name}")


register("Diya", 21)
register("Raj", -5)


# =================================
# TEST 5: retry
# =================================

attempts = 0

@retry(times=3)
def unstable():

    global attempts
    attempts += 1

    if attempts < 3:
        raise ValueError("Temporary failure")

    return "Done"


unstable()
```

```python
import time
import inspect
from functools import wraps

# ==========================================
# 1. curry(func)
# ==========================================

def curry(func):
    required_args = len(inspect.signature(func).parameters)

    def curried(*args):
        if len(args) >= required_args:
            return func(*args)

        return lambda *more: curried(*(args + more))

    return curried


# ==========================================
# 2. memoize_with_ttl(ttl_seconds)
# ==========================================

def memoize_with_ttl(ttl_seconds):
    def decorator(func):
        cache = {}

        @wraps(func)
        def wrapper(*args):
            current_time = time.time()

            if args in cache:
                result, timestamp = cache[args]

                if current_time - timestamp < ttl_seconds:
                    return result

            result = func(*args)
            cache[args] = (result, current_time)
            return result

        return wrapper

    return decorator


# ==========================================
# 3. once(func)
# ==========================================

def once(func):
    has_run = False
    saved_result = None

    @wraps(func)
    def wrapper(*args, **kwargs):
        nonlocal has_run, saved_result

        if not has_run:
            saved_result = func(*args, **kwargs)
            has_run = True

        return saved_result

    return wrapper


# ==========================================
# 4. throttle(func, min_interval)
# ==========================================

def throttle(min_interval):
    def decorator(func):
        last_called = 0

        @wraps(func)
        def wrapper(*args, **kwargs):
            nonlocal last_called

            current_time = time.time()

            if current_time - last_called >= min_interval:
                last_called = current_time
                return func(*args, **kwargs)

        return wrapper

    return decorator


# ==========================================
# 5. compose_async(*funcs)
# Skip pipeline if a step returns None
# ==========================================

def compose_async(*funcs):

    def composed(value):

        for func in funcs:
            value = func(value)

            if value is None:
                return None

        return value

    return composed


# ==========================================
# CURRY TESTS
# ==========================================

@curry
def add(a, b, c):
    return a + b + c

print("curry add:", add(2)(3)(5))
print("curry add:", add(5)(5)(5))


# ==========================================
# ONCE TEST
# ==========================================

@once
def load_config():
    print("Initializing config...")
    return {
        "debug": False,
        "version": "1.0"
    }

config1 = load_config()
print("Config:", config1)

config2 = load_config()
print("Config (cached):", config2)

if config1 is config2:
    print("once: result is same")


# ==========================================
# MEMOIZE WITH TTL TEST
# ==========================================

@memoize_with_ttl(2)
def expensive_square(x):
    print("Computing...")
    return x * x

expensive_square(10)
expensive_square(10)  # cached


# ==========================================
# THROTTLE TEST
# ==========================================

call_count = 0

@throttle(0.5)
def limited_call():
    global call_count
    call_count += 1

for _ in range(10):
    limited_call()
    time.sleep(0.2)

print(f"Throttle: {call_count} calls made in limited time")


# ==========================================
# DATA TRANSFORMATION PIPELINE
# ==========================================

@curry
def replace_word(old, new, text):
    return text.replace(old, new)

replace_python = replace_word("python", "PYTHON")
replace_world = replace_word("world", "WORLD")

def strip_text(text):
    return text.strip()

def lower_text(text):
    return text.lower()

def upper_text(text):
    return text.upper()

pipeline = compose_async(
    strip_text,
    lower_text,
    replace_python,
    replace_world,
    upper_text
)

result = pipeline("   Hello World Python   ")
print("Pipeline:", result)


# ==========================================
# NONE SKIP TEST
# ==========================================

def reject_empty(text):
    if not text:
        return None
    return text

safe_pipeline = compose_async(
    reject_empty,
    upper_text
)

safe_pipeline("")
safe_pipeline("hello")


# ==========================================
# COMPLETE
# ==========================================

print("Functional toolkit complete!")
```
