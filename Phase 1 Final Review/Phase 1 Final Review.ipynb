```python
# Word Frequency Counter
```

```python
text = "the cat sat on the mat the cat is fat"

# 1. Split text into words
words = text.split()

# 2. Find unique words while maintaining order
unique_words = []

for word in words:
    if word not in unique_words:
        unique_words.append(word)

# Variables for most frequent word
most_frequent_word = ""
highest_count = 0

# 3 & 4. Count occurrences and print
for unique_word in unique_words:

    count = 0

    for word in words:
        if word == unique_word:
            count += 1

    print(f"{unique_word}: {count}")

    # 5. Find most frequent word
    if count > highest_count:
        highest_count = count
        most_frequent_word = unique_word

# 5. Print most frequent word
print(f"Most frequent: {most_frequent_word} ({highest_count} times)")

# 6. Print total unique words
print(f"Unique words: {len(unique_words)}")
```

```python
# Leaderboard Sorter
```

```python
scores = [
    ("Diya", 8540, "India"),
    ("Alex", 9200, "USA"),
    ("Raj", 8540, "India"),
    ("Yuki", 7800, "Japan"),
    ("Sara", 9200, "USA"),
    ("Leo", 8100, "Brazil")
]

# 1. Sort by score descending, then name alphabetically
sorted_scores = sorted(scores, key=lambda x: (-x[1], x[0]))

# Variables for calculations
total_score = 0
top_3_countries = []
indian_players = []

# 2. Print ranked leaderboard
rank = 1

for name, score, country in sorted_scores:

    print(f"#{rank} {name:<8} | {score} pts | {country}")

    total_score += score

    # Store top 3 countries
    if rank <= 3:
        top_3_countries.append(country)

    # Store Indian players
    if country == "India":
        indian_players.append(name)

    rank += 1

# 3. Print top 3 countries
print(f"Top 3 countries: {', '.join(top_3_countries)}")

# 4. Print average score
average_score = round(total_score / len(scores), 2)
print(f"Average score: {average_score}")

# 5. Print Indian players
print(f"Indian players: {', '.join(indian_players)}")
```

```python
# Caesar Cipher Encoder & Decoder
```

```python
message = "Hello Python Learner"
shift = 3

encoded_message = ""

# ENCODING
for char in message:

    # For uppercase letters
    if char.isupper():

        new_char = chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
        encoded_message += new_char

    # For lowercase letters
    elif char.islower():

        new_char = chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
        encoded_message += new_char

    # Keep spaces/symbols unchanged
    else:
        encoded_message += char

print(f"Encoded: {encoded_message}")

# DECODING
decoded_message = ""

for char in encoded_message:

    if char.isupper():

        original_char = chr((ord(char) - ord('A') - shift) % 26 + ord('A'))
        decoded_message += original_char

    elif char.islower():

        original_char = chr((ord(char) - ord('a') - shift) % 26 + ord('a'))
        decoded_message += original_char

    else:
        decoded_message += char

print(f"Decoded: {decoded_message}")

# Shift used
print(f"Shift used: {shift}")
```

```python
# Number Analyser
```

```python
nums = [15, 28, 3, 42, 17, 56, 8, 91, 34, 72, 5, 63, 44, 29, 11]

even_numbers = []
odd_numbers = []
prime_numbers = []
perfect_squares = []

total = 0
count = 0

# ONE loop through the list
for num in nums:

    # Running total and count
    total += num
    count += 1

    # Even or Odd
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)

    # Prime check
    if num > 1:

        is_prime = True

        for i in range(2, num):

            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            prime_numbers.append(num)

    # Perfect square check
    root = int(num ** 0.5)

    if root * root == num:
        perfect_squares.append(num)

# Average
average = round(total / count, 2)

# Print results
print(f"Even: {even_numbers}")
print(f"Odd: {odd_numbers}")
print(f"Primes: {prime_numbers}")
print(f"Perfect squares: {perfect_squares}")
print(f"Average: {average}")
```

```python
# Contact Book Manager
```

```python
contacts = [
    ("Diya Desai", "9876543210", "diya@email.com", "Bhuj"),
    ("Raj Patel", "9988776655", "raj@email.com", "Ahmedabad"),
    ("Amit Shah", "9123456789", "amit@email.com", "Surat"),
    ("Sara Khan", "9876501234", "sara@email.com", "Bhuj"),
    ("Kiran Mehta", "9000011111", "kiran@email.com", "Vadodara")
]

# Variables
valid_phone_count = 0
valid_email_count = 0

all_phones_valid = True
all_emails_valid = True

cities = []

# 1. Print all contacts
count = 1

for name, phone, email, city in contacts:

    print(f"{count}. {name} | {phone} | {city}")

    count += 1

    # Store unique cities
    if city not in cities:
        cities.append(city)

print("--- Bhuj contacts ---")

# 2. Search by city
for name, phone, email, city in contacts:

    if city == "Bhuj":
        print(f"{name} | {phone}")

print("--- Phone validation ---")

# 3. Validate phone numbers
for name, phone, email, city in contacts:

    if len(phone) == 10 and phone.isdigit():
        valid_phone_count += 1
    else:
        all_phones_valid = False

print(f"All phones valid: {all_phones_valid}")

print("--- Email validation ---")

# 4. Validate emails
for name, phone, email, city in contacts:

    if "@" in email and "." in email:
        valid_email_count += 1
    else:
        all_emails_valid = False

print(f"All emails valid: {all_emails_valid}")

print("--- A-M names ---")

# 5. Names starting with A-M
for name, phone, email, city in contacts:

    first_letter = name[0].upper()

    if first_letter >= "A" and first_letter <= "M":
        print(name)

# Sort cities alphabetically
cities.sort()

# 6. Summary
print("--- Summary ---")

print(f"Total: {len(contacts)}")
print(f"Cities: {', '.join(cities)}")
print(f"Valid phones: {valid_phone_count}")
print(f"Valid emails: {valid_email_count}")
```
