```python
#A dictionary stores data as key-value pairs - like a real dictionary where a word (key) maps to its meaning (value):
```

```python
student = {
    "name": "Diya",
    "age": 21,
    "city": "baroda",
    "cgpa": 8.75
}
```

```python
#Accessing values:
```

```python
print(student["name"])        # Diya  ← by key
print(student.get("age"))     # 21
print(student.get("phone"))   # None  ← safe, no KeyError!
print(student.get("phone", "N/A"))  # N/A ← default value
```

```python
#The difference - [] vs .get():
```

```python
# student["phone"]       → KeyError! crashes if key missing
# student.get("phone")   → None, safe
```

```python
#Adding and updating:
```

```python
student["phone"] = "9876543210"   # add new key
student["age"] = 22               # update existing key
student.update({"city": "Surat", "grade": "A"})  # update multiple
```

```python
#Removing:
```

```python
removed = student.pop("grade")    # removes & returns value
del student["phone"]              # removes, no return
student.clear()                   # empties entire dict
```

```python
#The 3 views - keys, values, items:
```

```python
d = {"a": 1, "b": 2, "c": 3}
print(d.keys())    # dict_keys(['a', 'b', 'c'])
print(d.values())  # dict_values([1, 2, 3])
print(d.items())   # dict_items([('a',1),('b',2),('c',3)])

# Convert to list:
print(list(d.keys()))     # ['a', 'b', 'c']
print(list(d.values()))   # [1, 2, 3]
```

```python
#Looping through a dictionary:
```

```python
for key in student:               # loops over keys
    print(key, student[key])

for key, value in student.items():   # cleanest way!
    print(f"{key}: {value}")

for value in student.values():    # just values
    print(value)
```

```python
#Checking membership:
```

```python
print("name" in student)     # True  ← checks keys
print("Diya" in student)     # False ← values not checked
print("Diya" in student.values())  # True
```

```python
#Dictionary from two lists using zip():
```

```python
keys = ["name", "age", "city"]
vals = ["Diya", 21, "Bhuj"]
d = dict(zip(keys, vals))
print(d)   # {'name': 'Diya', 'age': 21, 'city': 'Bhuj'}
```

```python
#Loop through items
```

```python
prices = {
    "Rice": 60,
    "Oil": 180,
    "Sugar": 45,
    "Milk": 60,
    "Bread": 35
}

# 1. Loop through all items
for item, price in prices.items():
    print(f"{item} costs Rs.{price}")

# 2. Most expensive item
most_expensive = max(prices, key=prices.get)
print(f"Most expensive: {most_expensive} (Rs.{prices[most_expensive]})")

# 3. Cheapest item
cheapest = min(prices, key=prices.get)
print(f"Cheapest: {cheapest} (Rs.{prices[cheapest]})")

# 4. Total value
total = sum(prices.values())
print(f"Total: Rs.{total}")

# 5. Items above Rs.50
above_50 = []

for item, price in prices.items():
    if price > 50:
        above_50.append(item)

print("Above Rs.50:", ", ".join(above_50))
```

```python
#Word frequency with dict
```

```python
text = "the cat sat on the mat the cat is fat and the mat is flat"

# Split text into words
words = text.split()

# 1. Count frequency using dictionary
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# 2. Print words sorted by frequency (highest first)
sorted_words = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

for word, count in sorted_words:
    print(f"{word}: {count}")

# 3. Print top 3 most common words
top_3 = [word for word, count in sorted_words[:3]]
print("Top 3:", ", ".join(top_3))

# 4. Print words that appear exactly once
appear_once = []

for word, count in frequency.items():
    if count == 1:
        appear_once.append(word)

print("Appear once:", ", ".join(appear_once))

# 5. Print total unique words
print("Unique words:", len(frequency))
```

```python
#Nested dictionary - student database
```

```python
# Nested dictionary
db = {
    "D001": {
        "name": "Diya",
        "marks": {"maths": 88, "science": 92, "english": 76},
        "city": "Bhuj"
    },
    
    "D002": {
        "name": "Raj",
        "marks": {"maths": 72, "science": 65, "english": 80},
        "city": "Ahmedabad"
    },
    
    "D003": {
        "name": "Amit",
        "marks": {"maths": 95, "science": 98, "english": 92},
        "city": "Surat"
    }
}

# 1. Print student details
for student_id, details in db.items():
    
    total = sum(details["marks"].values())
    avg = total / len(details["marks"])
    
    print(f"{student_id} | {details['name']} | Total: {total} | Avg: {round(avg, 2)} | City: {details['city']}")

# 2. Find highest maths mark
top_student = ""
highest_maths = 0

for details in db.values():
    if details["marks"]["maths"] > highest_maths:
        highest_maths = details["marks"]["maths"]
        top_student = details["name"]

print(f"Top maths: {top_student} ({highest_maths})")

# 3. Find city with most students
city_count = {}

for details in db.values():
    city = details["city"]
    
    if city in city_count:
        city_count[city] += 1
    else:
        city_count[city] = 1

max_students = max(city_count.values())

top_cities = []

for city, count in city_count.items():
    if count == max_students:
        top_cities.append(city)

if len(top_cities) == len(city_count):
    print("Top city: Each city has 1 student")
else:
    print("Top city:", ", ".join(top_cities))

# 4. Add passed field
for student_id, details in db.items():
    
    total = sum(details["marks"].values())
    avg = total / len(details["marks"])
    
    details["passed"] = avg >= 70
    
    print(f"{student_id} {details['name']}: passed={details['passed']}")
```

```python
#Inventory management system
```

```python
# Inventory database
inventory = {
    "Rice": {"price": 60, "stock": 50, "category": "Grocery"},
    "Oil": {"price": 180, "stock": 10, "category": "Grocery"},
    "Shampoo": {"price": 250, "stock": 5, "category": "Personal"},
    "Bread": {"price": 35, "stock": 0, "category": "Bakery"},
    "Soap": {"price": 45, "stock": 30, "category": "Personal"},
    "Sugar": {"price": 45, "stock": 25, "category": "Grocery"}
}

# Customer order
order = {
    "Rice": 3,
    "Oil": 15,
    "Bread": 2,
    "Soap": 5,
    "Chips": 1
}

receipt = []
total_bill = 0

# Process order
for item, qty in order.items():

    # Item not found
    if item not in inventory:
        print(f"❌ {item}: not found")

    else:
        stock = inventory[item]["stock"]
        price = inventory[item]["price"]

        # Out of stock
        if stock == 0:
            print(f"❌ {item}: out of stock")

        # Ordered quantity greater than stock
        elif qty > stock:
            print(f"⚠️ {item}: only {stock} available, ordered {stock}")

            amount = stock * price
            receipt.append(f"{item} x{stock} = Rs.{amount}")

            total_bill += amount

            inventory[item]["stock"] = 0

        # Normal order
        else:
            amount = qty * price

            print(f"✅ {item} x{qty} = Rs.{amount}")

            receipt.append(f"{item} x{qty} = Rs.{amount}")

            total_bill += amount

            inventory[item]["stock"] -= qty

# Receipt
print("\n--- Receipt---")

for line in receipt:
    print(line)

print(f"Total: Rs.{total_bill}")

# Group by category
print("\n--- By Category ---")

categories = {}

for item, details in inventory.items():

    category = details["category"]

    if category not in categories:
        categories[category] = []

    categories[category].append(item)

for category, items in categories.items():
    print(f"{category}: {', '.join(items)}")

# Low stock alert
print("\n--- Low Stock (< 10) ---")

for item, details in inventory.items():

    if details["stock"] < 10:
        print(f"{item}: {details['stock']} left")

# Total inventory value
print("\n--- Inventory Value ---")

inventory_value = 0

for details in inventory.values():
    inventory_value += details["price"] * details["stock"]

print(f"Total value: Rs.{inventory_value}")
```
