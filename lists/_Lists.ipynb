```python
#Indexing - access by position:
```

```python
fruits = ["apple", "mango", "banana", "grape"]
print(fruits[0])    # apple  ← first
print(fruits[1])    # mango
print(fruits[-1])   # grape  ← last
print(fruits[-2])   # banana
```

```python
#Slicing - extract a portion:
```

```python
print(fruits[1:3])   # ['mango', 'banana']
print(fruits[:2])    # ['apple', 'mango']
print(fruits[2:])    # ['banana', 'grape']
print(fruits[::-1])  # ['grape', 'banana', 'mango', 'apple'] ← reversed!
```

```python
#Modifying lists - lists are mutable (changeable):
```

```python
fruits[0] = "pineapple"   # change item
print(fruits)   # ['pineapple', 'mango', 'banana', 'grape']
```

```python
#Adding items:
```

```python
fruits.append("kiwi")        # add to end
fruits.insert(1, "cherry")   # insert at index 1
fruits += ["lemon", "lime"]  # join two lists
```

```python
#Removing items:
```

```python
fruits.remove("mango")   # removes first match by value
fruits.pop()             # removes and returns last item
fruits.pop(0)            # removes and returns item at index 0
del fruits[2]            # delete by index, no return
```

```python
#Useful list operations:
```

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6]
print(len(nums))         # 8
print(sum(nums))         # 31
print(min(nums))         # 1
print(max(nums))         # 9
print(sorted(nums))      # [1, 1, 2, 3, 4, 5, 6, 9] ← new list
nums.sort()              # sorts in place
nums.reverse()           # reverses in place
print(nums.count(1))     # 2 ← how many times 1 appears
print(nums.index(5))     # finds index of first 5
print(3 in nums)         # True ← membership test
```

```python
#List analyser
```

```python
temps = [28, 35, 22, 40, 18, 33, 29, 41, 25, 37]

# 1. Average temperature
total = 0

for temp in temps:
    total += temp

average = total / len(temps)
print(round(average, 2))

# 2. Number of days above 30°C
count_above_30 = 0

for temp in temps:
    if temp > 30:
        count_above_30 += 1

print(count_above_30)

# 3. Sorted temperatures
sorted_temps = sorted(temps)
print(sorted_temps)

# 4. Reversed original list
reversed_temps = temps[::-1]
print(reversed_temps)

# 5. Count of 28
print(temps.count(28))
```

```python
#Second largest without sort
```

```python
nums = [15, 42, 8, 97, 23, 56, 97, 34]

largest = nums[0]
second_largest = nums[0]
largest_index = 0

# Find largest number and its index
for i in range(len(nums)):
    if nums[i] > largest:
        largest = nums[i]
        largest_index = i

# Find second largest (different from largest)
second_largest = float('-inf')

for num in nums:
    if num != largest and num > second_largest:
        second_largest = num

print("Largest:", largest)
print("Second largest:", second_largest)
print("Index of largest:", largest_index)
```

```python
#Student marks manager
```

```python
students = ["Diya", "Raj", "Amit", "Sara", "Kiran"]
marks = [88, 72, 95, 61, 79]

# 1. Print each student with marks
for i in range(len(students)):
    print(f"{students[i]}: {marks[i]}")

# 2. Find topper
highest_marks = max(marks)
topper_index = marks.index(highest_marks)

print(f"Topper: {students[topper_index]} ({highest_marks})")

# 3. Find student needing improvement
lowest_marks = min(marks)
lowest_index = marks.index(lowest_marks)

print(f"Needs improvement: {students[lowest_index]} ({lowest_marks})")

# 4. Class average
average = sum(marks) / len(marks)
print(f"Class average: {round(average, 2)}")

# 5. Students above average
above_average = []

for i in range(len(marks)):
    if marks[i] > average:
        above_average.append(students[i])

print("Above average:", ", ".join(above_average))

# 6. Add new student
students.append("Meera")
marks.append(84)

# 7. Updated total students count
print("Total students:", len(students))
```
