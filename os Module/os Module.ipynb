```python
#Virtual file system simulator
```

```python
import fnmatch

# Virtual File System
vfs = {
    "/": ["home", "etc", "usr"],

    "/home": ["diya", "raj"],

    "/home/diya": [
        "projects",
        "documents",
        "script.py",
        "notes.txt"
    ],

    "/home/diya/projects": [
        "day41.py",
        "day40.py",
        "utils.py"
    ],

    "/home/raj": ["work.py"],

    "/etc": ["config.txt", "hosts"],

    "/usr": ["bin", "lib"]
}


# 1. List contents
def vls(path):
    return vfs.get(path, "Path not found")


# 2. Check existence
def vexists(path):

    if path in vfs:
        return True

    parent = "/".join(path.rstrip("/").split("/")[:-1])

    if parent == "":
        parent = "/"

    name = path.split("/")[-1]

    return name in vfs.get(parent, [])


# 3. Check if file
def vis_file(path):

    name = path.split("/")[-1]

    return "." in name


# 4. Find matching files
def vfind(root, pattern):

    matches = []

    for path, contents in vfs.items():

        if path.startswith(root):

            for item in contents:

                if fnmatch.fnmatch(item, pattern):
                    matches.append(path.rstrip("/") + "/" + item)

    return matches


# 5. Print tree
def vtree(path, indent=0):

    if indent == 0:
        print(path)

    contents = vfs.get(path, [])

    for item in contents:

        full_path = path.rstrip("/") + "/" + item

        prefix = "│   " * indent + "├── "

        print(prefix + item)

        if full_path in vfs:
            vtree(full_path, indent + 1)


# TESTING

print("/home/diya contents:",
      vls("/home/diya"))

print()

print("/home/diya/projects exists:",
      vexists("/home/diya/projects"))

print("/home/fake exists:",
      vexists("/home/fake"))

print()

print("script.py is file:",
      vis_file("script.py"))

print("projects is file:",
      vis_file("projects"))

print()

print("Python files:",
      vfind("/", "*.py"))

print()

vtree("/")
```

```python
#Path analyser and organiser
```

```python
import os
from collections import defaultdict

files = [
    "/home/diya/photo.jpg",
    "/home/diya/report.pdf",
    "/home/diya/script.py",
    "/home/diya/data.csv",
    "/home/diya/notes.txt",
    "/home/diya/video.mp4",
    "/home/diya/backup.py",
    "/home/diya/image.png",
    "/home/diya/music.mp3",
    "/home/diya/spreadsheet.xlsx"
]

# Categories
categories = {
    "images": ["jpg", "png", "gif"],
    "docs": ["pdf", "txt", "docx"],
    "code": ["py", "js", "java"],
    "data": ["csv", "xlsx", "json"],
    "media": ["mp4", "mp3"]
}


# 1. Group by extension
extension_count = defaultdict(int)

# 2. Group by category
grouped_categories = defaultdict(list)

# 3. Suggested folder structure
folder_structure = defaultdict(list)

# 5. Renamed files
renamed_files = []

for file in files:

    filename = os.path.basename(file)

    name, ext = os.path.splitext(filename)

    clean_ext = ext[1:]   # remove dot

    # Count extensions
    extension_count[ext] += 1

    # Group by category
    for category, exts in categories.items():

        if clean_ext in exts:

            grouped_categories[category].append(filename)

            folder_structure[category].append(filename)

    # Rename simulation
    renamed_files.append("2025_" + filename)


# 4. Stats
total_files = len(files)

largest_category = max(
    grouped_categories,
    key=lambda x: len(grouped_categories[x])
)

largest_count = len(grouped_categories[largest_category])


# OUTPUT

print("By extension:")
print(dict(extension_count))

print()

for category, file_list in grouped_categories.items():
    print(f"{category.title()}: {file_list}")

print()

print("Suggested Folder Structure:")
print(dict(folder_structure))

print()

print(
    f"Total: {total_files} files | "
    f"Categories: {len(grouped_categories)} | "
    f"Largest: {largest_category} "
    f"({largest_count} files)"
)

print()

print("Renamed Files:")
print(", ".join(renamed_files))
```

```python
import os
import platform

# ---------------------------------------------------
# 1. Print OS Type
# ---------------------------------------------------

print("OS Type:", platform.system(), f"({os.name})")


# ---------------------------------------------------
# 2. Path separator and line separator
# ---------------------------------------------------

print("Path separator:", os.sep)

if os.linesep == "\n":
    print(r"Line separator: LF (\n)")
elif os.linesep == "\r\n":
    print(r"Line separator: CRLF (\r\n)")
else:
    print("Unknown line separator")


# ---------------------------------------------------
# 3. PATH variable info
# ---------------------------------------------------

path_var = os.environ.get("PATH", "")

path_dirs = path_var.split(os.pathsep)

print("PATH directories:", len(path_dirs))


# ---------------------------------------------------
# 4. Safe environment reader
# ---------------------------------------------------

def safe_env(key, default=None, validator=None):

    value = os.environ.get(key, default)

    # Required variable check
    if value is None:
        raise ValueError(
            f"{key} environment variable is required"
        )

    # Validation
    if validator:
        value = validator(value)

    return value


# ---------------------------------------------------
# Validators
# ---------------------------------------------------

def validate_port(value):

    port = int(value)

    if 1 <= port <= 65535:
        return port

    raise ValueError("Port must be between 1-65535")


def validate_env(value):

    allowed = [
        "development",
        "staging",
        "production"
    ]

    if value not in allowed:
        raise ValueError(
            f"APP_ENV must be one of {allowed}"
        )

    return value


def to_bool(value):

    return value.lower() in [
        "true",
        "1",
        "yes",
        "on"
    ]


# ---------------------------------------------------
# 5. Simulate environment variables
# ---------------------------------------------------

os.environ["DB_HOST"] = "localhost"
os.environ["DB_PORT"] = "5432"
os.environ["APP_ENV"] = "development"
os.environ["DEBUG"] = "false"

# SECRET_KEY intentionally NOT set


# ---------------------------------------------------
# Load Config
# ---------------------------------------------------

print("\nConfig loaded:")

db_host = safe_env("DB_HOST", "localhost")

db_port = safe_env(
    "DB_PORT",
    "5432",
    validate_port
)

app_env = safe_env(
    "APP_ENV",
    "development",
    validate_env
)

debug = safe_env(
    "DEBUG",
    "false",
    to_bool
)

print("  DB_HOST:", db_host)
print("  DB_PORT:", db_port)
print("  APP_ENV:", app_env)
print("  DEBUG:", debug)


# ---------------------------------------------------
# SECRET_KEY required
# ---------------------------------------------------

try:

    secret_key = safe_env("SECRET_KEY")

except ValueError as e:

    print("\nSecretKey error:", e)


# ---------------------------------------------------
# Set SECRET_KEY and reload
# ---------------------------------------------------

os.environ["SECRET_KEY"] = "s3cr3t_k3y_123"

secret_key = safe_env("SECRET_KEY")

print("\nWith SECRET_KEY set:")
print("  SECRET_KEY:", secret_key)
```

```python
import fnmatch
import copy


class FileSystemManager:

    def __init__(self, root_dir):

        self.root = root_dir

        # Virtual filesystem
        # Directories -> dict
        # Files -> string content
        self.fs = {
            root_dir: {}
        }

    # -----------------------------------------
    # Internal helper
    # -----------------------------------------

    def _get_node(self, path):

        parts = [p for p in path.strip("/").split("/")]

        node = self.fs[self.root]

        if parts and parts[0] == self.root:
            parts = parts[1:]

        for part in parts:

            if part not in node:
                return None

            node = node[part]

        return node

    # -----------------------------------------
    # Create Directory
    # -----------------------------------------

    def create_dir(self, path):

        parts = [p for p in path.strip("/").split("/")]

        node = self.fs[self.root]

        if parts and parts[0] == self.root:
            parts = parts[1:]

        for part in parts:

            if part not in node:
                node[part] = {}

            node = node[part]

    # -----------------------------------------
    # Create File
    # -----------------------------------------

    def create_file(self, path, content=""):

        parts = [p for p in path.strip("/").split("/")]

        filename = parts.pop()

        node = self.fs[self.root]

        if parts and parts[0] == self.root:
            parts = parts[1:]

        for part in parts:

            if part not in node:
                node[part] = {}

            node = node[part]

        node[filename] = content

    # -----------------------------------------
    # Read File
    # -----------------------------------------

    def read_file(self, path):

        node = self._get_node(path)

        if isinstance(node, str):
            return node

        return "File not found"

    # -----------------------------------------
    # Delete
    # -----------------------------------------

    def delete(self, path):

        parts = [p for p in path.strip("/").split("/")]

        name = parts.pop()

        node = self.fs[self.root]

        if parts and parts[0] == self.root:
            parts = parts[1:]

        for part in parts:
            node = node[part]

        if name in node:
            del node[name]

    # -----------------------------------------
    # Move / Rename
    # -----------------------------------------

    def move(self, src, dst):

        content = self._get_node(src)

        self.copy(src, dst)

        self.delete(src)

    # -----------------------------------------
    # Copy File
    # -----------------------------------------

    def copy(self, src, dst):

        content = self._get_node(src)

        self.create_file(dst, copy.deepcopy(content))

    # -----------------------------------------
    # Search
    # -----------------------------------------

    def search(self, pattern):

        matches = []

        def walk(node, current_path):

            for name, value in node.items():

                full_path = f"{current_path}/{name}"

                if isinstance(value, str):

                    if fnmatch.fnmatch(name, pattern):
                        matches.append(full_path)

                elif isinstance(value, dict):

                    walk(value, full_path)

        walk(self.fs[self.root], self.root)

        return matches

    # -----------------------------------------
    # Size
    # -----------------------------------------

    def size_of(self, path=None):

        if path is None:
            node = self.fs[self.root]
        else:
            node = self._get_node(path)

        total = 0

        def calc(node):

            nonlocal total

            if isinstance(node, str):

                total += len(node)

            elif isinstance(node, dict):

                for value in node.values():
                    calc(value)

        calc(node)

        return total

    # -----------------------------------------
    # Tree
    # -----------------------------------------

    def tree(self):

        print(self.root + "/")

        def walk(node, indent=0):

            for name, value in node.items():

                prefix = "│   " * indent + "├── "

                if isinstance(value, dict):

                    print(prefix + name + "/")

                    walk(value, indent + 1)

                else:

                    print(prefix + name)

        walk(self.fs[self.root])

    # -----------------------------------------
    # Stats
    # -----------------------------------------

    def stats(self):

        files = 0
        dirs = 0

        def count(node):

            nonlocal files, dirs

            for value in node.values():

                if isinstance(value, str):
                    files += 1

                elif isinstance(value, dict):
                    dirs += 1
                    count(value)

        count(self.fs[self.root])

        total_size = self.size_of()

        print(
            f"Stats: {files} files, "
            f"{dirs} dirs, "
            f"{total_size} bytes total"
        )


# =====================================================
# TESTING
# =====================================================

fsm = FileSystemManager("project")

# Create directories
fsm.create_dir("project/src")
fsm.create_dir("project/tests")

# Create files
fsm.create_file(
    "project/README.md",
    "# My Python Project"
)

fsm.create_file(
    "project/src/main.py",
    "print('Hello World')"
)

fsm.create_file(
    "project/src/utils.py",
    "def add(a,b): return a+b"
)

fsm.create_file(
    "project/setup.py",
    "setup configuration"
)

fsm.create_file(
    "project/tests/test_main.py",
    "def test_main(): pass"
)

print("Created project structure")

# Read file
print(
    "README content:",
    fsm.read_file("project/README.md")
)

# Move file
fsm.move(
    "project/setup.py",
    "project/src/setup.py"
)

print("Moved setup.py to src/setup.py")

# Search files
print(
    "Python files:",
    fsm.search("*.py")
)

# Size
print(
    "Total size:",
    fsm.size_of(),
    "bytes"
)

# Tree
fsm.tree()

# Stats
fsm.stats()
```
