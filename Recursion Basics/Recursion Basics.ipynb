What is recursion?

A function is recursive when it calls itself. Every recursive solution needs:

Base case — when to stop (no more recursion)
Recursive case — call itself with a smaller problem

```python
#The simplest example — countdown:
```

```python
def countdown(n):
    if n <= 0:          # base case
        print("Go!")
        return
    print(n)
    countdown(n - 1)    # recursive case — smaller problem

countdown(5)
# 5 4 3 2 1 Go!
```

```python
#Factorial — the classic:
```

```python
# Iterative:
def factorial_iter(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# Recursive — much cleaner:
def factorial(n):
    if n <= 1:           # base case: 0! = 1! = 1
        return 1
    return n * factorial(n - 1)   # n! = n × (n-1)!

# How it works:
# factorial(4)
# = 4 × factorial(3)
# = 4 × 3 × factorial(2)
# = 4 × 3 × 2 × factorial(1)
# = 4 × 3 × 2 × 1
# = 24
```

```python
#The call stack — how Python tracks recursion:
```

```python
def factorial(n):
    print(f"factorial({n}) called")
    if n <= 1:
        print(f"returning 1")
        return 1
    result = n * factorial(n - 1)
    print(f"factorial({n}) returning {result}")
    return result

factorial(4)
# factorial(4) called
# factorial(3) called
# factorial(2) called
# factorial(1) called
# returning 1
# factorial(2) returning 2
# factorial(3) returning 6
# factorial(4) returning 24
```

```python
#Fibonacci — two recursive calls:
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# fib(5) = fib(4) + fib(3)
# Very slow for large n — exponential time!
```

```python
#Memoization — fix the slowness:
```

```python
# With a cache dict:
memo = {}
def fib_memo(n):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]

# Or use Python's built-in cache:
from functools import lru_cache

@lru_cache(maxsize=None)
def fib_fast(n):
    if n <= 1:
        return n
    return fib_fast(n-1) + fib_fast(n-2)
```

```python
#Power function:
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    if exp % 2 == 0:                        # fast: even exponent
        return power(base, exp//2) ** 2
    return base * power(base, exp - 1)
```

```python
#Binary search — recursion vs iteration:
```

```python
def binary_search(arr, target, lo=0, hi=None):
    if hi is None:
        hi = len(arr) - 1
    if lo > hi:
        return -1                    # base case: not found
    mid = (lo + hi) // 2
    if arr[mid] == target:
        return mid                   # base case: found
    elif arr[mid] < target:
        return binary_search(arr, target, mid+1, hi)
    else:
        return binary_search(arr, target, lo, mid-1)
```

```python
#Classic recursion problems
```

```python
# 1. factorial(n) → n!
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


# 2. sum_to_n(n) → sum of 1+2+...+n
def sum_to_n(n):
    if n == 0:
        return 0
    return n + sum_to_n(n - 1)


# 3. power(base, exp) → base^exp
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)


# 4. reverse_string(s) → reverse s recursively
def reverse_string(s):
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]


# 5. count_down(n)
def count_down(n):
    if n == 0:
        print("Blast off!")
        return
    print(n)
    count_down(n - 1)


# Test
print(factorial(6))
print(sum_to_n(10))
print(power(2, 10))
print(reverse_string("Python"))
count_down(5)
```

```python
#Fibonacci with memoization
```

```python
from functools import lru_cache

# 1. Naive recursive Fibonacci
def fib_slow(n):
    if n <= 1:
        return n
    return fib_slow(n - 1) + fib_slow(n - 2)


# 2. Fibonacci with manual memoization
def fib_memo(n, memo=None):
    if memo is None:
        memo = {}

    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib_memo(n - 1, memo) + fib_memo(n - 2, memo)
    return memo[n]


# 3. Fibonacci with lru_cache
@lru_cache(maxsize=None)
def fib_fast(n):
    if n <= 1:
        return n
    return fib_fast(n - 1) + fib_fast(n - 2)


# Test fib(10)
print("fib_slow(10):", fib_slow(10))
print("fib_memo(10):", fib_memo(10))
print("fib_fast(10):", fib_fast(10))


# First 15 Fibonacci numbers
first_15 = [fib_fast(i) for i in range(15)]
print("First 15:", first_15)


# First Fibonacci number greater than 1000
n = 0
while True:
    fib_num = fib_fast(n)
    if fib_num > 1000:
        print("First fib > 1000:", fib_num)
        break
    n += 1


# Check if a number is Fibonacci
def is_fibonacci(num):
    n = 0
    while fib_fast(n) < num:
        n += 1
    return fib_fast(n) == num


print("144 is Fibonacci:", is_fibonacci(144))
```

```python
#Recursive data structures
```

```python
# 1. Flatten a deeply nested list
def flatten_deep(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(flatten_deep(item))
        else:
            result.append(item)

    return result


# 2. Sum all numbers at any nesting level
def deep_sum(lst):
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += deep_sum(item)
        else:
            total += item

    return total


# 3. Find depth of a nested dict tree
def tree_depth(tree):
    if not tree:
        return 0

    return 1 + max(tree_depth(subtree) for subtree in tree.values())


# 4. Count leaf nodes (empty dicts)
def count_leaves(tree):
    if not tree:
        return 1

    total = 0
    for subtree in tree.values():
        total += count_leaves(subtree)

    return total


# 5. List all root-to-leaf paths
def tree_paths(tree, path=""):
    paths = []

    for node, subtree in tree.items():
        current_path = node if path == "" else path + "→" + node

        if not subtree:  # leaf node
            paths.append(current_path)
        else:
            paths.extend(tree_paths(subtree, current_path))

    return paths


# Test Data
nested = [1, [2, [3, [4, 5]]], 6]

tree = {
    "A": {
        "B": {
            "D": {}
        },
        "C": {
            "E": {},
            "F": {}
        }
    }
}

# Tests
print(flatten_deep(nested))
print(deep_sum(nested))
print("Tree depth:", tree_depth(tree))
print("Leaf count:", count_leaves(tree))
print("Paths:", ", ".join(tree_paths(tree)))
```

```python
#Merge sort and Quick sort
```

```python
import random
import time

# Global counter for merge sort comparisons
comparisons = 0


# Merge helper
def merge(left, right):
    global comparisons

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        comparisons += 1

        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


# Merge Sort
def merge_sort(arr, depth=0, show_steps=False):
    if show_steps:
        print("  " * depth + f"Sorting: {arr}")

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid], depth + 1, show_steps)
    right = merge_sort(arr[mid:], depth + 1, show_steps)

    return merge(left, right)


# Quick Sort
def quick_sort(arr, depth=0, show_steps=False):
    if show_steps:
        print("  " * depth + f"Sorting: {arr}")

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return (
        quick_sort(left, depth + 1, show_steps)
        + middle
        + quick_sort(right, depth + 1, show_steps)
    )


# Test array
arr = [64, 34, 25, 12, 22, 11, 90]

print("Merge sort steps:")
comparisons = 0
merge_result = merge_sort(arr, show_steps=True)

quick_result = quick_sort(arr)
builtin_result = sorted(arr)

print("\nMerge sort:", merge_result)
print("Quick sort:", quick_result)
print("Built-in:", builtin_result)
print("All match:", merge_result == quick_result == builtin_result)
print("Merge sort comparisons:", comparisons)


# --------------------------------------------------
# Timing on 1000 random numbers
# --------------------------------------------------

data = [random.randint(1, 10000) for _ in range(1000)]

start = time.perf_counter()
merge_sort(data.copy())
merge_time = time.perf_counter() - start

start = time.perf_counter()
quick_sort(data.copy())
quick_time = time.perf_counter() - start

start = time.perf_counter()
sorted(data.copy())
builtin_time = time.perf_counter() - start

print("\nTiming on 1000 random numbers")
print(f"Merge Sort : {merge_time:.6f} sec")
print(f"Quick Sort : {quick_time:.6f} sec")
print(f"Built-in   : {builtin_time:.6f} sec")
```

```python
#Tower of Hanoi and permutations
```

```python
# --------------------------------------------------
# 1. Tower of Hanoi
# --------------------------------------------------

def hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return 1

    moves = 0

    moves += hanoi(n - 1, source, auxiliary, target)

    print(f"Move disk {n} from {source} to {target}")
    moves += 1

    moves += hanoi(n - 1, auxiliary, target, source)

    return moves


# --------------------------------------------------
# 2. Generate all permutations
# --------------------------------------------------

def permutations(lst):
    if len(lst) <= 1:
        return [lst]

    result = []

    for i in range(len(lst)):
        current = lst[i]
        remaining = lst[:i] + lst[i + 1:]

        for perm in permutations(remaining):
            result.append([current] + perm)

    return result


# --------------------------------------------------
# 3. Generate all r-length combinations
# --------------------------------------------------

def combinations(lst, r):
    if r == 0:
        return [[]]

    if len(lst) < r:
        return []

    if len(lst) == 0:
        return []

    first = lst[0]

    with_first = [
        [first] + combo
        for combo in combinations(lst[1:], r - 1)
    ]

    without_first = combinations(lst[1:], r)

    return with_first + without_first


# --------------------------------------------------
# 4. Power Set (all subsets)
# --------------------------------------------------

def power_set(lst):
    if not lst:
        return [[]]

    first = lst[0]
    rest_power = power_set(lst[1:])

    with_first = [[first] + subset for subset in rest_power]

    return rest_power + with_first


# --------------------------------------------------
# Tests
# --------------------------------------------------

print("Tower of Hanoi (3 disks):")
total_moves = hanoi(3, "A", "C", "B")

print(f"\nTotal moves: {total_moves} (2^3 - 1 = {2**3 - 1})")


# Permutations
perms = permutations([1, 2, 3])

print("\nPermutations of [1,2,3]:", len(perms))
print(perms)


# Combinations
combos = combinations([1, 2, 3, 4], 2)

print("\nCombinations(4,2):", len(combos))
print(combos)


# Power Set
ps = power_set([1, 2, 3])

print("\nPower set of [1,2,3]:", len(ps), "subsets")
print(ps)
```
