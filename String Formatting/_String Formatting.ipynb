```python
#3 ways to format strings:
```

```python
#1. % formatting (old style — avoid in new code):
```

```python
name = "Diya"
age = 21
print("Hello %s, you are %d years old" % (name, age))
print("Price: %.2f" % 99.999)   # Price: 100.00
```

```python
#2. format() method (Python 3+):
```

```python
print("Hello {}, you are {} years old".format(name, age))
print("Hello {0}, {0} again, {1}".format(name, age))   # reuse by index
print("Hello {n}, age {a}".format(n=name, a=age))      # by name
```

```python
#3. f-strings (Python 3.6+ — use this always):
```

```python
print(f"Hello {name}, you are {age} years old")
print(f"Next year: {age + 1}")          # expressions work!
print(f"Upper: {name.upper()}")         # method calls work!
print(f"Type: {type(age).__name__}")    # anything Python works!
```

```python
#Format specifiers — the power of : inside {}:
```

```python
# Numbers:
pi = 3.14159265
print(f"{pi:.2f}")      # 3.14  ← 2 decimal places
print(f"{pi:.4f}")      # 3.1416
print(f"{pi:10.2f}")    # '      3.14'  ← width 10, right-aligned
print(f"{pi:<10.2f}")   # '3.14      '  ← left-aligned
print(f"{pi:^10.2f}")   # '   3.14   '  ← center-aligned
print(f"{pi:010.2f}")   # '0000003.14'  ← zero-padded

# Integers:
n = 1234567
print(f"{n:,}")          # 1,234,567  ← comma separator
print(f"{n:_}")          # 1_234_567  ← underscore separator
print(f"{n:08d}")        # 01234567   ← zero-padded to 8 digits
print(f"{n:b}")          # binary
print(f"{n:x}")          # hex (lowercase)
print(f"{n:X}")          # hex (uppercase)
print(f"{n:o}")          # octal

# Strings:
text = "Python"
print(f"{text:>10}")     # '    Python'  ← right-align in 10
print(f"{text:<10}")     # 'Python    '  ← left-align
print(f"{text:^10}")     # '  Python  '  ← center
print(f"{text:*^10}")    # **Python**   ← center with * fill
print(f"{text:->10}")    # ----Python   ← right with - fill
```

```python
pi = 3.14159265
population = 1428627663
name = "Python"
ratio = 0.7834

# 1. pi to 2 decimal places
print(f"{pi:.2f}")

# 2. pi to 4 decimal places, width 12, right-aligned
print(f"{pi:12.4f}")

# 3. population with comma separator
print(f"{population:,}")

# 4. name centered in 20 chars with * fill
print(f"{name:*^20}")

# 5. ratio as percentage with 1 decimal place
print(f"{ratio:.1%}")

# 6. pi in scientific notation with 2 decimal places
print(f"{pi:.2e}")
```

```python
#Receipt formatter
```

```python
items = [
    ("Rice", 60, 3),
    ("Oil", 180, 2),
    ("Sugar", 45, 5),
    ("Milk", 60, 4),
    ("Bread", 35, 2)
]

grand_total = 0

print("-" * 32)
print(f"{'Item':<10}{'Qty':>6}{'Price':>10}{'Total':>10}")
print("-" * 32)

for name, price, qty in items:
    total = price * qty
    grand_total += total

    print(f"{name:<10}{qty:>6}{price:>10.2f}{total:>10.2f}")

print("-" * 32)
print(f"{'TOTAL':<22}{grand_total:>10.2f}")
print("-" * 32)
```

```python
#Table generator
```

```python
def print_table(headers, rows, col_widths):
    # Header row
    header_line = " | ".join(
        f"{header:<{width}}"
        for header, width in zip(headers, col_widths)
    )
    print(header_line)

    # Separator line
    separator = "-+-".join("-" * width for width in col_widths)
    print(separator)

    # Data rows
    for row in rows:
        formatted_row = []

        for value, width in zip(row, col_widths):
            # Numbers right-aligned, text left-aligned
            if isinstance(value, (int, float)):
                formatted_row.append(f"{value:>{width}}")
            else:
                formatted_row.append(f"{value:<{width}}")

        print(" | ".join(formatted_row))


# ---------------- STUDENT TABLE ----------------

headers = ["Name", "Score", "Grade", "Status"]

rows = [
    ["Diya", 88, "A", "Pass"],
    ["Raj", 72, "B", "Pass"],
    ["Amit", 95, "A+", "Pass"],
    ["Sara", 61, "C", "Pass"],
    ["Kiran", 39, "F", "Fail"]
]

col_widths = [10, 7, 7, 8]

print_table(headers, rows, col_widths)

print()


# ---------------- NUMBER BASE TABLE ----------------

headers2 = ["Dec", "Bin", "Oct", "Hex"]

rows2 = [
    [n, bin(n)[2:], oct(n)[2:], hex(n)[2:].upper()]
    for n in range(1, 17)
]

col_widths2 = [4, 10, 4, 4]

print_table(headers2, rows2, col_widths2)
```

```python
#Progress bar and dashboard
```

```python
def progress_bar(current, total, width=20, fill="█", empty="░"):
    percent = (current / total) * 100
    filled = int((current / total) * width)
    empty_space = width - filled

    bar = fill * filled + empty * empty_space

    print(f"Progress: {bar} {percent:.1f}% ({current}/{total})")


def dashboard(title, metrics):
    # Format metric lines
    lines = []

    for key, value in metrics.items():
        if isinstance(value, int):
            value = f"{value:,}"

        line = f"{key:<10} : {value:>7}"
        lines.append(line)

    # Find required box width
    content_width = max(len(title), *(len(line) for line in lines))
    box_width = content_width + 4

    # Top border
    print("╔" + "═" * box_width + "╗")

    # Title
    print(f"║ {title:^{content_width}} ║")

    # Middle border
    print("╠" + "═" * box_width + "╣")

    # Metrics
    for line in lines:
        print(f"║ {line:<{content_width}} ║")

    # Bottom border
    print("╚" + "═" * box_width + "╝")


# ---------------- TESTS ----------------

progress_bar(40, 100)
progress_bar(75, 100)
progress_bar(100, 100)

print()

dashboard(
    "Sales Dashboard",
    {
        "Revenue": 1234,
        "Customers": 89,
        "Orders": 156,
        "Avg Sale": 789.5
    }
)
```

```python
#Report card formatter
```

```python
students = [
    {
        "name": "Diya Desai",
        "id": "S001",
        "marks": {
            "Maths": 88,
            "Science": 92,
            "English": 76,
            "CS": 95
        }
    },
    {
        "name": "Raj Patel",
        "id": "S002",
        "marks": {
            "Maths": 72,
            "Science": 65,
            "English": 80,
            "CS": 70
        }
    },
    {
        "name": "Amit Shah",
        "id": "S003",
        "marks": {
            "Maths": 95,
            "Science": 98,
            "English": 92,
            "CS": 97
        }
    }
]


# ---------------- GRADE FUNCTION ----------------

def get_grade(score):
    if score >= 95:
        return "A+"
    elif score >= 85:
        return "A"
    elif score >= 75:
        return "B+"
    elif score >= 65:
        return "B"
    elif score >= 50:
        return "C"
    else:
        return "F"


# ---------------- PREPARE RANKS ----------------

totals = []

for student in students:
    total = sum(student["marks"].values())
    totals.append((student["id"], total))

# Sort descending by total
totals.sort(key=lambda x: x[1], reverse=True)

rank_map = {}

for rank, (student_id, total) in enumerate(totals, start=1):
    rank_map[student_id] = rank


# ---------------- REPORT CARD FUNCTION ----------------

def print_report_card(student):
    width = 37

    name = student["name"]
    sid = student["id"]
    marks = student["marks"]

    total = sum(marks.values())
    avg = total / len(marks)

    result = "Pass" if avg >= 40 else "Fail"
    rank = rank_map[sid]

    # Borders
    print("┌" + "─" * width + "┐")
    print(f"│ {'STUDENT REPORT CARD':^{width}} │")
    print("├" + "─" * width + "┤")

    # Student info
    print(f"│ {'Name':<5}: {name:<27} │")
    print(f"│ {'ID':<5}: {sid:<27} │")

    print("├" + "─" * width + "┤")

    # Header
    print(f"│ {'Subject':<12}{'Score':>7}{'Grade':>9} │")

    # Subject rows
    for subject, score in marks.items():
        grade = get_grade(score)

        print(
            f"│ {subject:<12}{score:>7}{grade:>9} │"
        )

    print("├" + "─" * width + "┤")

    # Summary
    summary = f"Total: {total}/400  Avg: {avg:.2f}%"
    print(f"│ {summary:<35} │")

    result_line = f"Result: {result}"
    print(f"│ {result_line:<35} │")

    rank_line = f"Rank: {rank}"
    print(f"│ {rank_line:<35} │")

    print("└" + "─" * width + "┘")


# ---------------- PRINT ALL REPORT CARDS ----------------

for student in students:
    print_report_card(student)
    print()
```
