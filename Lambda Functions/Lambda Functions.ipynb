```python
#Basic syntax:
```

```python
# Regular function:
def square(x):
    return x ** 2

# Lambda equivalent:
square = lambda x: x ** 2

print(square(5))  
```

```python
#Anatomy of a lambda:
```

lambda  parameters  :  expression
lambda     x, y     :   x + y

Key rules:

Only ONE expression (no statements, no if blocks, no loops)
Always returns the expression result (implicit return)
Can have zero, one, or many parameters

greet = lambda: "Hello!"          # zero params
double = lambda x: x * 2          # one param
add = lambda x, y: x + y          # two params
clamp = lambda x, lo, hi: max(lo, min(x, hi))  # three params

```python
#Lambdas with default args:
```

```python
power = lambda x, exp=2: x ** exp
print(power(3))      # 9
print(power(3, 3))   # 27
```

```python
#Conditional expression in lambda:
```

```python
absolute = lambda x: x if x >= 0 else -x
classify = lambda n: "even" if n % 2 == 0 else "odd"
```

```python
#Where lambdas shine — as arguments to other functions:
```

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

# sorted with key:
print(sorted(nums))                          # [1,1,2,3,4,5,6,9]
print(sorted(nums, key=lambda x: -x))        # descending

words = ["banana", "fig", "apple", "kiwi"]
print(sorted(words, key=lambda w: len(w)))   # by length
print(sorted(words, key=lambda w: w[-1]))    # by last letter

# sorted with multiple criteria:
people = [("Diya",21),("Raj",25),("Amit",21),("Sara",25)]
print(sorted(people, key=lambda p: (p[1], p[0])))
# sort by age, then name: [('Amit',21),('Diya',21),('Raj',25),('Sara',25)]
```

```python
#With map() - apply to every item:
```

```python
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
print(squared)   # [1, 4, 9, 16, 25]
```

```python
#With filter() - keep matching items:
```

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)   # [2, 4, 6, 8]
```

```python
#With max() / min() / sorted() - custom key:
```

```python
products = [
    {"name": "Rice", "price": 60},
    {"name": "Oil",  "price": 180},
    {"name": "Sugar","price": 45}
]
cheapest = min(products, key=lambda p: p["price"])
print(cheapest["name"])   # Sugar
```

```python
#Lambda vs def - when to use which:
```

lambda → one-liner, used once, passed as argument
def    → complex logic, reused, needs docstring, multiple lines

```python
#Common beginner mistake - assigning lambdas:
```

```python
# PEP8 says don't do this — use def instead:
f = lambda x: x + 1   # bad style if you're naming it

# Just write:
def f(x): return x + 1   # preferred
```

```python
# Lambda expressions

double = lambda x: x * 2

is_even = lambda x: x % 2 == 0

first_char = lambda s: s[0]

full_name = lambda first, last: f"{first} {last}"

clamp = lambda x, lo, hi: max(lo, min(x, hi))

# Tests
print(double(7))

print(is_even(4))
print(is_even(7))

print(first_char("Python"))

print(full_name("Diya", "Desai"))

print(clamp(15, 0, 10))
print(clamp(-5, 0, 10))
```

```python
#sorted() with lambda keys
```

```python
students = [
    ("Diya", 21, 88),
    ("Raj", 25, 72),
    ("Amit", 21, 95),
    ("Sara", 25, 61),
    ("Kiran", 22, 79)
]

# 1. By name alphabetically
by_name = sorted(students, key=lambda s: s[0])

# 2. By score descending (highest first)
by_score_desc = sorted(students, key=lambda s: s[2], reverse=True)

# 3. By age, then name
by_age_name = sorted(students, key=lambda s: (s[1], s[0]))

# 4. By score then age (lowest score first, then youngest)
by_score_age = sorted(students, key=lambda s: (s[2], s[1]))

print(by_name)
print(by_score_desc)
print(by_age_name)
print(by_score_age)
```

```python
#map() and filter() with lambda
```

```python
prices = [120, 450, 89, 670, 234, 55, 890, 345]
names = [" diya ", "RAJ", "amit ", " SARA"]

# 1. Apply 10% discount to all prices
discounted = list(map(lambda p: p * 0.9, prices))
print(discounted)

# 2. Filter prices above 300
above_300 = list(filter(lambda p: p > 300, prices))
print(above_300)

# 3. Filter prices between 100 and 500
between_100_500 = list(filter(lambda p: 100 <= p <= 500, prices))
print(between_100_500)

# 4. Clean all names (strip + title case)
clean_names = list(map(lambda n: n.strip().title(), names))
print(clean_names)

# 5. Get lengths of all cleaned names
name_lengths = list(map(lambda n: len(n), clean_names))
print(name_lengths)

# 6. Chain: keep only even prices, then square them
even_squared = list(
    map(lambda p: p ** 2,
        filter(lambda p: p % 2 == 0, prices))
)
print(even_squared)
```

```python
#Lambda as function factory
```

```python
# 1. Function factory: multiplier
make_multiplier = lambda n: (lambda x: x * n)

# 2. Function factory: adder
make_adder = lambda n: (lambda x: x + n)

# 3. Function factory: range checker
make_between = lambda lo, hi: (lambda x: lo <= x <= hi)

# 4. Function factory: formatter
make_formatter = lambda prefix, suffix="": (
    lambda text: f"{prefix}{text}{suffix}"
)

# Pipeline
pipeline = [
    make_adder(10),
    make_multiplier(2),
    make_adder(-5)
]

numbers = [1, 2, 3, 4, 5]

results = []
for num in numbers:
    value = num
    for func in pipeline:
        value = func(value)
    results.append(value)

print(results)

# Range checker
is_teen = make_between(13, 17)

ages = [10, 14, 16, 20]
for age in ages:
    print(is_teen(age))

# Formatter
badge = make_formatter("[", "]")

words = ["Python", "AI", "Dev"]
for word in words:
    print(badge(word))
```

```python
#Data pipeline with lambdas
```

```python
employees = [
    {"name":"Diya","dept":"Eng","salary":85000,"years":3},
    {"name":"Raj","dept":"Mkt","salary":65000,"years":7},
    {"name":"Amit","dept":"Eng","salary":92000,"years":5},
    {"name":"Sara","dept":"HR","salary":58000,"years":2},
    {"name":"Kiran","dept":"Mkt","salary":72000,"years":4},
    {"name":"Meera","dept":"Eng","salary":78000,"years":6}
]

# 1. Filter: only Engineering dept
eng_employees = filter(
    lambda e: e["dept"] == "Eng",
    employees
)

# 2. Map: add bonus field
eng_with_bonus = map(
    lambda e: {
        **e,
        "bonus": e["salary"] * 0.1 * e["years"]
    },
    eng_employees
)

# 3. Sort: by bonus descending
sorted_eng = sorted(
    eng_with_bonus,
    key=lambda e: e["bonus"],
    reverse=True
)

# 4. Print each employee
list(map(
    lambda e: print(
        f'{e["name"]} | {e["dept"]} | Bonus: Rs.{e["bonus"]}'
    ),
    sorted_eng
))

# 5. Total bonus payout
total_bonus = sum(
    map(lambda e: e["bonus"], sorted_eng)
)

print(f"Total Eng bonus: Rs.{total_bonus}")
```
