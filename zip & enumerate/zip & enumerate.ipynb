```python
#zip — pair up iterables:
```

```python
names  = ["Diya", "Raj", "Amit"]
scores = [88, 72, 95]
grades = ["A", "B", "A+"]

# Basic zip:
paired = list(zip(names, scores))
# [('Diya',88), ('Raj',72), ('Amit',95)]

# Three iterables:
full = list(zip(names, scores, grades))
# [('Diya',88,'A'), ('Raj',72,'B'), ('Amit',95,'A+')]

# zip stops at shortest:
long  = [1, 2, 3, 4, 5]
short = ["a", "b", "c"]
print(list(zip(long, short)))   # [(1,'a'),(2,'b'),(3,'c')] ← 4,5 dropped!
```

```python
#zip_longest — use fill value for shorter:
```

```python
from itertools import zip_longest

list(zip_longest(long, short, fillvalue="?"))
# [(1,'a'),(2,'b'),(3,'c'),(4,'?'),(5,'?')]
```

```python
#Unzipping — with * operator:
```

```python
pairs = [("Diya",88), ("Raj",72), ("Amit",95)]

names, scores = zip(*pairs)   # unzip!
print(names)    # ('Diya', 'Raj', 'Amit')
print(scores)   # (88, 72, 95)
```

```python
#zip basics and dict building
```

```python
from itertools import zip_longest

keys = ["name", "age", "city", "cgpa", "is_student"]
values = ["Diya", 21, "Bhuj", 8.75, True]

subjects = ["Maths", "Science", "English", "CS", "Hindi"]
scores = [88, 92, 76, 95, 71]

# 1. Build profile dictionary
profile = dict(zip(keys, values))

# 2. Build subject → score dictionary
subject_scores = dict(zip(subjects, scores))

# 3. Print each subject-score pair
for subject, score in subject_scores.items():
    print(f"{subject}: {score}")

# 4. Unzip back into two tuples
unzipped_subjects, unzipped_scores = zip(*subject_scores.items())

# 5. zip vs zip_longest
a = [1, 2, 3, 4, 5]
b = [10, 20]

normal_zip = list(zip(a, b))
longest_zip = list(zip_longest(a, b, fillvalue=0))

print("\nProfile:", profile)
print("Subject Scores:", subject_scores)
print("Unzipped subjects:", unzipped_subjects)
print("Unzipped scores:", unzipped_scores)
print("zip (stops early):", normal_zip)
print("zip_longest (fills):", longest_zip)
```

```python
#enumerate patterns
```

```python
menu = ["Vada Pav", "Samosa", "Pav Bhaji", "Chai", "Lassi"]
prices = [15, 20, 60, 10, 30]
orders = ["Samosa", "Chai", "Vada Pav", "Lassi"]

# 1. Print menu with 1-based numbering
for i, (item, price) in enumerate(zip(menu, prices), start=1):
    print(f"{i}. {item} - Rs.{price}")

# 2. Find index of "Pav Bhaji" using enumerate
for index, item in enumerate(menu):
    if item == "Pav Bhaji":
        print(f"\nPav Bhaji is at index: {index}")
        break

# 3. Print all items with their price using enumerate + prices list
print("\nMenu with prices:")
for i, item in enumerate(menu):
    print(f"{i}: {item} - Rs.{prices[i]}")

# 4. Simulate order processing
print("\nOrder Processing:")
for order_no, order in enumerate(orders, start=1):
    print(f"Order #{order_no}: {order}")

# 5. Find the most expensive item
max_index, max_price = max(enumerate(prices), key=lambda x: x[1])
print(f"\nMost expensive: {menu[max_index]} at Rs.{max_price}")
```

```python
#Matrix operations with zip
```

```python
matrix_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix_b = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

# 1. Transpose matrix_a
transpose = [list(row) for row in zip(*matrix_a)]

print("Transposed:")
for row in transpose:
    print(row)

# 2. Add matrices element-wise
matrix_sum = [
    [a + b for a, b in zip(row_a, row_b)]
    for row_a, row_b in zip(matrix_a, matrix_b)
]

print("\nMatrix sum:")
for row in matrix_sum:
    print(row)

# 3. Find column sums
column_sums = [sum(col) for col in zip(*matrix_a)]
print("\nColumn sums:", column_sums)

# 4. Find row with maximum sum
max_row = max(matrix_a, key=sum)
print(f"Row with max sum: {max_row} (sum={sum(max_row)})")

# 5. Check if row 0 mirrors row 2 reversed
is_mirror = matrix_a[0] == matrix_a[2][::-1]
print(f"Row 0 mirrors row 2 reversed: {is_mirror}")

# 6. Print matrix_a nicely formatted
print("\nFormatted matrix:")
for row in matrix_a:
    print("".join(f"{num:3}" for num in row))
```

```python
#Leaderboard and ranking system
```

```python
from math import sqrt

players = ["Diya", "Raj", "Amit", "Sara", "Kiran", "Meera"]
round1 = [88, 72, 95, 61, 79, 84]
round2 = [91, 85, 87, 78, 92, 76]
round3 = [79, 88, 93, 82, 85, 90]

# 1. Calculate total score per player
player_data = [
    {
        "name": name,
        "r1": r1,
        "r2": r2,
        "r3": r3,
        "total": r1 + r2 + r3
    }
    for name, r1, r2, r3 in zip(players, round1, round2, round3)
]

# 2. Rank players (competition ranking with ties)
sorted_players = sorted(player_data, key=lambda p: p["total"], reverse=True)

rank = 0
prev_total = None

for pos, player in enumerate(sorted_players, start=1):
    if player["total"] != prev_total:
        rank = pos
    player["rank"] = rank
    prev_total = player["total"]

# 3. Show score improvement round-by-round per player
print("Improvements:")
for name, r1, r2, r3 in zip(players, round1, round2, round3):
    print(
        f"{name:6} | "
        f"R1→R2: {r2-r1:+3d} | "
        f"R2→R3: {r3-r2:+3d} | "
        f"Overall: {r3-r1:+3d}"
    )

# 4. Print leaderboard
print("\nFinal Leaderboard:")
for p in sorted_players:
    print(
        f"#{p['rank']:<2} "
        f"{p['name']:<6} | "
        f"Total: {p['total']} | "
        f"Rounds: {p['r1']}, {p['r2']}, {p['r3']}"
    )

# 5. Find who improved the most (round3 - round1)
improvements = [
    (name, r3 - r1)
    for name, r1, r3 in zip(players, round1, round3)
]

most_improved = max(improvements, key=lambda x: x[1])

# 6. Find consistent players (std dev < 7)
def std_dev(scores):
    mean = sum(scores) / len(scores)
    variance = sum((x - mean) ** 2 for x in scores) / len(scores)
    return sqrt(variance)

consistent = [
    name
    for name, r1, r2, r3 in zip(players, round1, round2, round3)
    if std_dev([r1, r2, r3]) < 7
]

print(f"\nMost improved: {most_improved[0]} (+{most_improved[1]})")
print("Consistent players (std<7):", ", ".join(consistent))
```

```python
#Data alignment and merge system
```

```python
students_db = {
    "S001": {"name": "Diya", "age": 21},
    "S002": {"name": "Raj", "age": 22},
    "S003": {"name": "Amit", "age": 20},
    "S004": {"name": "Sara", "age": 23}
}

exam_results = [
    ("S001", 88),
    ("S003", 95),
    ("S002", 72),
    ("S005", 85)
]

attendance = [
    ("S001", 92),
    ("S002", 78),
    ("S004", 88),
    ("S003", 95)
]

# 1. Align exam results with student names
print("Exam results aligned:")
for num, (sid, score) in enumerate(exam_results, start=1):
    name = students_db.get(sid, {}).get("name", "UNKNOWN")
    print(f"{num}. {name} ({sid}): {score}")

# 2. Merge attendance and exam results by student ID
exam_dict = dict(exam_results)
attendance_dict = dict(attendance)

common_ids = exam_dict.keys() & attendance_dict.keys()

merged_records = []
for sid in common_ids:
    name = students_db.get(sid, {}).get("name", "UNKNOWN")
    exam = exam_dict[sid]
    att = attendance_dict[sid]
    combined = exam * 0.7 + att * 0.3

    merged_records.append({
        "sid": sid,
        "name": name,
        "exam": exam,
        "attendance": att,
        "combined": combined
    })

print("\nMerged records:")
for record in sorted(merged_records, key=lambda x: x["sid"]):
    print(
        f"{record['sid']}: "
        f"{record['name']} | "
        f"Exam:{record['exam']} | "
        f"Attendance:{record['attendance']} | "
        f"Combined:{record['combined']:.1f}"
    )

# 3. Students present in all three datasets
db_ids = set(students_db.keys())
exam_ids = set(exam_dict.keys())
attendance_ids = set(attendance_dict.keys())

all_three = db_ids & exam_ids & attendance_ids

print("\nIn all 3 datasets:", ", ".join(sorted(all_three)))

# 4. Rank students by combined score
ranking = sorted(
    merged_records,
    key=lambda x: x["combined"],
    reverse=True
)

# 5. Formatted ranking report
print("\nFinal ranking:")
for rank, student in enumerate(ranking, start=1):
    print(
        f"#{rank} "
        f"{student['name']:<5} "
        f"({student['sid']}): "
        f"{student['combined']:.1f}"
    )
```
