```python
def f(*args):    # args is a TUPLE of all extra positional args
    print(args, type(args))

def g(**kwargs): # kwargs is a DICT of all extra keyword args
    print(kwargs, type(kwargs))

f(1, 2, 3)       # (1, 2, 3) <class 'tuple'>
g(a=1, b=2)      # {'a':1,'b':2} <class 'dict'>
```

```python
#The name args/kwargs is just convention - the * is what matters:
```

```python
def add(*numbers):     # totally fine
    return sum(numbers)

def show(**settings):  # totally fine
    print(settings)
```

```python
#Forwarding args - passing through to another function:
```

```python
def logger(func_name, *args, **kwargs):
    print(f"Calling {func_name} with {args} {kwargs}")

def process(*args, **kwargs):
    logger("process", *args, **kwargs)  # forward everything!
    return sum(args)

process(1, 2, 3, verbose=True)
# Calling process with (1,2,3) {'verbose': True}
```

```python
def first_last_rest(first, last, *rest):
    print(f"First: {first}, Last: {last}")
    print(f"Rest: {rest}")

first_last_rest(1, 10, 2, 3, 4, 5)
# First: 1, Last: 10
# Rest: (2, 3, 4, 5)
```

```python
#Keyword-only args - forced after *:
```

```python
def send(message, *recipients, cc=None, urgent=False):
    print(f"To: {recipients}")
    print(f"CC: {cc}")
    print(f"Urgent: {urgent}")
    print(f"Msg: {message}")

send("Hi!", "Diya", "Raj", cc="Boss", urgent=True)
# cc and urgent MUST be keyword — they come after *recipients
```

```python
def register(name, email, *, age, city):
    # age and city MUST be keyword args — no positional allowed
    print(name, email, age, city)

register("Diya", "d@e.com", age=21, city="Bhuj")
# register("Diya", "d@e.com", 21, "Bhuj")  ← TypeError!
```

```python
#**Decorators preview - the most important use of *args/kwargs:
```

```python
def log_calls(func):
    def wrapper(*args, **kwargs):   # capture ANY call
        print(f"Calling {func.__name__}")
        result = func(*args, **kwargs)  # forward to original
        print(f"Done. Returned: {result}")
        return result
    return wrapper

@log_calls
def add(a, b):
    return a + b

add(3, 5)
# Calling add
# Done. Returned: 8
```

```python
#Combining everything - real-world API style:
```

```python
def api_call(endpoint, method="GET", *path_params,
             timeout=30, retry=3, **query_params):
    url = f"/{endpoint}/" + "/".join(str(p) for p in path_params)
    params = "&".join(f"{k}={v}" for k,v in query_params.items())
    print(f"{method} {url}?{params} (timeout={timeout}, retry={retry})")

api_call("users", "GET", 42, "posts",
         timeout=10, page=1, limit=20)
# GET /users/42/posts?page=1&limit=20 (timeout=10, retry=3)
```

```python
# Function to calculate statistics
def stats(*numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    total = sum(numbers)
    average = total / len(numbers)

    print(f"Min:{minimum} Max:{maximum} Sum:{total} Avg:{average}")


# Function to merge multiple dictionaries
def merge_dicts(*dicts):
    merged = {}

    for d in dicts:
        merged.update(d)   # later dict overrides earlier ones

    return merged


# Test cases
stats(10, 25, 8, 42, 17, 33)
stats(1, 2, 3)

print(merge_dicts({"a":1}, {"b":2}, {"a":99, "c":3}))
```

```python
def send_email(subject, *recipients, cc=None, bcc=None,
               urgent=False, signature="Best regards"):

    # Add urgent tag if needed
    if urgent:
        subject = subject + " [URGENT]"

    print(f"Subject: {subject}")
    print(f"To: {', '.join(recipients)}")
    print(f"CC: {cc}")
    print(f"BCC: {bcc}")
    print(f"Signed: {signature}")
    print("---")


# Function calls
send_email(
    "Meeting",
    "diya@co.com",
    "raj@co.com",
    urgent=True,
    cc="boss@co.com"
)

send_email(
    "Hello",
    "amit@co.com",
    signature="Cheers"
)
```

```python
# Timing wrapper function
def timed_call(func, *args, **kwargs):
    result = func(*args, **kwargs)

    print(f"Args:{len(args)} Kwargs:{len(kwargs)} → Result:{result}")

    return result


# Retry function
def retry(func, times=3, *args, **kwargs):
    for attempt in range(1, times + 1):
        try:
            return func(*args, **kwargs)

        except Exception:
            print(f"Attempt {attempt} failed")

    print(f"All {times} attempts failed")
    return None


# Custom multiply function
def multiply(*nums):
    result = 1

    for num in nums:
        result *= num

    return result


# Flaky function
def flaky(n):
    if n < 3:
        raise ValueError("Number too small")

    return n * 10


# Testing timed_call
timed_call(sorted, [5, 3, 1, 4, 2], reverse=False)
timed_call(sum, [1, 2, 3, 4, 5])
timed_call(multiply, 3, 4, 5)

# Testing retry
print(retry(flaky, 2, 2))
print(retry(flaky, 3, 3))
```

```python
# Default settings
DEFAULTS = {
    "host": "localhost",
    "port": 8000,
    "debug": False,
    "workers": 4,
    "timeout": 30,
    "log_level": "INFO"
}


# Build config with overrides
def default_config(**overrides):
    config = DEFAULTS.copy()
    config.update(overrides)
    return config


# Validate config
def validate_config(config):
    errors = []

    # Check port
    if not (1 <= config["port"] <= 65535):
        errors.append("port out of range")

    # Check workers
    if not (1 <= config["workers"] <= 32):
        errors.append("workers out of range")

    # Check log level
    valid_levels = ["DEBUG", "INFO", "WARNING", "ERROR"]

    if config["log_level"] not in valid_levels:
        errors.append("invalid log_level")

    return (len(errors) == 0, errors)


# Print config
def print_config(config, *, show_defaults=True, prefix=""):
    items = []

    for key, value in config.items():

        # Skip default values if requested
        if not show_defaults and value == DEFAULTS[key]:
            continue

        items.append(f"{prefix}{key}={value}")

    print(" ".join(items))


# ---------------- TESTING ----------------

# Config 1
config1 = default_config(port=5000, debug=True)

valid, errors = validate_config(config1)

if valid:
    print("✅ Config 1 valid")
    print_config(config1)
else:
    print("❌ Config 1 invalid:", ", ".join(errors))


# Config 2
config2 = default_config(
    port=9000,
    debug=True,
    workers=8,
    log_level="DEBUG"
)

valid, errors = validate_config(config2)

if valid:
    print("✅ Config 2 valid")
    print_config(config2, show_defaults=False)
else:
    print("❌ Config 2 invalid:", ", ".join(errors))


# Config 3
config3 = default_config(
    port=70000,
    workers=50
)

valid, errors = validate_config(config3)

if valid:
    print("✅ Config 3 valid")
    print_config(config3)
else:
    print("❌ Config 3 invalid:", ", ".join(errors))
```

```python
#  HELPER FUNCTIONS

# Prime checker
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True


# Flatten nested lists
def flatten(data):
    result = []

    for item in data:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)

    return result


# Safe divide
def safe_divide(a, b):
    return a / b


#  MINI TEST FRAMEWORK 

passed = 0
failed = 0


def assert_equal(actual, expected, *, msg=""):
    global passed, failed

    if actual == expected:
        print(f"✅ PASS: {msg}")
        passed += 1
    else:
        print(f"❌ FAIL: {msg}")
        print(f"Expected: {expected}, Got: {actual}")
        failed += 1


def assert_true(condition, *, msg=""):
    global passed, failed

    if condition:
        print(f"✅ PASS: {msg}")
        passed += 1
    else:
        print(f"❌ FAIL: {msg}")
        failed += 1


def assert_raises(func, *args, exception=Exception, **kwargs):
    global passed, failed

    try:
        func(*args, **kwargs)

        print("❌ FAIL: Exception was not raised")
        failed += 1

    except exception:
        print("✅ PASS: divide by zero raises")
        passed += 1

    except Exception as e:
        print(f"❌ FAIL: Wrong exception raised: {e}")
        failed += 1


def run_tests(test_name, *test_funcs):
    global passed, failed

    passed = 0
    failed = 0

    for test in test_funcs:
        test()

    total = passed + failed

    print(f'\n=== Mini Test Results: "{test_name}" ===')
    print(f"Passed: {passed} | Failed: {failed} | Total: {total}")


#  TEST FUNCTIONS 

def test_prime_true():
    assert_true(is_prime(2), msg="2 is prime")


def test_prime_false():
    assert_equal(is_prime(4), False, msg="4 is not prime")


def test_flatten():
    data = [1, [2, [3, 4]], 5]
    assert_equal(flatten(data), [1, 2, 3, 4, 5],
                 msg="flatten works")


def test_safe_divide():
    assert_equal(safe_divide(10, 2), 5.0,
                 msg="safe divide works")


def test_divide_by_zero():
    assert_raises(safe_divide, 10, 0,
                  exception=ZeroDivisionError)


#  RUN ALL TESTS

run_tests(
    "My Tests",
    test_prime_true,
    test_prime_false,
    test_flatten,
    test_safe_divide,
    test_divide_by_zero
)
```
