```python
#break - exit the loop immediately:
```

```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
# Output: 1 2 3 4
# Loop stops the moment i hits 5
```

```python
#continue - skip this iteration, keep looping:
```

```python
for i in range(1, 8):
    if i == 4:
        continue
    print(i)
# Output: 1 2 3 5 6 7
# 4 is skipped, loop continues
```

```python
#break in a while loop:
```

```python
count = 0
while True:          # infinite loop!
    count += 1
    if count == 5:
        break        # only exit is the break
print("Stopped at", count)   # Stopped at 5
```

```python
#continue in a while loop:
```

```python
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue     # skip even numbers
    print(n)
# Output: 1 3 5 7 9
```

```python
#break with else - the most misunderstood Python feature:
```

```python
# The else block runs ONLY if loop finished WITHOUT hitting break
for i in range(1, 6):
    if i == 10:       # never true
        break
else:
    print("No break happened!")   # this runs

for i in range(1, 6):
    if i == 3:
        break
else:
    print("This won't print")    # break happened → else skipped
```

```python
#Real use case - searching:
```

```python
names = ["Raj", "Diya", "Amit", "Sara"]
target = "Amit"

for name in names:
    if name == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} not found.")
# Output: Found Amit!
```

```python
#Nested loops with break - only breaks inner loop:
```

```python
for i in range(3):
    for j in range(3):
        if j == 1:
            break        # only exits inner loop!
    print(f"i={i}")      # this still runs
# i=0, i=1, i=2
```

```python
#continue for filtering - very common pattern:
```

```python
numbers = [4, -2, 7, -5, 3, -8, 9]
print("Positive numbers only:")
for n in numbers:
    if n < 0:
        continue     # skip negatives
    print(n)
# 4 7 3 9
```

```python
#Skip the spaces
```

```python
sentence = "Hello World Python"

for ch in sentence:
    if ch == " ":
        continue
    print(ch)

print("Done! No spaces printed.")
```

```python
#Prime number finder
```

```python
num = 37

for i in range(2, num):
    if num % i == 0:
        print(num, "is not prime")
        break
else:
    print(num, "is prime")
```

```python
#Shopping filter + early stop
```

```python
products = [
    ("Laptop", 85000, 3),
    ("Phone", 25000, 8),
    ("Tablet", 35000, 0),
    ("Watch", 8000, 5),
    ("Camera", 45000, 2),
    ("Headphones", 3000, 12)
]

budget = 30000
count = 0

for name, price, stock in products:
    if stock == 0:
        continue

    if price > budget:
        continue

    print(f"{name} - Rs.{price} ({stock} in stock)")
    count += 1

    if count == 2:
        break

print("Search complete.")
```

```python
#Password validator with while/break
```

```python
passwords = ["abc", "hello123", "Py@123!", "short", "GoodPass9#"]

attempt = 0

while True:
    if attempt >= len(passwords):
        print("No valid password found.")
        break

    password = passwords[attempt]
    print("Trying:", password)

    failed_rules = []

    if len(password) < 8:
        failed_rules.append("too short")

    has_digit = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
            break

    if not has_digit:
        failed_rules.append("no digit")

    has_uppercase = False
    for ch in password:
        if ch.isupper():
            has_uppercase = True
            break

    if not has_uppercase:
        failed_rules.append("no uppercase")

    if len(failed_rules) == 0:
        print("Valid password found:", password)
        break
    else:
        print("Failed:", ", ".join(failed_rules))

    attempt += 1
```
