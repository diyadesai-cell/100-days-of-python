```python
#Basic syntax:
```

```python
# Regular loop:
squares = []
for x in range(1, 6):
    squares.append(x ** 2)

# List comprehension — same result, one line:
squares = [x ** 2 for x in range(1, 6)]
print(squares)   # [1, 4, 9, 16, 25]
```

```python
#Nested list comprehension:
```

```python
# Flatten a 2D list:
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [n for row in matrix for n in row]
# [1, 2, 3, 4, 5, 6, 7, 8, 9]

# All pairs:
pairs = [(x, y) for x in range(1,4) for y in range(1,4) if x != y]
# [(1,2),(1,3),(2,1),(2,3),(3,1),(3,2)]
```

```python
#Conditional expression (if-else) in comprehension:
```

```python
# Note: if-else goes BEFORE for, plain if goes AFTER
result = ["even" if x%2==0 else "odd" for x in range(1,6)]
# ['odd', 'even', 'odd', 'even', 'odd']

# Clamp values:
clamped = [min(100, max(0, x)) for x in [-5, 50, 110, 75, -10]]
# [0, 50, 100, 75, 0]
```

```python
#List comp vs map/filter - same thing, different style:
```

```python
nums = [1, 2, 3, 4, 5]

# map style:
squared = list(map(lambda x: x**2, nums))

# comprehension style (more Pythonic):
squared = [x**2 for x in nums]

# filter style:
evens = list(filter(lambda x: x%2==0, nums))

# comprehension style:
evens = [x for x in nums if x%2==0]
```

```python
# 1. Squares of 1 to 10
squares = [n ** 2 for n in range(1, 11)]

# 2. Even numbers from 1 to 20
evens = [n for n in range(1, 21) if n % 2 == 0]

# 3. Cubes of odd numbers from 1 to 15
odd_cubes = [n ** 3 for n in range(1, 16) if n % 2 != 0]

words = ["hello", "world", "python", "rocks"]

# 4. Uppercase versions
uppercase_words = [word.upper() for word in words]

# 5. Lengths of each word
word_lengths = [len(word) for word in words]

print(squares)
print(evens)
print(odd_cubes)
print(uppercase_words)
print(word_lengths)
```

```python
#Filter and transform
```

```python
nums = [15, -3, 42, 0, -8, 27, 61, -14, 33, 9]
sentences = [
    "Python is great",
    "I love coding",
    "sky is blue",
    "AI is future",
    "keep learning"
]

# 1. All positive numbers
positive_nums = [n for n in nums if n > 0]

# 2. Absolute values of all negative numbers
negative_abs = [abs(n) for n in nums if n < 0]

# 3. Numbers divisible by 3
div_by_3 = [n for n in nums if n % 3 == 0]

# 4. Sentences that contain the word "is"
contains_is = [s for s in sentences if "is" in s.split()]

# 5. First word of each sentence
first_words = [s.split()[0] for s in sentences]

# 6. Sentences with more than 2 words
more_than_2_words = [s for s in sentences if len(s.split()) > 2]

print(positive_nums)
print(negative_abs)
print(div_by_3)
print(contains_is)
print(first_words)
print(more_than_2_words)
```

```python
#Nested and paired comprehensions
```

```python
# 1. Flatten the matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]

# 2. (row, col) pairs for a 3x3 grid where row != col
coordinates = [(row, col) for row in range(3) for col in range(3) if row != col]

# 3. Multiplication table strings from 1 to 3
table = [f"{i}x{j}={i*j}" for i in range(1, 4) for j in range(1, 4)]

# 4. Name-score strings using zip()
names = ["Diya", "Raj", "Amit"]
scores = [88, 72, 95]
name_scores = [f"{name}: {score}" for name, score in zip(names, scores)]

# 5. All pairs (a, b) from [1..5] where a + b == 6
pairs = [(a, b) for a in range(1, 6) for b in range(1, 6) if a + b == 6]

print(flattened)
print(coordinates)
print(table)
print(name_scores)
print(pairs)
```

```python
#Comprehension with if-else
```

```python
temps_c = [0, 15, 28, -5, 37, 22, 42, -10, 19, 35]
words = ["Python", "java", "RUBY", "go", "JavaScript", "swift", "KOTLIN"]
scores = [45, 82, 91, 55, 73, 68, 39, 95, 61, 77]

# 1. Temperature categories
temp_labels = [
    "hot" if t > 30 else "warm" if t >= 15 else "cold"
    for t in temps_c
]

# 2. Title case words, add ⭐ if length > 5
normalized_words = [
    w.title() + ("⭐" if len(w) > 5 else "")
    for w in words
]

# 3. Grades
grades = [
    "A" if s >= 90 else
    "B" if s >= 75 else
    "C" if s >= 60 else
    "D" if s >= 40 else
    "F"
    for s in scores
]

# 4. Passing scores, otherwise 0
passing_scores = [s if s >= 60 else 0 for s in scores]

# 5. FizzBuzz from 1 to 20
fizzbuzz = [
    "FizzBuzz" if n % 15 == 0 else
    "Fizz" if n % 3 == 0 else
    "Buzz" if n % 5 == 0 else
    str(n)
    for n in range(1, 21)
]

print(temp_labels)
print(normalized_words)
print(grades)
print(passing_scores)
print(fizzbuzz)
```

```python
#Real data processing pipeline
```

```python
raw = [
    "TXN001,Diya,2024-01-15,850.00,food",
    "TXN002,Raj,2024-01-15,1200.50,electronics",
    "TXN003,Diya,2024-01-16,320.00,food",
    "TXN004,Amit,2024-01-16,450.75,clothing",
    "TXN005,Raj,2024-01-17,890.00,electronics",
    "TXN006,Sara,2024-01-17,150.00,food",
    "TXN007,Diya,2024-01-18,2100.00,electronics"
]

# 1. Parse into list of dictionaries
transactions = [
    {
        "id": parts[0],
        "name": parts[1],
        "date": parts[2],
        "amount": float(parts[3]),
        "category": parts[4]
    }
    for parts in [row.split(",") for row in raw]
]

# 2. Transactions above Rs.500
above_500 = [t["id"] for t in transactions if t["amount"] > 500]

# 3. Unique names (maintain order)
customers = list(dict.fromkeys([t["name"] for t in transactions]))

# 4. Total spent per person
totals = [
    f"{name}: Rs.{sum(t['amount'] for t in transactions if t['name'] == name)}"
    for name in customers
]

# 5. Most expensive transaction per category
categories = list(dict.fromkeys([t["category"] for t in transactions]))

most_expensive = [
    max(
        [t for t in transactions if t["category"] == category],
        key=lambda x: x["amount"]
    )
    for category in categories
]

# Output
print(f"{len(transactions)} transactions parsed")

print("Above 500:", ", ".join(above_500))

print("Customers:", ", ".join(customers))

for total in totals:
    print(total)

for txn in sorted(most_expensive, key=lambda x: x["category"]):
    print(
        f"{txn['category']}: {txn['id']} (Rs.{txn['amount']})"
    )
```
