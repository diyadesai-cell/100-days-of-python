```python
#Multiple return values - actually a tuple:
```

```python
def divmod_custom(a, b):
    return a // b, a % b    # returns tuple (quotient, remainder)

q, r = divmod_custom(17, 5)   # unpack
print(q, r)    # 3 2

result = divmod_custom(17, 5)
print(result)         # (3, 2)
print(result[0])      # 3
print(type(result))   # <class 'tuple'>
```

```python
def even_odd_split(nums):
    evens = [n for n in nums if n % 2 == 0]
    odds  = [n for n in nums if n % 2 != 0]
    return evens, odds   # tuple of two lists

e, o = even_odd_split([1,2,3,4,5,6])
print(e)   # [2,4,6]
print(o)   # [1,3,5]
```

```python
def analyse(text):
    words = text.split()
    return {
        "word_count": len(words),
        "char_count": len(text),
        "unique_words": len(set(words)),
        "longest": max(words, key=len)
    }

stats = analyse("the quick brown fox jumps")
print(stats["longest"])    # jumps
```

```python
def multiplier(factor):
    def multiply(n):
        return n * factor
    return multiply       # returns the inner function!

double = multiplier(2)
triple = multiplier(3)
print(double(5))    # 10
print(triple(5))    # 15
```

```python
def classify_bmi(weight_kg, height_m):
    # Early return for invalid input
    if weight_kg <= 0 or height_m <= 0:
        return "Invalid input"

    # Calculate BMI
    bmi = weight_kg / (height_m ** 2)

    # Classify BMI
    if bmi < 18.5:
        return "Underweight"

    if bmi < 25:
        return "Normal weight"

    if bmi < 30:
        return "Overweight"

    return "Obese"


# Test cases
print(classify_bmi(70, 1.75))
print(classify_bmi(50, 1.75))
print(classify_bmi(90, 1.75))
print(classify_bmi(120, 1.75))
print(classify_bmi(-5, 1.75))
```

```python
def split_numbers(nums):
    positives = []
    negatives = []
    zeros = []

    for num in nums:
        if num > 0:
            positives.append(num)
        elif num < 0:
            negatives.append(num)
        else:
            zeros.append(num)

    return positives, negatives, zeros


def analyse_text(text):
    words_list = text.split()

    words = len(words_list)
    chars = len(text)
    sentences = text.count(".") + text.count("!") + text.count("?")

    total_word_length = 0
    for word in words_list:
        clean_word = word.strip(".!?,'\"")
        total_word_length += len(clean_word)

    avg_word_length = round(total_word_length / words, 2)

    return {
        "words": words,
        "chars": chars,
        "sentences": sentences,
        "avg_word_length": avg_word_length
    }


# Test
nums = [5, -3, 0, 8, -1, 0, 4, -7, 0, 2]

p, n, z = split_numbers(nums)

print("Pos:", p)
print("Neg:", n)
print("Zero:", z)

stats = analyse_text("Hello world. Python is great! Isn't it?")

for k, v in stats.items():
    print(f"{k}: {v}")
```

```python
#Chained function calls
```

```python
def clean(text):
    return text.strip().lower()


def remove_punct(text):
    punctuations = ".,!?;:"

    for p in punctuations:
        text = text.replace(p, "")

    return text


def tokenize(text):
    return text.split()


def remove_stopwords(words):
    stopwords = ["the", "a", "an", "is", "in", "on", "at", "to", "and", "or"]

    filtered_words = []

    for word in words:
        if word not in stopwords:
            filtered_words.append(word)

    return filtered_words


def count_freq(words):
    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    # Sort dictionary by count descending
    sorted_freq = dict(
        sorted(freq.items(), key=lambda item: item[1], reverse=True)
    )

    return sorted_freq


# Raw text
raw = " The quick brown FOX jumps over the lazy DOG! The dog barked. "

# Chaining function calls
result = count_freq(
    remove_stopwords(
        tokenize(
            remove_punct(
                clean(raw)
            )
        )
    )
)

# Print top 5 frequencies
count = 0
for word, freq in result.items():
    print(f"{word}: {freq}")

    count += 1
    if count == 5:
        break
```

```python
# 1. Returns a power function
def make_power(exp):
    def power(num):
        return num ** exp

    return power


# 2. Returns a discount function
def make_discount(pct):
    def apply_discount(price):
        return price - (price * pct / 100)

    return apply_discount


# 3. Returns a validator function
def make_validator(min_val, max_val):
    def validate(value):
        return min_val <= value <= max_val

    return validate


# 4. Returns a greeting function
def make_greeting(greeting, punctuation="!"):
    def greet(name):
        return f"{greeting}, {name}{punctuation}"

    return greet


# Creating functions
square = make_power(2)
cube = make_power(3)

ten_off = make_discount(10)
twenty_off = make_discount(20)

is_valid_age = make_validator(0, 120)
is_valid_score = make_validator(0, 100)

hello = make_greeting("Hello")
namaste = make_greeting("Namaste", "🙏")


# Testing
print(square(5))
print(cube(3))

print(ten_off(1000))
print(twenty_off(1000))

print(is_valid_age(25))
print(is_valid_score(150))

print(hello("Diya"))
print(namaste("Raj"))
```

```python
#Result object pattern
```

```python
import math


# Success result
def ok(value):
    return {
        "success": True,
        "value": value,
        "error": None
    }


# Error result
def err(message):
    return {
        "success": False,
        "value": None,
        "error": message
    }


# Safe divide
def safe_divide(a, b):
    if b == 0:
        return err("Division by zero")

    return ok(a / b)


# Safe square root
def safe_sqrt(n):
    if n < 0:
        return err("Cannot sqrt negative")

    return ok(round(n ** 0.5, 4))


# Safe logarithm
def safe_log(n):
    if n <= 0:
        return err("Cannot log zero or negative")

    return ok(round(math.log(n), 4))


# Chain function
def chain(result, func):
    if result["success"]:
        return func(result["value"])

    return result


# Test cases
results = [
    safe_divide(10, 2),
    safe_divide(10, 0),
    safe_sqrt(16),
    safe_sqrt(-4),
    chain(safe_divide(100, 4), safe_sqrt),
    chain(safe_divide(10, 0), safe_sqrt)
]


# Print results
for r in results:
    if r["success"]:
        print(f"✅ {r['value']}")
    else:
        print(f"❌ {r['error']}")
```
