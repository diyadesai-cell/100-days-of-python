```python
#Python has 3 logical operators: and, or, not
```

```python
#and - both conditions must be True:
```

```python
age = 20
has_id = True
print(age >= 18 and has_id)    # True  ← both True
print(age >= 18 and not has_id) # False ← second is False
```

```python
#or - at least one condition must be True:
```

```python
is_weekend = False
is_holiday = True
print(is_weekend or is_holiday)  # True  ← one is True
print(is_weekend or False)       # False ← both False
```

```python
#not - flips True to False and vice versa:
```

```python
is_raining = False
print(not is_raining)    # True
print(not True)          # False
print(not 0)             # True  ← 0 is falsy, not 0 = True
```

```python
#Truth tables - memorise these:
```

```python
#and
#True  and True  → True
#True  and False → False
#False and True  → False
#False and False → False
```

```python
#or:
#True  or True  → True
#True  or False → True
#False or True  → True
#False or False → False
```

```python
#Operator precedence - not first, then and, then or:
```

```python
print(True or False and False)   # True  ← and runs first: False and False=False, then True or False=True
print(not False or False)        # True  ← not runs first: not False=True, then True or False=True
print(not (True or False))       # False ← brackets override order
```

```python
#Short-circuit evaluation - Python stops early:
```

```python
# With "and" — if first is False, second is never checked
# With "or"  — if first is True, second is never checked
x = 5
print(x > 10 and x / 0 > 1)   # False ← never divides by zero!
print(x < 10 or x / 0 > 1)    # True  ← never divides by zero!
```

```python
#Eligibility checker A job application has these rules: age = 24 experience = 2 has_degree = True cgpa = 7.8
```

```python
age = 24
experience = 2
has_degree = True
cgpa = 7.8

junior_dev = (21 <= age <= 28) and has_degree
senior_dev = (experience >= 3) or (cgpa >= 8.0)
intern = (age <= 25) and (not has_degree)

any_role = junior_dev or senior_dev or intern

print(junior_dev)
print(senior_dev)
print(intern)
print(any_role)
```

```python
x = 5
y = 10
z = 0
print(x > 3 and y < 20 or z == 1)
print(x > 3 and (y < 20 or z == 1))
print(not x == 5 or y > 8)
print(not (x == 5 or y > 8))
print(x > 0 and y > 0 and not z > 0)
```

```python
#Loan eligibility system
```

```python
age = 29
salary = 45000
credit_score = 720
existing_loans = 1
is_employed = True

basic_eligible = (21 <= age <= 60) and is_employed and salary >= 30000

good_credit = (credit_score >= 750) or (
    credit_score >= 700 and existing_loans == 0
)

high_risk = (existing_loans >= 2) or (credit_score < 600)

final_approval = (
    basic_eligible and
    not high_risk and
    (good_credit or salary >= 50000)
)

print(basic_eligible)
print(good_credit)
print(high_risk)
print(final_approval)
print("Loan approved:", final_approval)
```
