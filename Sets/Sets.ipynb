```python
#Set operations the real power:
```

```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}

# Union — all elements from both:
print(a | b)          # {1,2,3,4,5,6,7}
print(a.union(b))     # same

# Intersection — only common elements:
print(a & b)          # {3,4,5}
print(a.intersection(b))

# Difference — in a but NOT in b:
print(a - b)          # {1,2}
print(a.difference(b))

# Symmetric difference — in either but NOT both:
print(a ^ b)          # {1,2,6,7}
print(a.symmetric_difference(b))
```

```python
#Subset and superset:
```

```python
a = {1, 2, 3}
b = {1, 2, 3, 4, 5}

print(a.issubset(b))    # True  ← all of a is in b
print(b.issuperset(a))  # True  ← b contains all of a
print(a.isdisjoint({6,7,8}))  # True ← no common elements
print(a <= b)   # True ← subset operator
print(b >= a)   # True ← superset operator
```

```python
#Set basics and deduplication
```

```python
nums = [3,1,4,1,5,9,2,6,5,3,5,8,9,7,9,3,2,3]

# 1. Convert to a set
unique_nums = set(nums)
print("Set:", unique_nums)

# 2. Number of unique values
print("Unique count:", len(unique_nums))

# 3. Check if 7 is in the set
print("7 in set:", 7 in unique_nums)

# 4. Check if 10 is in the set
print("10 in set:", 10 in unique_nums)

# 5. Sorted list of unique values
sorted_unique = sorted(unique_nums)
print("Sorted unique:", sorted_unique)

# 6. Count duplicates removed
duplicates_removed = len(nums) - len(unique_nums)
print("Duplicates removed:", duplicates_removed)
```

```python
#Set operations
```

```python
python_students = {"Diya","Raj","Amit","Sara","Kiran"}
java_students = {"Raj","Meera","Amit","Priya","Kiran"}

# 1. Students in BOTH courses
both = sorted(python_students & java_students)
print("Both:", ", ".join(both))

# 2. Students in EITHER course
either = sorted(python_students | java_students)
print("Either:", ", ".join(either))

# 3. Only in Python
only_python = sorted(python_students - java_students)
print("Only Python:", ", ".join(only_python))

# 4. Only in Java
only_java = sorted(java_students - python_students)
print("Only Java:", ", ".join(only_java))

# 5. In exactly ONE course
exactly_one = sorted(python_students ^ java_students)
print("Exactly one:", ", ".join(exactly_one))

# 6. Check subset
print("Python subset of Java:", python_students.issubset(java_students))
```

```python
#Common elements finder
```

```python
group_a = {"Diya","Raj","Amit","Sara","Leo"}
group_b = {"Raj","Meera","Diya","Priya","Leo"}
group_c = {"Amit","Leo","Diya","Kiran","Raj"}
group_d = {"Sara","Diya","Leo","Raj","Arjun"}

groups = {
    "group_a": group_a,
    "group_b": group_b,
    "group_c": group_c,
    "group_d": group_d
}

# 1. People in ALL 4 groups
all_four = sorted(group_a & group_b & group_c & group_d)
print("In all 4:", ", ".join(all_four))

# 2. People in at least 3 groups
all_people = set.union(*groups.values())

at_least_3 = []
for person in all_people:
    count = sum(person in group for group in groups.values())
    if count >= 3:
        at_least_3.append(person)

print("In 3+ groups:", ", ".join(sorted(at_least_3)))

# 3. People in exactly 1 group
exactly_1 = []
for person in all_people:
    count = sum(person in group for group in groups.values())
    if count == 1:
        exactly_1.append(person)

print("In exactly 1:", ", ".join(sorted(exactly_1)))

# 4. Groups sharing at least 3 members with group_a
shared_groups = []

for name, group in groups.items():
    if name != "group_a":
        common = group_a & group
        if len(common) >= 3:
            shared_groups.append(name)

print("Groups sharing 3+ with A:", ", ".join(shared_groups))

# 5. Total unique people
print("Total unique people:", len(all_people))
```

```python
#Duplicate detector
```

```python
transactions = [
    ("TXN001","Diya",500),
    ("TXN002","Raj",1200),
    ("TXN001","Diya",500),
    ("TXN003","Amit",800),
    ("TXN002","Raj",1200),
    ("TXN004","Sara",300),
    ("TXN005","Diya",500),
    ("TXN003","Amit",800)
]

seen_ids = set()
duplicate_ids = set()
unique_customers = set()
duplicate_customers = set()
deduped_transactions = []

# Process transactions
for txn_id, customer, amount in transactions:
    
    # Unique customers
    unique_customers.add(customer)
    
    # Detect duplicates
    if txn_id in seen_ids:
        duplicate_ids.add(txn_id)
        duplicate_customers.add(customer)
    else:
        seen_ids.add(txn_id)
        deduped_transactions.append((txn_id, customer, amount))

# 1. Duplicate transaction IDs
print("Duplicate IDs:", ", ".join(sorted(duplicate_ids)))

# 2. Unique customers
print("Unique customers:", ", ".join(sorted(unique_customers)))

# 3. Customers with duplicate transactions
print("Customers with duplicates:", ", ".join(sorted(duplicate_customers)))

# 4. Total amount of UNIQUE transactions only
total = sum(amount for _, _, amount in deduped_transactions)
print("Total (unique only): Rs." + str(total))

# 5. Clean deduped transaction list
print("Deduped transactions:")

for txn_id, customer, amount in deduped_transactions:
    print(f"{txn_id} | {customer} | Rs.{amount}")
```
