```python
# Estimate pi using Monte Carlo:
def estimate_pi(n_samples):
    inside = sum(
        1 for _ in range(n_samples)
        if random.random()**2 + random.random()**2 <= 1
    )
    return 4 * inside / n_samples

print(estimate_pi(100000))   # ≈ 3.14...
```

```python
import random
import string


# 1. Generate PIN
def generate_pin(length=4):
    return ''.join(random.choices(string.digits, k=length))


# 2. Generate Password
def generate_password(length=12,
                      use_upper=True,
                      use_digits=True,
                      use_symbols=True):

    lowercase = string.ascii_lowercase
    uppercase = string.ascii_uppercase if use_upper else ""
    digits = string.digits if use_digits else ""
    symbols = "!@#$%^&*" if use_symbols else ""

    all_chars = lowercase + uppercase + digits + symbols

    # Weights:
    # lowercase more common than symbols
    weights = []

    for ch in all_chars:
        if ch in lowercase:
            weights.append(5)
        elif ch in uppercase:
            weights.append(3)
        elif ch in digits:
            weights.append(2)
        else:
            weights.append(1)

    password = ''.join(
        random.choices(all_chars, weights=weights, k=length)
    )

    return password


# 3. Generate OTP
def generate_otp(length=6):
    return ''.join(random.choices(string.digits, k=length))


# 4. Generate Username
def generate_username(adjectives, nouns):
    adjective = random.choice(adjectives)
    noun = random.choice(nouns)
    number = random.randint(10, 99)

    return f"{adjective}{noun}{number}"


# 5. Password Strength Checker
def strength_check(password):

    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_symbol = any(c in string.punctuation for c in password)

    score = sum([has_upper, has_lower, has_digit, has_symbol])

    if len(password) >= 12 and score == 4:
        return "Strong"
    elif len(password) >= 8 and score >= 3:
        return "Medium"
    else:
        return "Weak"


# ---------------- TESTING ----------------

random.seed(99)

print("PIN:", generate_pin())

passwords = []

for i in range(3):
    pwd = generate_password()
    passwords.append(pwd)
    print("Password:", pwd)

print("OTP:", generate_otp())

adjectives = ["happy", "smart", "cool", "fast"]
nouns = ["Tiger", "Lion", "Coder", "Eagle"]

username = generate_username(adjectives, nouns)
print("Username:", username)

print()

for pwd in passwords:
    print("Strength:", strength_check(pwd))
```

```python
import random
import math

# Reproducibility
random.seed(42)

# =====================================================
# 1. Monte Carlo estimation of π
# =====================================================

samples = 100000
inside_circle = 0

for _ in range(samples):
    x = random.random()
    y = random.random()

    if x**2 + y**2 <= 1:
        inside_circle += 1

pi_estimate = 4 * inside_circle / samples

print("Monte Carlo π estimate:", round(pi_estimate, 4))
print("Theoretical π:", round(math.pi, 5))

print()

# =====================================================
# 2. Probability of at least one 6 in 3 dice rolls
# =====================================================

success = 0

for _ in range(samples):
    dice = [random.randint(1, 6) for _ in range(3)]

    if 6 in dice:
        success += 1

prob_estimate = success / samples

# Theoretical probability
# P(at least one 6) = 1 - P(no 6)
theoretical_prob = 1 - (5/6)**3

print("P(at least one 6 in 3 dice):",
      f"{prob_estimate * 100:.1f}%")

print("Theoretical:",
      f"{theoretical_prob * 100:.2f}%")

print()

# =====================================================
# 3. Birthday paradox simulation
# =====================================================

trials = 10000
shared_birthday_count = 0

for _ in range(trials):

    birthdays = [random.randint(1, 365) for _ in range(23)]

    if len(birthdays) != len(set(birthdays)):
        shared_birthday_count += 1

birthday_estimate = shared_birthday_count / trials

# Theoretical probability
p_no_match = 1

for i in range(23):
    p_no_match *= (365 - i) / 365

birthday_theoretical = 1 - p_no_match

print("Birthday paradox (n=23):",
      f"{birthday_estimate * 100:.1f}%")

print("Theoretical:",
      f"{birthday_theoretical * 100:.2f}%")
```

```python
#Game simulation engine
```

```python
import random

# Reproducibility
random.seed(42)

WIN_SCORE = 100


# =====================================================
# Strategies
# =====================================================

def conservative(turn_score, total):
    return turn_score >= 15 or (total + turn_score >= WIN_SCORE)


def aggressive(turn_score, total):
    return turn_score >= 25 or (total + turn_score >= WIN_SCORE)


def random_strategy(turn_score, total):
    if turn_score > 5:
        return random.random() < 0.30
    return False


# =====================================================
# Single game simulation
# =====================================================

def play_game(strategy1, strategy2):

    scores = [0, 0]
    turns = 0
    current_player = 0

    strategies = [strategy1, strategy2]

    while scores[0] < WIN_SCORE and scores[1] < WIN_SCORE:

        turns += 1
        turn_score = 0

        while True:

            roll = random.randint(1, 6)

            # Rolled a 1 → lose turn score
            if roll == 1:
                turn_score = 0
                break

            turn_score += roll

            # Strategy decides whether to hold
            hold = strategies[current_player](
                turn_score,
                scores[current_player]
            )

            if hold:
                scores[current_player] += turn_score
                break

        # Check winner
        if scores[current_player] >= WIN_SCORE:
            return current_player, turns

        # Switch player
        current_player = 1 - current_player


# =====================================================
# Run many simulations
# =====================================================

def simulate_games(strategy1, strategy2,
                   name1, name2,
                   games=1000):

    wins = [0, 0]
    total_turns = 0

    for _ in range(games):

        winner, turns = play_game(strategy1, strategy2)

        wins[winner] += 1
        total_turns += turns

    avg_turns = total_turns / games

    print(f"Simulating {games} games: {name1} vs {name2}")

    print(f"{name1.capitalize()} wins: "
          f"{wins[0] / games * 100:.1f}%")

    print(f"{name2.capitalize()} wins: "
          f"{wins[1] / games * 100:.1f}%")

    print(f"Avg game length: {avg_turns:.1f} turns")

    print()


# =====================================================
# Run simulations
# =====================================================

simulate_games(
    conservative,
    aggressive,
    "conservative",
    "aggressive"
)

simulate_games(
    conservative,
    random_strategy,
    "conservative",
    "random"
)
```

```python
import random
import math

# Reproducibility
random.seed(42)

# =====================================================
# 1. ASCII Histogram
# =====================================================

def histogram(data, bins=10, width=30):

    min_val = min(data)
    max_val = max(data)

    bin_size = (max_val - min_val) / bins

    counts = [0] * bins

    # Count values in bins
    for value in data:

        index = int((value - min_val) / bin_size)

        # Handle max value edge case
        if index == bins:
            index -= 1

        counts[index] += 1

    max_count = max(counts)

    print("Histogram:\n")

    for i in range(bins):

        start = min_val + i * bin_size
        end = start + bin_size

        bar_length = int((counts[i] / max_count) * width)

        print(f"{start:.2f}-{end:.2f}: {'█' * bar_length}")


# =====================================================
# 2. Central Limit Theorem Demonstration
# =====================================================

def sample_mean_distribution(population,
                             sample_size,
                             n_samples):

    means = []

    for _ in range(n_samples):

        sample = random.sample(population, sample_size)

        sample_mean = sum(sample) / len(sample)

        means.append(sample_mean)

    print(f"\nHistogram of {n_samples} sample means:\n")

    histogram(means, bins=10)

    print("\nCentral Limit Theorem confirmed:")
    print("Sample means are approximately normally distributed")

    return means


# =====================================================
# 3. Bootstrap Confidence Interval
# =====================================================

def bootstrap_confidence_interval(data,
                                  stat_func,
                                  n_bootstrap=1000,
                                  confidence=0.95):

    stats = []

    n = len(data)

    for _ in range(n_bootstrap):

        # Resample with replacement
        sample = random.choices(data, k=n)

        stats.append(stat_func(sample))

    stats.sort()

    lower_index = int((1 - confidence) / 2 * n_bootstrap)
    upper_index = int((1 + confidence) / 2 * n_bootstrap)

    return (stats[lower_index], stats[upper_index])


# =====================================================
# 4. Stratified Sampling
# =====================================================

def stratified_sample(groups, n_per_group):

    result = {}

    for group_name, members in groups.items():

        result[group_name] = random.sample(
            members,
            min(n_per_group, len(members))
        )

    return result


# =====================================================
# TESTING
# =====================================================

# Exponential-ish population
population = [
    random.expovariate(1)
    for _ in range(10000)
]

# CLT demonstration
means = sample_mean_distribution(
    population,
    sample_size=50,
    n_samples=1000
)

print("\n" + "=" * 50)

# Bootstrap CI
data = [23, 45, 12, 67, 34, 89, 11, 56, 78, 90]

mean_func = lambda x: sum(x) / len(x)

ci = bootstrap_confidence_interval(
    data,
    mean_func,
    n_bootstrap=1000,
    confidence=0.95
)

print("\n95% Bootstrap Confidence Interval for Mean:")
print(f"({ci[0]:.2f}, {ci[1]:.2f})")

print("\n" + "=" * 50)

# Stratified sampling
student_groups = {
    "Eng": ["Diya", "Meera", "Amit", "Rohan"],
    "Mkt": ["Raj", "Kiran", "Sara", "Neha"],
    "HR":  ["Anjali", "Vikram", "Pooja"]
}

sampled = stratified_sample(student_groups, 2)

print("\nStratified Sample (2 from each group):\n")

for group, students in sampled.items():
    print(f"{group}: {students}")
```
