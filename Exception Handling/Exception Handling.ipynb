```python
import math

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Cannot divide by zero")

def safe_int(s):
    try:
        return int(s)
    except ValueError:
        print(f"Not a valid number: {s}")

def safe_index(lst, i):
    try:
        return lst[i]
    except IndexError:
        print(f"Index {i} out of range")

def safe_key(d, k):
    try:
        return d[k]
    except KeyError:
        print(f"Key '{k}' not found")

def safe_sqrt(n):
    try:
        if n < 0:
            raise ValueError
        return math.sqrt(n)
    except ValueError:
        print("Cannot sqrt negative number")

# Tests:
print(safe_divide(10, 2))    # 5.0
safe_divide(10, 0)            # Cannot divide by zero
print(safe_int("42"))         # 42
safe_int("hello")             # Not a valid number: hello
print(safe_index([1,2,3], 2)) # 3
safe_index([1,2,3], 10)       # Index 10 out of range
d = {"city": "Bhuj"}
print(safe_key(d, "city"))    # Bhuj
safe_key(d, "phone")          # Key 'phone' not found
print(safe_sqrt(16))          # 4.0
safe_sqrt(-4)                 # Cannot sqrt negative numberimport math

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Cannot divide by zero")

def safe_int(s):
    try:
        return int(s)
    except ValueError:
        print(f"Not a valid number: {s}")

def safe_index(lst, i):
    try:
        return lst[i]
    except IndexError:
        print(f"Index {i} out of range")

def safe_key(d, k):
    try:
        return d[k]
    except KeyError:
        print(f"Key '{k}' not found")

def safe_sqrt(n):
    try:
        if n < 0:
            raise ValueError
        return math.sqrt(n)
    except ValueError:
        print("Cannot sqrt negative number")

# Tests:
print(safe_divide(10, 2))    # 5.0
safe_divide(10, 0)            # Cannot divide by zero
print(safe_int("42"))         # 42
safe_int("hello")             # Not a valid number: hello
print(safe_index([1,2,3], 2)) # 3
safe_index([1,2,3], 10)       # Index 10 out of range
d = {"city": "Bhuj"}
print(safe_key(d, "city"))    # Bhuj
safe_key(d, "phone")          # Key 'phone' not found
print(safe_sqrt(16))          # 4.0
safe_sqrt(-4)                 # Cannot sqrt negative number
```

```python
#Q2 — Easy: try/except/else/finally
```

```python
def process_data(data_str):
    try:
        nums = [int(x) for x in data_str.split(",") if x]
        avg = sum(nums) / len(nums)
    except ValueError:
        print("Invalid data: contains non-numbers")
    except ZeroDivisionError:
        print("Empty data: no numbers to average")
    else:
        print(f"Success! Average: {avg:.2f}")
    finally:
        print("Processing complete.")

process_data("10,20,30,40")  # Success! Average: 25.00
process_data("10,abc,30")    # Invalid data...
process_data("")              # Empty data...
```

```python
#Q3 — Medium: Custom exceptions
```

```python
class BankError(Exception):
    pass

class InsufficientFundsError(BankError):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient funds: need Rs.{amount}, have Rs.{balance}")

class InvalidAmountError(BankError):
    def __init__(self, amount):
        super().__init__(f"Invalid amount: Rs.{amount} (must be positive)")

class AccountLockedError(BankError):
    def __init__(self):
        super().__init__("Account is locked. Contact support.")

def withdraw(balance, amount, locked=False):
    if locked:
        raise AccountLockedError()
    if amount <= 0:
        raise InvalidAmountError(amount)
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

# Tests:
tests = [(1000,500,{}),(1000,500,{"locked":True}),(1000,-100,{}),(500,2000,{})]
for a, b, kw in tests:
    try:
        r = withdraw(a, b, **kw)
        print(f"Withdrew Rs.{b}. Remaining: Rs.{r}")
    except BankError as e:
        print(f"{type(e).__name__}: {e}")
```

```python
#Q4 — Hard: Robust CSV parser
```

```python
raw_data = [
    "Diya,21,8.75,True",
    "Raj,twenty,7.2,False",
    "Amit,25,,True",
    "Sara,22,9.1,maybe",
    "Kiran,19,8.0,False",
    ",18,7.5,True"
]

def parse_record(line):
    parts = line.split(",")
    if not parts[0].strip():
        raise ValueError("name field empty")
    name = parts[0].strip()
    try:
        age = int(parts[1])
    except ValueError:
        raise ValueError(f"age field invalid ({parts[1]})")
    try:
        if not parts[2]:
            raise ValueError("missing")
        cgpa = float(parts[2])
    except ValueError:
        raise ValueError("cgpa field missing")
    if parts[3] not in ["True", "False"]:
        raise ValueError(f"is_student field invalid ({parts[3]})")
    return {"name": name, "age": age, "cgpa": cgpa, "is_student": parts[3]=="True"}

ok = fail = 0
for line in raw_data:
    try:
        r = parse_record(line)
        print(f"✅ {r['name']}: age={r['age']}, cgpa={r['cgpa']}, student={r['is_student']}")
        ok += 1
    except ValueError as e:
        print(f"❌ {line.split(',')[0]}: {e}")
        fail += 1
print(f"---\nParsed: {ok} | Failed: {fail}")
```

```python
#Q5 — Expert: Exception-safe pipeline
```

```python
class ValidationError(Exception):
    def __init__(self, field, value, reason):
        super().__init__(f"{field}: {value} ({reason})")

class TransformError(Exception):
    def __init__(self, step, value, reason):
        super().__init__(f"{step}: {value} ({reason})")

def validate(record):
    if not record.get("name","").strip():
        raise ValidationError("name", "", "required field empty")
    age = record.get("age")
    if not isinstance(age, int) or not 0 <= age <= 120:
        raise ValidationError("age", age, "must be 0-120")
    return record

def transform(record):
    r = record.copy()
    r["name"] = r["name"].strip().upper()
    return r

def enrich(record):
    r = record.copy()
    parts = r["name"].split()
    if len(parts) < 2:
        raise TransformError("enrich", r["name"], "missing name parts")
    r["initials"] = ".".join(p[0] for p in parts) + "."
    r["age_group"] = "minor" if r["age"]<18 else "adult" if r["age"]<60 else "senior"
    return r

def run_pipeline(records):
    ok = fail = 0
    for i, rec in enumerate(records, 1):
        try:
            r = validate(rec)
            r = transform(r)
            r = enrich(r)
            print(f"✅ {r['name']} | age={r['age']} | group={r['age_group']} | initials={r['initials']}")
            ok += 1
        except (ValidationError, TransformError) as e:
            print(f"❌ Record {i}: {type(e).__name__} - {e}")
            fail += 1
    print(f"---\nSucceeded: {ok} | Failed: {fail}")

records = [
    {"name":"Diya Desai","age":21},
    {"name":"Amit Shah","age":25},
    {"name":"Bad Guy","age":-5},
    {"name":"","age":18},
    {"name":"SingleName","age":30}
]
run_pipeline(records)
```
