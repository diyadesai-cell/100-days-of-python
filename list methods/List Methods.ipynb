```python
#Adding items:
```

```python
nums = [3, 1, 4]
nums.append(5)          # [3, 1, 4, 5] — add to end
nums.insert(1, 99)      # [3, 99, 1, 4, 5] — insert at index
nums.extend([7, 8])     # [3, 99, 1, 4, 5, 7, 8] — add multiple
```

```python
#Difference: append vs extend:
```

```python
a = [1, 2, 3]
a.append([4, 5])    # [1, 2, 3, [4, 5]] ← adds list AS one item!
b = [1, 2, 3]
b.extend([4, 5])    # [1, 2, 3, 4, 5]  ← adds items individually
```

```python
#Removing items:
```

```python
nums = [5, 3, 8, 3, 9]
nums.remove(3)      # [5, 8, 3, 9] — removes FIRST match only
last = nums.pop()   # removes & returns last → 9
item = nums.pop(1)  # removes & returns index 1 → 8
nums.clear()        # [] — removes everything
```

```python
#Searching:
```

```python
fruits = ["apple", "mango", "banana", "mango"]
print(fruits.index("mango"))    # 1 — first occurrence
print(fruits.count("mango"))    # 2 — total occurrences
print("kiwi" in fruits)         # False
```

```python
#Sorting:
```

```python
nums = [5, 2, 8, 1, 9]
nums.sort()                  # [1, 2, 5, 8, 9] — in place, ascending
nums.sort(reverse=True)      # [9, 8, 5, 2, 1] — descending
print(sorted(nums))          # new sorted list, original unchanged

words = ["banana", "apple", "cherry"]
words.sort(key=len)          # sort by string length!
# ['apple', 'banana', 'cherry']
```

```python
#Reversing:
```

```python
nums = [1, 2, 3, 4, 5]
nums.reverse()          # [5, 4, 3, 2, 1] — in place
print(nums[::-1])       # new reversed list, original unchanged
```

```python
#Copying - the trap every beginner falls into:
```

```python
a = [1, 2, 3]
b = a           # NOT a copy — both point to same list!
b.append(4)
print(a)        # [1, 2, 3, 4] ← a changed too!

# Correct ways to copy:
b = a.copy()        # method 1
b = a[:]            # method 2 — slice copy
b = list(a)         # method 3
```

```python
#Other useful methods:
```

```python
nums = [3, 1, 4, 1, 5]
print(nums.copy())      # returns a shallow copy
nums2 = [6, 7]
nums2.extend(nums)      # adds all nums to nums2
```

```python
data = [5, 2, 8, 1, 9, 3]
top3 = sorted(data, reverse=True)[:3]
print(top3)    # [9, 8, 5]
```

```python
#append vs extend
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

# 1. Append [4, 5] to list a
a.append([4, 5])
print(a)

# 2. Extend list b with [4, 5]
b.extend([4, 5])
print(b)

# 3. Print lengths
print(len(a))
print(len(b))

# 4. Difference explanation
print("append added 1 items, extend added 2 items")
```

```python
#pop and remove
```

```python
queue = ["Alice", "Bob", "Charlie", "Bob", "Diana"]

# 1. Remove first "Bob"
queue.remove("Bob")
print(queue)

# 2. Pop last person
last_person = queue.pop()
print("Popped:", last_person)
print(queue)

# 3. Pop person at index 0
first_person = queue.pop(0)
print("Popped:", first_person)
print(queue)

# 4. Final queue length
print(len(queue))
```

```python
#Sort by multiple criteria
```

```python
words = ["banana", "fig", "apple", "kiwi", "cherry", "date"]

# 1. Alphabetically
print(sorted(words))

# 2. Reverse alphabetically
print(sorted(words, reverse=True))

# 3. By word length (shortest first)
print(sorted(words, key=len))

# 4. By word length (longest first)
print(sorted(words, key=len, reverse=True))

# Original list remains unchanged
print(words)
```

```python
#The copy trap
```

```python
# 1. Create original list
original = [10, 20, 30]

# 2. Create reference
ref = original

# 3. Append 99 to ref
ref.append(99)

# 4. Print both lists
print(original)
print(ref)

# 5. Create another original list
original2 = [10, 20, 30]

# 6. Create real copy
real_copy = original2.copy()

# 7. Append 99 to copied list
real_copy.append(99)

# 8. Print both lists
print(original2)
print(real_copy)

# 9 and 10. Print comparison results
print("Reference changed original:", original == ref)
print("Copy changed original:", original2 == real_copy)
```

```python
#Playlist manager
```

```python
playlist = ["Shape of You", "Blinding Lights", "Levitating", "Stay", "Peaches"]

# 1. Add "Montero" to end
playlist.append("Montero")
print(playlist)

# 2. Add "drivers license" at position 2
playlist.insert(2, "drivers license")
print(playlist)

# 3. Remove "Stay"
playlist.remove("Stay")
print(playlist)

# 4. Sort alphabetically
playlist.sort()
print(playlist)

# 5. Print top 3 songs
print(playlist[:3])

# 6. Find index of "Peaches"
print(playlist.index("Peaches"))

# 7. Count songs with length > 8 characters
count = 0

for song in playlist:
    if len(song) > 8:
        count += 1

print(count)

# 8. Reverse playlist in place
playlist.reverse()
print(playlist)

# 9. Pop last song
now_playing = playlist.pop()
print("Now playing:", now_playing)

# 10. Final playlist and count
print(playlist)
print(len(playlist))
```
