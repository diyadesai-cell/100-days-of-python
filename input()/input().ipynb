```python
#Greet the user
name = input("Enter your name: ")
print(f"Hello, {name}! Welcome to Python.")
```

```python
#Age next year
age = input("Enter your age: ")
next_age = int(age) + 1
print(f"Next year you will be {next_age} years old.")
```

```python
#Simple bill splitter
total = input("Enter total bill: ")
people = input("Enter number of people: ")
each = float(total) / int(people)
print(f"Each person pays: {each} rupees")
```

```python
#Temperature converter
celsius = input("Enter temperature in Celsius: ")
celsius = int(celsius)
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

```python
#Mini user profile
name = input("Enter your name: ")
age = input("Enter your age: ")
cgpa = input("Enter your CGPA: ")
age_int = int(age)
is_adult = age_int >= 18
print("--- User Profile ---")
print(f"Name : {name}")
print(f"Age  : {age_int}")
print(f"CGPA : {cgpa}")
print(f"Adult: {is_adult}")
```
