```python
#String methods are functions that belong to strings. You call them using a dot: string.method()
```

```python
#Case methods
```

```python
text = "hello WORLD"
print(text.upper())       
print(text.lower())      
print(text.title())       
print(text.capitalize())  
print(text.swapcase())    
```

```python
#Whitespace methods:
```

```python
msg = "   hello   "
print(msg.strip())    # "hello"   ← removes both sides
print(msg.lstrip())   # "hello   " ← left only
print(msg.rstrip())   # "   hello" ← right only
```

```python
#Search methods:
```

```python
s = "I love Python and Python loves me"
print(s.find("Python"))     # 7  ← index of first match (-1 if not found)
print(s.count("Python"))    # 2  ← how many times it appears
print(s.startswith("I"))    # True
print(s.endswith("me"))     # True
```

```python
#Replace and split:
```

```python
s = "I love Java"
print(s.replace("Java", "Python"))   

csv = "apple,mango,banana"
print(csv.split(",")) 

words = ["Hello", "World"]
print(" ".join(words))  
```

```python
#Check content:
```

```python
print("hello".isalpha())    # True  ← only letters
print("123".isdigit())      # True  ← only digits
print("  ".isspace())       # True  ← only spaces
print("hello123".isalnum()) # True  ← letters or digits
```

```python
#Chaining methods - you can stack them:
```

```python
text = "  hello world  "
print(text.strip().title())  
```

```python
#Search and count
```

```python
sentence = "Python is great and python is easy"

# Convert to lowercase for case-insensitive counting
lower_sentence = sentence.lower()
print(lower_sentence.count("python")) 
print(sentence.index("great"))        
print(sentence.startswith("Python"))   
```

```python
#CSV to list to sentence
```

```python
data = "mango,apple,banana,grape"
# Split into a list
fruits = data.split(",")
print(fruits)
# Join with " | "
joined = " | ".join(fruits)

print(joined)
```

```python
#Username validator
```

```python
username = " Dev_Coder123 "
# Step 1: Strip whitespace
cleaned = username.strip()
print(cleaned)

# Step 2: Convert to lowercase
cleaned = cleaned.lower()
print(cleaned)

# Step 3: Replace "_" with "-"
cleaned = cleaned.replace("_", "-")
print(cleaned)

# Step 4: Check if alphanumeric
print(cleaned.isalnum())

# Step 5: Print total character count
print(len(cleaned))
```
