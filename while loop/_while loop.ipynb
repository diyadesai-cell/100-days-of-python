```python
#A while loop keeps running its block as long as the condition is True:
```

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
#Anatomy of a while loop:
```

```python
# 1. Initialize the variable BEFORE the loop
count = 1

# 2. Condition — checked before every iteration
while count <= 5:

    # 3. Body — what runs each time
    print(count)

    # 4. Update — MUST change the variable or loop runs forever!
    count += 1
```

```python
#Infinite loop — the most common beginner mistake:
```

```python
# DANGER — never do this without a break!
count = 1
while count <= 5:
    print(count)
    # forgot count += 1 → runs forever!
```

```python
#while with else — runs when condition becomes False:
```

```python
n = 1
while n <= 3:
    print(n)
    n += 1
else:
    print("Loop finished!")
# Output: 1, 2, 3, Loop finished!
```

```python
#Countdown:
```

```python
n = 5
while n > 0:
    print(n)
    n -= 1
print("Blast off!")
```

```python
#User input loop - keep asking until valid:
```

```python
# Simulated version (since we can't use real input here)
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    print(f"Attempt {attempts + 1}: Enter password")
    attempts += 1
print("Too many attempts!")
```

```python
#Accumulator pattern — building a sum:
```

```python
total = 0
n = 1
while n <= 10:
    total += n    # adds 1+2+3+...+10
    n += 1
print(total)    
```

```python
#break - exit the loop immediately:
```

```python
n = 1
while n <= 10:
    if n == 5:
        break     # stops at 5
    print(n)
    n += 1
```

```python
#continue - skip current iteration:
```

```python
n = 0
while n < 6:
    n += 1
    if n == 3:
        continue   # skips printing 3
    print(n)
```

```python
i = 1

while i <= 10:
    print(i)
    i += 1

print("Done!")
```

```python
#Sum of first N numbers
```

```python
n = 100
sum = 0
i = 1

while i <= n:
    sum += i
    i += 1

print("Sum =", sum)
```

```python
#Multiplication table
```

```python
num = 7
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

```python
#Find first power of 2 over N
```

```python
n = 1000
power = 1

while power <= n:
    print(power)
    power *= 2

print("First power of 2 greater than 1000:", power)
```

```python
#Number guessing game (simulated)
```

```python
secret = 42
guesses = [10, 67, 35, 42, 55]

attempts = 0
max_attempts = 5

while attempts < max_attempts:
    guess = guesses[attempts]

    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("Correct! You got it in", attempts + 1, "attempts!")
        break

    attempts += 1

if attempts == max_attempts:
    print("Game over! The number was", secret)
```
