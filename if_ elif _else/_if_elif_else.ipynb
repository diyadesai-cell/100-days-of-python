```python
#Basic if
```

```python
#The colon : and indentation (4 spaces) are mandatory. Python uses indentation instead of {} brackets like other languages.
```

```python
age = 20
if age >= 18:
    print("You are an adult")
```

```python
#if / else - two paths:
```

```python
score = 45
if score >= 50:
    print("Passed!")
else:
    print("Failed!")
```

```python
#if / elif / else - multiple paths
```

```python
marks = 78

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
elif marks >= 40:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
#Python checks from top to bottom and stops at the first condition that is True
```

```python
#Nested if - if inside an if:
```

```python
age = 20
has_ticket = True

if age >= 18:
    if has_ticket:
        print("Entry allowed")
    else:
        print("Need a ticket")
else:
    print("Too young")
```

```python
#One-line if (ternary) - short and clean:
```

```python
x = 10
result = "even" if x % 2 == 0 else "odd"
print(result)  
```

```python
#Combining with logical operators:
```

```python
age = 25
salary = 40000

if age >= 21 and salary >= 30000:
    print("Loan approved")
elif age >= 21 or salary >= 30000:
    print("Conditional approval")
else:
    print("Rejected")
```

```python
age = 14

if age < 5:
    print("Free")
elif age <= 12:
    print("Rs. 100")
elif age <= 17:
    print("Rs. 150")
elif age <= 59:
    print("Rs. 250")
else:
    print("Rs. 120")
```

```python
#ATM machine
```

```python
balance = 5000
withdraw = 3500
pin_correct = True

if not pin_correct:
    print("Wrong PIN. Access denied.")
else:
    if withdraw > balance:
        print("Insufficient funds.")
    elif withdraw > 0:
        print("Dispensing Rs.", withdraw)
        print("New balance: Rs.", balance - withdraw)
    else:
        print("Invalid amount.")
```

```python
#Traffic light + speed fine system
```

```python
light = "green"
speed = 95
zone = "highway"

if light == "red":
    print("Stop! You ran a red light. Fine: Rs. 5000")

elif light == "yellow":
    print("Slow down!")

elif light == "green":
    if zone == "school" and speed > 25:
        print("School zone speeding! Fine: Rs. 3000")

    elif zone == "city" and speed > 60:
        print("City speeding! Fine: Rs. 1000")

    elif zone == "highway" and speed > 120:
        print("Highway speeding! Fine: Rs. 2000")

    else:
        print("Drive safe!")

else:
    print("Unknown signal.")
```
