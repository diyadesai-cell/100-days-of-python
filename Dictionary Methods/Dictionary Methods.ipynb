```python
#The setdefault pattern - most underused method:
```

```python
data = [
    {"name": "Diya", "city": "Bhuj"},
    {"name": "Raj", "city": "Ahmedabad"},
    {"name": "atik", "city": "Bhuj"},
    {"name": "Sara", "city": "Surat"}
]

groups = {}

for item in data:
    groups.setdefault(item["city"], []).append(item["name"])

print(groups)
```

```python
#Merging dictionaries:
```

```python
a = {"x": 1, "y": 2}
b = {"y": 99, "z": 3}

# Method 1 — update (modifies a):
a.update(b)       # {"x":1, "y":99, "z":3} ← b overwrites a

# Method 2 — unpacking (Python 3.5+):
merged = {**a, **b}   # new dict, b's values win on conflict

# Method 3 — | operator (Python 3.9+):
merged = a | b
```

```python
#Iterating patterns:
```

```python
d = {"maths": 88, "science": 92, "english": 76}

# Pattern 1 — keys only:
for subject in d:
    print(subject)

# Pattern 2 — values only:
for score in d.values():
    print(score)

# Pattern 3 — both (most common):
for subject, score in d.items():
    print(f"{subject}: {score}")

# Pattern 4 — with index:
for i, (key, val) in enumerate(d.items(), 1):
    print(f"{i}. {key}: {val}")
```

```python
config = {
    "theme": "dark",
    "language": "English",
    "font_size": 14
}

# 1. Get theme
print(config.get("theme"))

# 2. Get volume safely with default 50
print(config.get("volume", 50))

# 3. Add notifications if not exists
print(config.setdefault("notifications", True))

# 4. setdefault on existing key
print(config.setdefault("theme", "light"))

# 5. Pop font_size
print(config.pop("font_size"))

# 6. Safely pop username
print(config.pop("username", "guest"))

# 7. Print final config
print(config)
```

```python
#Merging dictionaries
```

```python
defaults = {
    "theme": "light",
    "lang": "English",
    "font": 14,
    "timeout": 30
}

user_prefs = {
    "theme": "dark",
    "font": 18,
    "notifications": True
}

# 1. Merge dictionaries
merged = defaults | user_prefs

# 2. Print merged result
print(merged)

# 3. Print customised keys
print("Customised:", ", ".join(user_prefs.keys()))

# 4. Print kept default settings
kept_defaults = []

for key in defaults:
    if key not in user_prefs:
        kept_defaults.append(key)

print("Kept defaults:", ", ".join(kept_defaults))

# 5. Total settings
print("Total settings:", len(merged))
```

```python
#Grouping with setdefault
```

```python
employees = [
    {"name": "Diya", "dept": "Engineering", "salary": 85000},
    {"name": "Raj", "dept": "Marketing", "salary": 65000},
    {"name": "Amit", "dept": "Engineering", "salary": 92000},
    {"name": "Sara", "dept": "HR", "salary": 58000},
    {"name": "Kiran", "dept": "Marketing", "salary": 72000},
    {"name": "Meera", "dept": "Engineering", "salary": 78000}
]

# 1. Group employees by department
groups = {}

for emp in employees:
    groups.setdefault(emp["dept"], []).append(emp)

# Variables for highest paying department
highest_avg = 0
highest_dept = ""

# Total payroll
total_payroll = 0

# 2. Print names and average salary
for dept, emp_list in groups.items():

    names = []
    total_salary = 0

    for emp in emp_list:
        names.append(emp["name"])
        total_salary += emp["salary"]
        total_payroll += emp["salary"]

    avg_salary = total_salary / len(emp_list)

    print(f"{dept}: {', '.join(names)} | Avg salary: Rs.{avg_salary}")

    # 3. Find highest paying department
    if avg_salary > highest_avg:
        highest_avg = avg_salary
        highest_dept = dept

print(f"Highest paying dept: {highest_dept}")

# 4. Print total payroll
print(f"Total payroll: Rs.{total_payroll}")
```

```python
#Dict as a counter election results
```

```python
votes = [
    "Raj", "Diya", "Raj", "Amit", "Diya",
    "Raj", "Sara", "Diya", "Raj", "Amit",
    "Diya", "Raj", "Sara", "Raj", "Diya"
]

# 1. Count votes using .get()
results = {}

for vote in votes:
    results[vote] = results.get(vote, 0) + 1

# Total votes
total_votes = len(votes)

# 2. Sort results by votes descending
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)

for candidate, count in sorted_results:
    percentage = round((count / total_votes) * 100, 2)
    print(f"{candidate}: {count} votes ({percentage}%)")

# 3. Declare winner
winner = sorted_results[0][0]
winner_votes = sorted_results[0][1]

print(f"Winner: {winner}")

# 4. Majority win check
majority = winner_votes > (total_votes / 2)

print(f"Majority win: {majority}")

# 5. Runner-up
runner_up = sorted_results[1][0]

print(f"Runner-up: {runner_up}")

# 6. Candidates with less than 3 votes
low_votes = []

for candidate, count in results.items():
    if count < 3:
        low_votes.append(candidate)

print("Less than 3 votes:", ", ".join(low_votes))

# 7. Remove last entry using popitem()
removed = results.popitem()

print("Removed entry:", removed)
```

```python
#Multi-level config system
```

```python
base_config = {
    "app": {
        "name": "MyApp",
        "version": "1.0",
        "debug": False
    },

    "db": {
        "host": "localhost",
        "port": 5432,
        "name": "mydb"
    },

    "cache": {
        "enabled": True,
        "ttl": 300
    }
}

env_config = {
    "app": {
        "version": "2.0",
        "debug": True
    },

    "db": {
        "host": "prod-server",
        "password": "secret"
    },

    "email": {
        "smtp": "mail.server.com",
        "port": 587
    }
}

# Deep merge
merged = {}

# Copy base config first
for section, settings in base_config.items():
    merged[section] = settings.copy()

new_sections = []
modified_keys = []

# Merge env config
for section, settings in env_config.items():

    # New section
    if section not in merged:
        merged[section] = settings.copy()
        new_sections.append(section)

    else:
        # Update nested keys
        for key, value in settings.items():

            # Track modified keys
            if key in merged[section]:
                if merged[section][key] != value:
                    modified_keys.append(f"{section}.{key}")

            merged[section][key] = value

# 2. Print merged config
for section, settings in merged.items():

    print(f"--- {section} ---")

    for key, value in settings.items():
        print(f"{key}: {value}")

# 3. New sections
print("New sections:", ", ".join(new_sections))

# 4. Modified keys
print("Modified keys:", ", ".join(modified_keys))

# 5. Flat config
print("Flat config:")

for section, settings in merged.items():

    for key, value in settings.items():
        print(f"{section}.{key} = {value}")
```
