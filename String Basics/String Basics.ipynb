```python
#A string is any text wrapped in quotes. Python treats single ' and double " quotes the same way.
```

```python
name = "Diya"
city = 'baroda'
print(name,city)
```

```python
#String with quotes inside - use the other type
```

```python
msg1 = "It's a great day!"     # single quote inside double → fine
msg2 = 'He said "hello"'       # double quote inside single → fine
msg3 = "She said \"wow\""      # escape with backslash → also fine
print(msg1)
print(msg2)
print(msg3)
```

```python
#Multiline strings — use triple quotes
```

```python
bio = """My name is Diya.
I am from vadodara.
I love Python."""
print(bio)
```

```python
#String length with len()
```

```python
name = "Diya"
print(len(name))   
```

```python
#Indexing - access individual characters
```

```python
name = "Python"
print(name[0])    # P  ← first character (index starts at 0!)
print(name[1])    # y
print(name[-1])   # n  ← last character (negative index goes from end)
print(name[-2])   # o
```

```python
#Slicing  extract a part of a string
```

```python
name = "Python"
print(name[0:3])   # Pyt   ← from index 0 up to (not including) 3
print(name[2:5])   # tho
print(name[:3])    # Pyt   ← start from beginning
print(name[3:])    # hon   ← go till end
print(name[::-1])  # nohtyP ← reverse the string!
```
