```python
#4 types of parameters:
```

```python
#1. Positional - order matters:
```

```python
def greet(name, age):
    print(f"{name} is {age}")

greet("Diya", 21)    # correct
greet(21, "Diya")    # wrong order — prints "21 is Diya"
```

```python
#2. Default - fallback value:
```

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Diya")           # Hello, Diya!
greet("atik", "Hi")      # Hi, atik!
```

```python
#3. *args - unlimited positional arguments:
```

```python
def add_all(*args):
    print(args)          # tuple: (1, 2, 3, 4, 5)
    return sum(args)

print(add_all(1, 2, 3))        # 6
print(add_all(1, 2, 3, 4, 5)) # 15
print(add_all(10))              # 10
```

```python
#4. **kwargs - unlimited keyword arguments:
```

```python
def show_info(**kwargs):
    print(kwargs)   # dict: {'name':'Diya','age':21}
    for key, val in kwargs.items():
        print(f"{key}: {val}")

show_info(name="Diya", age=21, city="Bhuj")
```

```python
#Combining all 4 - order matters:
```

```python
def full_func(pos1, pos2, default="hi", *args, **kwargs):
    print(pos1, pos2, default, args, kwargs)

# Rule: positional → default → *args → **kwargs
```

```python
#Unpacking when calling:
```

```python
def add(a, b, c):
    return a + b + c

nums = [1, 2, 3]
print(add(*nums))          # unpacks list → add(1,2,3)

info = {"a": 1, "b": 2, "c": 3}
print(add(**info))         # unpacks dict → add(a=1,b=2,c=3)
```

```python
#Keyword-only arguments (after *):
```

```python
def greet(name, *, greeting="Hello", punctuation="!"):
    print(f"{greeting}, {name}{punctuation}")

greet("Diya")                          # Hello, Diya!
greet("Diya", greeting="Hi")          # Hi, Diya!
# greet("Diya", "Hi")  ← TypeError! greeting must be keyword
```

```python
#Positional-only arguments (before /):
```

```python
def add(a, b, /):
    return a + b

add(1, 2)           # works
# add(a=1, b=2)     # TypeError! must be positional
```

```python
#Real-world pattern - flexible logger:
```

```python
def log(level="INFO", *messages, sep=" ", **metadata):
    msg = sep.join(str(m) for m in messages)
    meta = " | ".join(f"{k}={v}" for k,v in metadata.items())
    print(f"[{level}] {msg} | {meta}")

log("ERROR", "File", "not", "found", sep="-", code=404, user="Diya")
# [ERROR] File-not-found | code=404 | user=Diya
```

```python
#*args - unlimited inputs
```

```python
# Function to calculate total
def total(*args):
    return sum(args)


# Function to find longest string
def longest(*args):
    longest_word = args[0]

    for word in args:
        if len(word) > len(longest_word):
            longest_word = word

    return longest_word


# Test calls
print(total(1, 2, 3))
print(total(10, 20, 30, 40, 50))
print(total(7))

print(longest("cat", "elephant", "dog", "rhinoceros", "ox"))
print(longest("hi", "hello"))
```

```python
#**kwargs - named inputs
```

```python
# Function to build profile
def build_profile(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

    print("---")


# Function to create HTML-like tag
def make_tag(tag, **attributes):
    attrs = ""

    for key, value in attributes.items():
        # Replace class_ with class
        if key == "class_":
            key = "class"

        attrs += f' {key}="{value}"'

    return f"<{tag}{attrs}>"


# Test calls
build_profile(name="Diya", age=21, city="Bhuj", cgpa=8.75)

print(make_tag("img", src="photo.jpg", width="200", height="150"))
```

```python
#Mixing *args and **kwargs
```

```python
# Function with positional argument, *args and **kwargs
def order_food(restaurant, *items, **options):
    print(f"Order from: {restaurant}")

    print(f"Items: {', '.join(items)}")

    print("Options:")
    for key, value in options.items():
        print(f"  {key}: {value}")

    print(f"Total items: {len(items)}")


# Function call
order_food(
    "Dominos",
    "Pizza",
    "Burger",
    "Coke",
    address="MG Road",
    extra_cheese=True,
    delivery_time="30min"
)
```

```python
#Argument unpacking
```

```python
import math

# Given data
coords = (10, 20, 30)

person = {
    "name": "Diya",
    "age": 21,
    "city": "Bhuj"
}


# Function to calculate 3D distance
def distance3d(x, y, z):
    return round(math.sqrt(x**2 + y**2 + z**2), 2)


# Function to introduce a person
def introduce(name, age, city):
    print(f"Hi! I am {name}, {age} years old from {city}")


# Calling functions using unpacking
print(distance3d(*coords))

introduce(**person)


# Generic apply function
def apply(func, args=(), kwargs={}):
    return func(*args, **kwargs)


# Using apply() for both functions
print(apply(distance3d, args=coords))

apply(introduce, kwargs=person)
```

```python
#Flexible data pipeline
```

```python
# Function to validate values
def validate(*values, min_val=0, max_val=100, allow_none=False):
    valid_list = []
    invalid_list = []

    for value in values:

        # Handle None values
        if value is None:
            if allow_none:
                valid_list.append(value)
            else:
                invalid_list.append(value)

        # Check valid numeric range
        elif min_val <= value <= max_val:
            valid_list.append(value)

        else:
            invalid_list.append(value)

    return valid_list, invalid_list


# Function to transform values
def transform(*values, multiply=1, add=0, round_to=2):
    transformed = []

    for value in values:
        result = round(value * multiply + add, round_to)
        transformed.append(result)

    return transformed


# Complete pipeline function
def pipeline(data, **transform_opts):

    # Validate data
    valid, invalid = validate(*data, min_val=0, max_val=100)

    # Transform valid values
    transformed = transform(*valid, **transform_opts)

    # Print summary
    print("--- Pipeline Summary ---")
    print(f"Input: {len(data)} values")
    print(f"Valid: {len(valid)} | Invalid: {len(invalid)}")
    print(f"Transformed: {transformed}")


# Test validate()
valid, invalid = validate(
    85, 102, 45, None, 67, -5, 90,
    min_val=0,
    max_val=100
)

print("Valid:", valid)
print("Invalid:", invalid)


# Test transform()
result = transform(
    *valid,
    multiply=1.1,
    add=5,
    round_to=1
)

print("Transformed:", result)


# Test full pipeline
pipeline(
    [85, 102, 45, None, 67, -5, 90],
    multiply=1.1,
    add=5,
    round_to=1
)
```
