```python
#Format specifiers - full reference:
```

```python
n = 3.14159
print(f"{n:.2f}")      # 3.14 — 2 decimal places
print(f"{n:10.2f}")    # '      3.14' — width 10
print(f"{n:<10.2f}")   # '3.14      ' — left align
print(f"{n:^10.2f}")   # '   3.14   ' — center
print(f"{n:0>10.2f}")  # '0000003.14' — zero fill right
print(f"{n:*<10.2f}")  # '3.14******' — star fill left

num = 1234567
print(f"{num:,}")      # 1,234,567
print(f"{num:_}")      # 1_234_567
print(f"{num:b}")      # binary
print(f"{num:x}")      # hex
print(f"{num:08b}")    # 00010010... zero-padded binary

ratio = 0.857
print(f"{ratio:.1%}")  # 85.7%
```

```python
width = 15
precision = 3
value = 3.14159

print(f"{value:{width}.{precision}f}")  # '          3.142'

# Dynamic alignment:
align = "<"
print(f"{'Python':{align}10}")   # 'Python    '
align = ">"
print(f"{'Python':{align}10}")   # '    Python'
```

```python
#Alignment and padding
```

```python
players = [
    ("Diya", 9200, 1),
    ("Raj", 8540, 2),
    ("Amit", 8100, 3),
    ("Sara", 7800, 4),
    ("Kiran", 7200, 5)
]

max_score = max(score for _, score, _ in players)
avg_score = sum(score for _, score, _ in players) / len(players)

for name, score, rank in players:
    percentage = (score / max_score) * 100

    # Scale bar to 20 chars
    filled = round((score / max_score) * 20)
    bar = "█" * filled + "░" * (20 - filled)

    print(
        f"#{rank:>2} "
        f"{name:<10} "
        f"{score:>6,} pts  "
        f"{bar}  "
        f"{percentage:>6.1f}%"
    )

print(f"\nAverage score: {avg_score:,.0f}")
```

```python
#Dynamic format specs
```

```python
def format_value(value, **opts):
    width = opts.get("width", 10)
    align = opts.get("align", ">")
    decimals = opts.get("decimals", 2)
    fill = opts.get("fill", " ")
    pct = opts.get("pct", False)
    comma = opts.get("comma", False)

    # Handle numbers
    if isinstance(value, (int, float)):

        # Percentage formatting
        if pct:
            value *= 100
            number = f"{value:.{decimals}f}%"
            return f"{number:{fill}{align}{width}}"

        # Integer formatting
        if isinstance(value, int):
            comma_flag = "," if comma else ""
            return f"{value:{fill}{align}{width}{comma_flag}}"

        # Float formatting
        comma_flag = "," if comma else ""
        return f"{value:{fill}{align}{width}{comma_flag}.{decimals}f}"

    # String formatting
    return f"{str(value):{fill}{align}{width}}"


data = [
    ("Revenue", 1234567.89, {"comma": True, "width": 15}),
    ("Growth", 0.1567, {"pct": True, "decimals": 1, "width": 10}),
    ("Units", 42891, {"comma": True, "width": 12}),
    ("Avg Price", 288.5, {"decimals": 2, "width": 12}),
    ("Label", "Q4-2024", {"align": "<", "width": 12})
]

for label, value, options in data:
    print(f"{label:<10}: {format_value(value, **options)}")
```

```python
#f-string powered CLI dashboard
```

```python
system = {
    "cpu": 67.3,
    "memory": 78.5,
    "disk": 45.2,
    "network_in": 125.6,
    "network_out": 45.3,
    "uptime_hours": 127,
    "processes": 342
}

# Title
print(f"{' SYSTEM MONITOR ':=^46}")

warnings = 0
criticals = 0


# Function to build progress bar + status
def metric_line(name, value):
    global warnings, criticals

    filled = int(value // 10)

    # Partial block
    partial = "▓" if value % 10 >= 5 else ""

    bar = "█" * filled + partial
    bar = bar.ljust(10, "░")

    # Status
    if value > 80:
        status = "HIGH"
        criticals += 1
    elif value > 60:
        status = "WARN"
        warnings += 1
    else:
        status = "OK"
    
    print(f"{name:<8}: {bar}  {value:>5.1f}%  [{status:<4}]")


# Metrics
metric_line("CPU", system["cpu"])
metric_line("Memory", system["memory"])
metric_line("Disk", system["disk"])

# Network
print(f"{'Net In':<8}: {system['network_in']:>6.1f} KB/s")
print(f"{'Net Out':<8}: {system['network_out']:>6.1f} KB/s")

# Uptime
days = system["uptime_hours"] // 24
hours = system["uptime_hours"] % 24
print(f"{'Uptime':<8}: {days}d {hours}h")

# Processes
print(f"{'Procs':<8}: {system['processes']}")

# Footer
print("=" * 46)

# Summary
print(f"Status: {warnings} warnings, {criticals} critical")
```

```python
# 1. Mini template engine
def render(template, **context):
    result = template

    for key, value in context.items():
        placeholder = f"{{{{{key}}}}}"
        result = result.replace(placeholder, str(value))

    return result


# 2. Table renderer
def render_table(headers, rows, **style):
    border_char = style.get("border_char", "|")
    header_sep = style.get("header_sep", "-")
    col_width = style.get("col_width", 10)

    # Header row
    header_line = border_char
    for h in headers:
        header_line += f" {h:<{col_width}} {border_char}"

    print(header_line)

    # Separator line
    print(header_sep * len(header_line))

    # Data rows
    for row in rows:
        row_line = border_char
        for item in row:
            row_line += f" {str(item):<{col_width}} {border_char}"

        print(row_line)


# 3. Report renderer
def render_report(title, sections):
    print(f"{title:=^50}")

    for heading, content_dict in sections:
        print(f"\n[{heading}]")

        for key, value in content_dict.items():
            print(f"{key:<15}: {value}")

    print("=" * 50)


# ---------------- TESTING ----------------

template = (
    "Dear {{name}}, your order #{{order_id}} "
    "of Rs.{{amount}} is {{status}}."
)

print(
    render(
        template,
        name="Diya",
        order_id="ORD-001",
        amount="1,234",
        status="shipped"
    )
)

print()

render_table(
    ["Name", "Score", "Grade"],
    [
        ["Diya", 88, "A"],
        ["Raj", 72, "B"],
        ["Amit", 95, "A+"]
    ],
    col_width=10,
    border_char="│",
    header_sep="─"
)
```
