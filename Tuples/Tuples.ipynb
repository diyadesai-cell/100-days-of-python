```python
#Indexing and slicing - same as lists:
```

```python
colors = ("red", "green", "blue", "yellow")
print(colors[0])      # red
print(colors[-1])     # yellow
print(colors[1:3])    # ('green', 'blue')
print(colors[::-1])   # ('yellow', 'blue', 'green', 'red')
```

```python
#Immutability - you CANNOT change a tuple:
```

```python
colors[0] = "pink"   # TypeError! tuples don't support assignment
colors.append("purple")  # AttributeError! no such method
```

```python
nums = (3, 1, 4, 1, 5, 9, 1)
print(nums.count(1))    # 3 — how many times 1 appears
print(nums.index(5))    # 4 — index of first 5
```

```python
#Tuple unpacking - Python's most elegant feature:
```

```python
point = (10, 20)
x, y = point          # unpack into variables
print(x, y)           # 10 20

name, age, city = ("Diya", 21, "baroda")
print(name, age, city)

# Swap variables using tuple unpacking:
a, b = 5, 10
a, b = b, a    # swap!
print(a, b)    # 10 5

# Ignore values with _:
first, _, last = (1, 2, 3)
print(first, last)    # 1 3
```

```python
#Extended unpacking with *:
```

```python
first, *rest = (1, 2, 3, 4, 5)
print(first)   # 1
print(rest)    # [2, 3, 4, 5]  ← becomes a list!

*start, last = (1, 2, 3, 4, 5)
print(start)   # [1, 2, 3, 4]
print(last)    # 5

a, *b, c = (1, 2, 3, 4, 5)
print(a, b, c)  # 1 [2, 3, 4] 5
```

```python
# Create tuple
person = ("Diya", 21, "vadodara", 8.75)

# 1. Full tuple
print(person)

# 2. Type of tuple
print(type(person))

# 3. Name (index 0)
print(person[0])

# 4. Last item using negative index
print(person[-1])

# 5. Slice of age and city
print(person[1:3])

# 6. Length of tuple
print(len(person))
```

```python
# Tuple unpacking
rgb = (255, 128, 0)

# 1. Unpack values
red, green, blue = rgb

# 2. Print with labels
print("Red:", red)
print("Green:", green)
print("Blue:", blue)

# Extended unpacking
data = (1, 2, 3, 4, 5)

# 3. Unpack first, middle, last
first, *middle, last = data

# 4. Print values
print(first)
print(middle)
print(last)
```

```python
# Tuple of student records
students = (
    ("Diya", 88, "A"),
    ("atik", 72, "B"),
    ("tisha", 95, "A"),
    ("kashish", 61, "C"),
    ("ayush", 79, "B")
)

# Variables
a_count = 0
topper_name = ""
highest_score = 0

# Loop through records
for name, score, grade in students:
    print(f"{name} | Score: {score} | Grade: {grade}")
    
    # Count A grades
    if grade == "A":
        a_count += 1
    
    # Find highest scorer
    if score > highest_score:
        highest_score = score
        topper_name = name

# Final output
print(f"A students: {a_count}")
print(f"Topper: {topper_name} ({highest_score})")
```

```python
# Original tuple
coords = (10, 20, 30)

# 1. Print original tuple and type
print(coords)
print(type(coords))

# 2. Convert to list, modify, convert back to tuple
temp_list = list(coords)
temp_list[1] = 99
new_coords = tuple(temp_list)

# 3. Print new tuple
print(new_coords)

# 4. Concatenate tuples
bigger_tuple = coords + (40, 50)

# 5. Print concatenated tuple
print(bigger_tuple)

# 6. Original tuple unchanged
print("Original unchanged:", coords)

# Tuple for count() and index()
numbers = (10, 20, 10, 30, 10)

# 7. Count how many times 10 appears
print(numbers.count(10))

# 8. Find index of 30
print(numbers.index(30))
```

```python
# GPS route data
route = (
    (19.076, 72.877, "Mumbai"),
    (28.704, 77.102, "Delhi"),
    (12.972, 77.594, "Bangalore"),
    (22.572, 88.363, "Kolkata"),
    (13.082, 80.270, "Chennai")
)

# 1. Print each waypoint
for i, (lat, lng, city) in enumerate(route, start=1):
    print(f"Stop {i}: {city} ({lat}, {lng})")

# 2. Northernmost city
north_city = max(route, key=lambda x: x[0])

# 3. Easternmost city
east_city = max(route, key=lambda x: x[1])

print(f"Northernmost: {north_city[2]} ({north_city[0]}°N)")
print(f"Easternmost: {east_city[2]} ({east_city[1]}°E)")

# 4. Average latitude and longitude
avg_lat = sum(lat for lat, lng, city in route) / len(route)
avg_lng = sum(lng for lat, lng, city in route) / len(route)

print(f"Avg lat: {avg_lat:.3f}, Avg lng: {avg_lng:.3f}")

# 5. Extended unpacking
first_stop, *middle_stops, last_stop = route

# 6. Print start and end
print(f"Start: {first_stop[2]} → End: {last_stop[2]}")

# 7. Print middle stops
middle_names = ", ".join(city for lat, lng, city in middle_stops)
print(f"Middle stops: {middle_names}")
```
