```python
#Basic syntax:
```

```python
# Regular loop:
squares = {}
for x in range(1, 6):
    squares[x] = x ** 2

# Dict comprehension — same result:
squares = {x: x**2 for x in range(1, 6)}
print(squares)   # {1:1, 2:4, 3:9, 4:16, 5:25}
```

```python
#Nested dict comprehension:
```

```python
# Build a multiplication table as nested dict:
table = {i: {j: i*j for j in range(1, 4)} for i in range(1, 4)}
print(table[2][3])   # 6
# {1:{1:1,2:2,3:3}, 2:{1:2,2:4,3:6}, 3:{1:3,2:6,3:9}}
```

```python
#Grouping with dict comprehension:
```

```python
words = ["cat", "dog", "elephant", "ant", "emu"]
by_length = {length: [w for w in words if len(w) == length]
             for length in set(len(w) for w in words)}
# {3:['cat','dog','ant'], 8:['elephant'], 3:... }
```

```python
#Dict comp basics
```

```python
# 1. Map numbers 1-8 to their cubes
cubes = {n: n**3 for n in range(1, 9)}
print(cubes)

# 2. Map each word to its length
words = ["python", "java", "rust", "go", "javascript"]
lengths = {word: len(word) for word in words}
print(lengths)

# 3. Keep only items costing more than Rs.100
prices = {"Rice": 60, "Oil": 180, "Sugar": 45, "Shampoo": 250, "Milk": 60}
expensive = {item: price for item, price in prices.items() if price > 100}
print(expensive)

# 4. Flip the prices dict
flipped = {price: item for item, price in prices.items()}
print(flipped)

# 5. Map each number 1-10 to "even" or "odd"
parity = {n: ("even" if n % 2 == 0 else "odd") for n in range(1, 11)}
print(parity)
```

```python
#From two lists and transforming
```

```python
subjects = ["Maths", "Science", "English", "CS", "Hindi"]
scores = [88, 92, 76, 95, 71]

# 1. Subject → Score
subject_scores = {sub: score for sub, score in zip(subjects, scores)}
print(subject_scores)

# 2. Subject → Grade
grades = {
    sub: ("A+" if score >= 90
          else "A" if score >= 80
          else "B" if score >= 70
          else "C")
    for sub, score in zip(subjects, scores)
}
print(grades)

# 3. Subject → Pass/Fail
result = {
    sub: ("Pass" if score >= 75 else "Fail")
    for sub, score in zip(subjects, scores)
}
print(result)

# 4. Only subjects with score >= 80
top_subjects = {
    sub: score
    for sub, score in zip(subjects, scores)
    if score >= 80
}
print(top_subjects)

# 5. Subject → Percentage (score/100)
percentages = {
    sub: score / 100
    for sub, score in zip(subjects, scores)
}
print(percentages)
```

```python
#Word frequency and grouping
```

```python
text = "the quick brown fox jumps over the lazy dog the fox"
words = text.split()

# 1. Word → Count
frequency = {word: words.count(word) for word in set(words)}
print(frequency)

# 2. Word → First Index
first_index = {word: words.index(word) for word in set(words)}
print(first_index)

# 3. Group Words by Length
grouped = {
    length: [word for word in words if len(word) == length]
    for length in {len(word) for word in words}
}
print(grouped)

# 4. Word → Appears More Than Once?
duplicates = {word: words.count(word) > 1 for word in set(words)}
print(duplicates)

# 5. Count → List of Words Having That Count
reverse_lookup = {
    count: [word for word in set(words) if words.count(word) == count]
    for count in {words.count(word) for word in set(words)}
}
print(reverse_lookup)
```

```python
#Nested dict comprehension
```

```python
# 1. Multiplication table as nested dict
table = {
    i: {j: i * j for j in range(1, 4)}
    for i in range(1, 4)
}

print(table)
print(table[3][2])


# 2. Student report dictionary
data = [
    ("Diya", [88, 92, 76]),
    ("Raj", [72, 65, 80]),
    ("Amit", [95, 98, 92])
]

students = {
    name: {
        "scores": scores,
        "total": sum(scores),
        "avg": round(sum(scores) / len(scores), 2),
        "grade": (
            "A+" if sum(scores) / len(scores) >= 90
            else "A" if sum(scores) / len(scores) >= 80
            else "B" if sum(scores) / len(scores) >= 70
            else "C"
        )
    }
    for name, scores in data
}

print(students)


# 3. Invert a nested dictionary
original = {
    "fruits": {"apple": 1, "mango": 2},
    "vegs": {"potato": 3, "onion": 4}
}

inverted = {
    item: category
    for category, items in original.items()
    for item in items
}

print(inverted)
```

```python
#Sales analytics engine
```

```python
sales = [
    {"rep":"Diya","region":"North","product":"Laptop","amount":85000,"month":"Jan"},
    {"rep":"Raj","region":"South","product":"Phone","amount":25000,"month":"Jan"},
    {"rep":"Diya","region":"North","product":"Tablet","amount":35000,"month":"Feb"},
    {"rep":"Amit","region":"East","product":"Laptop","amount":92000,"month":"Jan"},
    {"rep":"Raj","region":"South","product":"Laptop","amount":88000,"month":"Feb"},
    {"rep":"Diya","region":"North","product":"Phone","amount":22000,"month":"Feb"},
    {"rep":"Amit","region":"East","product":"Phone","amount":28000,"month":"Feb"}
]

# 1. Total sales per rep
sales_per_rep = {
    rep: sum(s["amount"] for s in sales if s["rep"] == rep)
    for rep in {s["rep"] for s in sales}
}
print(sales_per_rep)

# 2. Total sales per region
sales_per_region = {
    region: sum(s["amount"] for s in sales if s["region"] == region)
    for region in {s["region"] for s in sales}
}
print(sales_per_region)

# 3. Total sales per product
sales_per_product = {
    product: sum(s["amount"] for s in sales if s["product"] == product)
    for product in {s["product"] for s in sales}
}
print(sales_per_product)

# 4. Total sales per month
sales_per_month = {
    month: sum(s["amount"] for s in sales if s["month"] == month)
    for month in {s["month"] for s in sales}
}
print(sales_per_month)

# 5. Best product per rep (highest single sale)
best_product = {
    rep: max(
        (s for s in sales if s["rep"] == rep),
        key=lambda x: x["amount"]
    )
    for rep in {s["rep"] for s in sales}
}

for rep, sale in best_product.items():
    print(f"{rep} best: {sale['product']} (Rs.{sale['amount']})")
```
