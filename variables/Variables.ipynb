```python
#Reassigning variables — you can change a variable's value anytime:
```

```python
score = 10
score = 20      # score is now 2
score = score + 5   # score is now 25
print(score)
```

```python
#Multiple assignment — assign multiple variables in one line:
```

```python
x, y, z = 1, 2, 3
a = b = c = 0    # all three are 0
print(a,b,c)
```

```python
#Printing variables:
```

```python
name = "Raj"
age = 21
print(name)           # Raj
print("Name:", name)  # Name: Raj
print(name, "is", age, "years old")  # Raj is 21 years old
```
