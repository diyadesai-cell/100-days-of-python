```python
#Keyword arguments:
```

```python
def describe(name, age, city):
    print(f"{name}, {age}, from {city}")

describe("Diya", 21, "Bhuj")               # positional
describe(city="Bhuj", name="Diya", age=21) # keyword — any order!
describe("Diya", city="Bhuj", age=21)      # mixed
```

```python
#Functions are first-class objects:
```

```python
def square(x):
    return x ** 2

f = square         # assign function to variable
print(f(5))        # 25

nums = [3, 1, 4, 1, 5]
print(sorted(nums, key=square))   # use as argument!
```

```python
#Scope - local vs global:
```

```python
x = 10   # global

def change():
    x = 99   # local — doesn't touch global x!
    print(x)  # 99

change()
print(x)   # still 10

def change_global():
    global x
    x = 99    # NOW changes global x

change_global()
print(x)   # 99
```

```python
#Early return pattern:
```

```python
def divide(a, b):
    if b == 0:
        return None    # guard clause — exit early
    return a / b
```

```python
# Function to analyse a list
def analyse_list(nums):
    minimum = min(nums)
    maximum = max(nums)
    total = sum(nums)
    count = len(nums)
    average = total / count

    return minimum, maximum, total, average, count


# Function to check palindrome
def is_palindrome(text):
    return text == text[::-1]


# List data
data = [23, 45, 12, 67, 34, 89, 11, 56]

# Unpacking returned values
minimum, maximum, total, average, count = analyse_list(data)

# Printing results
print(f"Min: {minimum} | Max: {maximum} | Sum: {total} | Avg: {average:.2f} | Count: {count}")

# Testing palindrome function
words = ["racecar", "hello", "madam"]

for word in words:
    print(f"{word} → {is_palindrome(word)}")
```

```python
#Function toolkit
```

```python
# 1. Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32


# 2. Count vowels
def count_vowels(text):
    vowels = "aeiou"
    count = 0

    for char in text.lower():
        if char in vowels:
            count += 1

    return count


# 3. Check prime number
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True


# 4. Flatten nested list
def flatten(nested):
    flat = []

    for sublist in nested:
        for item in sublist:
            flat.append(item)

    return flat


# Test calls
print(celsius_to_fahrenheit(100))

print(count_vowels("Hello Python"))

print([n for n in range(2, 20) if is_prime(n)])

print(flatten([[1, 2, 3], [4, 5], [6, 7, 8, 9]]))
```

```python
#Student grade system as functions
```

```python
# Function to get grade
def get_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    else:
        return "F"


# Function to get result
def get_result(percentage):
    return "Pass" if percentage >= 60 else "Fail"


# Function to analyse a student
def analyse_student(name, marks_list):
    total = sum(marks_list)
    max_marks = len(marks_list) * 100
    percentage = round((total / max_marks) * 100, 2)

    return {
        "name": name,
        "total": total,
        "percentage": percentage,
        "grade": get_grade(percentage),
        "result": get_result(percentage)
    }


# Function for class summary
def class_summary(students_data):
    total_pct = 0
    pass_count = 0
    fail_count = 0

    topper = students_data[0]["name"]
    topper_pct = students_data[0]["percentage"]

    for student in students_data:
        total_pct += student["percentage"]

        if student["result"] == "Pass":
            pass_count += 1
        else:
            fail_count += 1

        if student["percentage"] > topper_pct:
            topper_pct = student["percentage"]
            topper = student["name"]

    avg_pct = round(total_pct / len(students_data), 2)

    return {
        "avg_pct": avg_pct,
        "topper": topper,
        "topper_pct": topper_pct,
        "pass_count": pass_count,
        "fail_count": fail_count
    }


# Function to print report card
def print_report(name, marks):
    student = analyse_student(name, marks)

    print(f"--- {student['name']} ---")
    print(
        f"Total: {student['total']}/{len(marks)*100} | "
        f"{student['percentage']}% | "
        f"Grade: {student['grade']} | "
        f"{student['result']}"
    )


# Test data
data = [
    ("Diya", [88, 92, 76, 85]),
    ("atik", [72, 65, 80, 70]),
    ("Amit", [95, 98, 92, 97])
]

# Print reports
for name, marks in data:
    print_report(name, marks)

# Generate class summary
summary = class_summary([analyse_student(n, m) for n, m in data])

print(f"Class avg: {summary['avg_pct']}% | Topper: {summary['topper']}")
```
