```python
#Nesting means putting one if block inside another. Each level needs one more indent:
```

```python
age = 20
has_id = True

if age >= 18:
    print("Age check passed")
    if has_id:
        print("ID check passed")
        print("Entry allowed!")
    else:
        print("No ID — entry denied")
else:
    print("Too young — entry denied")
```

```python
#Three levels deep:
```

```python
is_member = True
has_points = True
points = 850

if is_member:
    if has_points:
        if points >= 1000:
            print("Gold tier reward!")
        elif points >= 500:
            print("Silver tier reward!")
        else:
            print("Keep collecting points!")
    else:
        print("No points yet")
else:
    print("Join membership first")
```

```python
#When to nest vs when to use and:
```

```python
# Using nested if:
if age >= 18:
    if has_id:
        print("Allowed")

# Equivalent using and — cleaner for simple cases:
if age >= 18 and has_id:
    print("Allowed")

# But nesting is better when each level has its OWN else:
if age >= 18:
    if has_id:
        print("Allowed")
    else:
        print("Need ID")   # ← this else only makes sense nested
else:
    print("Too young")     # ← this else is for age check only
```

```python
#Real pattern - early exit / guard clause:
```

```python
balance = 1000
amount = 500
pin = "1234"
entered_pin = "1234"

if entered_pin != pin:
    print("Wrong PIN")
else:
    if amount <= 0:
        print("Invalid amount")
    else:
        if amount > balance:
            print("Insufficient funds")
        else:
            balance -= amount
            print(f"Success! New balance: {balance}")
```

```python
#Flattening with elif - avoid deep nesting when possible:
```

```python
# Deep nested (harder to read):
x = 5 
if x > 0:
    if x < 10:
        print("Single digit positive")

# Flat version (cleaner):
if x > 0 and x < 10:
    print("Single digit positive")

# Even better with chained:
if 0 < x < 10:
    print("Single digit positive")
```

```python
#Shopping discount system
```

```python
is_member = True
cart_total = 1500
has_coupon = True

if not is_member:
    final_amount = cart_total
    print(f"No discount. Pay Rs. {cart_total}")

else:
    if cart_total < 500:
        discount = 0.05
        print("5% member discount")

    elif cart_total < 1000:
        discount = 0.10
        print("10% member discount")

    else:
        if has_coupon:
            discount = 0.20
            print("20% discount applied!")
        else:
            discount = 0.15
            print("15% discount applied!")

    final_amount = cart_total - (cart_total * discount)

print(f"Final amount: Rs. {round(final_amount, 2)}")
```

```python
#Food delivery fee calculator
```

```python
is_app_user = True
order_total = 349
distance_km = 4
is_raining = True
has_free_delivery = False

if not is_app_user:
    delivery_fee = 60

else:
    if has_free_delivery:
        delivery_fee = 0
        print("Free delivery!")

    else:
        # Base fee
        if order_total >= 500:
            delivery_fee = 0

        elif order_total >= 300:
            delivery_fee = 20

        else:
            delivery_fee = 40

        # Distance charges
        if distance_km <= 3:
            delivery_fee += 0

        elif distance_km <= 7:
            delivery_fee += 15

        else:
            delivery_fee += 30

        # Rain surge
        if is_raining:
            delivery_fee += 10

print(f"Delivery fee: Rs. {delivery_fee}")
```
