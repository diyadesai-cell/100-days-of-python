```python
#Comparison operators always return True or False (a boolean).
```

```python
#The most common beginner mistake - = vs ==:
```

```python
x = 5       # assignment — stores value
x == 5      # comparison — checks if equal → True
```

```python
#Comparing strings:
```

```python
print("apple" == "apple")   # True
print("apple" == "Apple")   # False ← case sensitive!
print("apple" != "mango")   # True
print("banana" > "apple")   # True ← compares alphabetically!
```

```python
#Comparing different types:
```

```python
print(5 == 5.0)     # True  ← int and float can be equal
print(5 == "5")     # False ← string and int are never equal
print(True == 1)    # True  ← bool is secretly an int!
print(False == 0)   # True
```

```python
#Chaining comparisons — Python's superpower:
```

```python
age = 20
print(18 <= age <= 60)    # True ← works perfectly in Python!
# Same as: age >= 18 and age <= 60
x = 5
print(1 < x < 10)        # True
```

```python
#Storing comparison results in variables:
```

```python
score = 85
passed = score >= 50
print(passed)          # True
print(type(passed))    # <class 'bool'>
```

```python
#Comparisons with in and not in:
```

```python
name = "Diya"
print("D" in name)        # True
print("z" not in name)    # True
fruits = ["apple", "mango"]
print("mango" in fruits)  # True
```

```python
#Store comparisons Given: price = 499, budget = 500
```

```python
price = 499
budget = 500
can_afford = price <= budget
is_expensive = price > 300
exact_match = price == budget
print(can_afford)
print(is_expensive)
print(exact_match)
```
