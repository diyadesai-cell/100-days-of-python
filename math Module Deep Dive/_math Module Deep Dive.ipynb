```python
import math


# 1. Circle
def circle_area(r):
    return math.pi * r ** 2

def circle_circumference(r):
    return 2 * math.pi * r


# 2. Sphere
def sphere_volume(r):
    return (4 / 3) * math.pi * r ** 3

def sphere_surface_area(r):
    return 4 * math.pi * r ** 2


# 3. Cylinder
def cylinder_volume(r, h):
    return math.pi * r ** 2 * h

def cylinder_surface_area(r, h):
    return 2 * math.pi * r * (r + h)


# 4. Cone
def cone_volume(r, h):
    return (1 / 3) * math.pi * r ** 2 * h

def cone_slant_height(r, h):
    return math.sqrt(r ** 2 + h ** 2)


# 5. Triangle using Heron's Formula
def triangle_area_heron(a, b, c):
    s = (a + b + c) / 2
    return math.sqrt(s * (s - a) * (s - b) * (s - c))


# ---------- TESTS ----------

# Circle
print("Circle area:", round(circle_area(5), 4))
print("Circle circumference:", round(circle_circumference(5), 4))

# Sphere
print("Sphere volume:", round(sphere_volume(3), 4))
print("Sphere surface area:", round(sphere_surface_area(3), 4))

# Cylinder
print("Cylinder volume:", round(cylinder_volume(4, 10), 4))
print("Cylinder surface area:", round(cylinder_surface_area(4, 10), 4))

# Cone
print("Cone volume:", round(cone_volume(3, 4), 4))
print("Cone slant height:", round(cone_slant_height(3, 4), 4))

# Triangle
print("Triangle area:", round(triangle_area_heron(3, 4, 5), 4))
```

```python
import math


# 1. Angle Between Two 2D Vectors
def angle_between_vectors(v1, v2):
    dot = v1[0] * v2[0] + v1[1] * v2[1]

    mag1 = math.sqrt(v1[0]**2 + v1[1]**2)
    mag2 = math.sqrt(v2[0]**2 + v2[1]**2)

    cos_theta = dot / (mag1 * mag2)

    # Avoid floating-point precision issues
    cos_theta = max(-1, min(1, cos_theta))

    angle_rad = math.acos(cos_theta)

    return math.degrees(angle_rad)


# 2. Distance Functions
def distance_2d(p1, p2):
    return math.sqrt(
        (p2[0] - p1[0])**2 +
        (p2[1] - p1[1])**2
    )


def distance_3d(p1, p2):
    return math.sqrt(
        (p2[0] - p1[0])**2 +
        (p2[1] - p1[1])**2 +
        (p2[2] - p1[2])**2
    )


# 3. Bearing Between GPS Coordinates
def bearing(lat1, lon1, lat2, lon2):

    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)

    dlon = math.radians(lon2 - lon1)

    x = math.sin(dlon) * math.cos(lat2)

    y = (
        math.cos(lat1) * math.sin(lat2)
        - math.sin(lat1) * math.cos(lat2) * math.cos(dlon)
    )

    initial_bearing = math.atan2(x, y)

    bearing_deg = (math.degrees(initial_bearing) + 360) % 360

    return bearing_deg


# 4. Projectile Motion
g = 9.81


def max_height(v, angle_deg):
    angle_rad = math.radians(angle_deg)
    return (v**2 * math.sin(angle_rad)**2) / (2 * g)


def range_distance(v, angle_deg):
    angle_rad = math.radians(angle_deg)
    return (v**2 * math.sin(2 * angle_rad)) / g


def time_of_flight(v, angle_deg):
    angle_rad = math.radians(angle_deg)
    return (2 * v * math.sin(angle_rad)) / g


# ---------- TESTS ----------

print(
    "Angle between (1,0) and (0,1):",
    round(angle_between_vectors((1, 0), (0, 1)), 1),
    "°"
)

print(
    "Angle between (1,1) and (1,0):",
    round(angle_between_vectors((1, 1), (1, 0)), 1),
    "°"
)

print(
    "2D distance (0,0)→(3,4):",
    round(distance_2d((0, 0), (3, 4)), 1)
)

print(
    "3D distance (0,0,0)→(1,1,1):",
    round(distance_3d((0, 0, 0), (1, 1, 1)), 3)
)

print(
    "Max height (v=20, 45°):",
    round(max_height(20, 45), 2),
    "m"
)

print(
    "Range (v=20, 45°):",
    round(range_distance(20, 45), 2),
    "m"
)

print(
    "Time of flight:",
    round(time_of_flight(20, 45), 2),
    "s"
)


# ---------- ASCII SINE WAVE ----------

print("\nSine wave:")

for deg in range(0, 360, 30):

    sine = math.sin(math.radians(deg))

    blocks = "█" * int(abs(sine) * 5)

    print(f"{deg:3}°: {sine:5.2f} |{blocks}")
```

```python
import math


# 1. Compound Interest
def compound_interest(P, r, n, t):
    """
    P = principal
    r = annual interest rate (decimal)
    n = compounds per year
    t = years
    """

    amount = P * (1 + r / n) ** (n * t)
    interest = amount - P

    return amount, interest


# 2. Loan EMI
def loan_emi(principal, annual_rate, months):

    r = annual_rate / 12 / 100   # monthly interest rate

    emi = (
        principal * r * (1 + r) ** months
    ) / (
        (1 + r) ** months - 1
    )

    total_payment = emi * months

    return emi, total_payment


# 3. Investment Doubling Time
def investment_doubling_time(rate_percent):

    # Rule of 72
    rule72 = 72 / rate_percent

    # Exact logarithmic calculation
    r = rate_percent / 100

    exact = math.log(2) / math.log(1 + r)

    return rule72, exact


# 4. Net Present Value (NPV)
def npv(rate, cashflows):

    total = 0

    for t, cf in enumerate(cashflows):
        total += cf / ((1 + rate) ** t)

    return total


# 5. Internal Rate of Return (IRR)
def irr_estimate(cashflows, guess=0.1, iterations=100):

    rate = guess

    for _ in range(iterations):

        # NPV
        npv_value = 0

        for t, cf in enumerate(cashflows):
            npv_value += cf / ((1 + rate) ** t)

        # Derivative of NPV
        derivative = 0

        for t, cf in enumerate(cashflows[1:], start=1):
            derivative += -t * cf / ((1 + rate) ** (t + 1))

        # Newton-Raphson update
        rate = rate - (npv_value / derivative)

    return rate


# ---------- TESTS ----------

# Compound Interest
amount, interest = compound_interest(
    P=100000,
    r=0.08,
    n=1,
    t=5
)

print("Compound Interest:")
print("  Principal: Rs.100,000")
print("  Rate: 8% for 5 years")
print("  Final: Rs.", round(amount, 1))
print("  Interest earned: Rs.", round(interest, 1))


# Loan EMI
emi, total = loan_emi(
    principal=500000,
    annual_rate=9,
    months=240
)

print("\nLoan EMI:")
print("  Loan: Rs.500,000 at 9% for 240 months")
print("  Monthly EMI: Rs.", round(emi, 2))
print("  Total payment: Rs.", round(total, 1))


# Doubling Time
rule72, exact = investment_doubling_time(8)

print("\nDoubling time at 8%:")
print("  Rule of 72:", round(rule72, 1), "years")
print("  Exact:", round(exact, 2), "years")


# NPV
cashflows = [-10000, 3000, 4000, 5000, 6000]

npv_value = npv(0.10, cashflows)

print("\nNPV at 10%: Rs.", round(npv_value, 2))


# IRR
irr = irr_estimate(cashflows)

print("IRR estimate:", round(irr * 100, 1), "%")
```

```python
import math


# 1. Descriptive Statistics
def descriptive_stats(data):

    n = len(data)

    # Mean
    mean = sum(data) / n

    # Median
    sorted_data = sorted(data)

    if n % 2 == 0:
        median = (
            sorted_data[n // 2 - 1] +
            sorted_data[n // 2]
        ) / 2
    else:
        median = sorted_data[n // 2]

    # Variance (population variance)
    variance = sum((x - mean) ** 2 for x in data) / n

    # Standard deviation
    std_dev = math.sqrt(variance)

    # Skewness
    skewness = (
        sum((x - mean) ** 3 for x in data) / n
    ) / (std_dev ** 3)

    # Kurtosis
    kurtosis = (
        sum((x - mean) ** 4 for x in data) / n
    ) / (std_dev ** 4)

    return {
        "mean": mean,
        "median": median,
        "variance": variance,
        "std_dev": std_dev,
        "skewness": skewness,
        "kurtosis": kurtosis,
        "min": min(data),
        "max": max(data)
    }


# 2. Normal PDF
def normal_pdf(x, mu, sigma):

    coefficient = 1 / (sigma * math.sqrt(2 * math.pi))

    exponent = math.exp(
        -((x - mu) ** 2) / (2 * sigma ** 2)
    )

    return coefficient * exponent


# 3. Normal CDF using erf
def normal_cdf(x, mu=0, sigma=1):

    z = (x - mu) / (sigma * math.sqrt(2))

    return 0.5 * (1 + math.erf(z))


# 4. Binomial Probability
def combinations_probability(n, k, p):

    combinations = math.factorial(n) / (
        math.factorial(k) *
        math.factorial(n - k)
    )

    probability = (
        combinations *
        (p ** k) *
        ((1 - p) ** (n - k))
    )

    return probability


# 5. Z-score
def z_score(x, data):

    stats = descriptive_stats(data)

    mean = stats["mean"]
    std_dev = stats["std_dev"]

    return (x - mean) / std_dev


# ---------- TEST DATA ----------

data = [23, 45, 67, 12, 89, 34, 56, 78, 90, 11]

stats = descriptive_stats(data)

print("Descriptive Statistics:")
print("  Mean:", round(stats["mean"], 1))
print("  Median:", round(stats["median"], 1))
print("  Variance:", round(stats["variance"], 2))
print("  Std Dev:", round(stats["std_dev"], 2))
print("  Min:", stats["min"], ", Max:", stats["max"])
print("  Skewness:", round(stats["skewness"], 3))
print("  Kurtosis:", round(stats["kurtosis"], 3))


# ---------- NORMAL PDF ----------

pdf = normal_pdf(65, 65, 15)

print(
    "\nNormal PDF at x=65 (mu=65,sigma=15):",
    round(pdf, 5)
)


# ---------- PROBABILITY SCORE > 80 ----------

prob_above_80 = 1 - normal_cdf(80, 65, 15)

print(
    "P(score > 80):",
    round(prob_above_80 * 100, 2),
    "%"
)


# ---------- BINOMIAL PROBABILITY ----------

binom = combinations_probability(10, 3, 0.3)

print(
    "Binomial P(X=3) for n=10, p=0.3:",
    round(binom, 4)
)


# ---------- Z-SCORE ----------

z = z_score(90, data)

print(
    "Z-score of 90 in dataset:",
    round(z, 2)
)
```
