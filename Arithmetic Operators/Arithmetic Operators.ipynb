```python
#The three must understand deeply:
#/ always returns a float, even if perfectly divisible:
```

```python
print(10 / 2)    # 5.0  ← not 5!
```

```python
#// floor division — divides and cuts off decimals:
```

```python
print(10 // 3)   # 3   ← not 3.33
print(-7 // 2)   # -4  ← rounds DOWN (toward negative infinity!)
```

```python
#% modulus — gives the remainder after division:
```

```python
print(10 % 3)    # 1   ← because 10 = 3×3 + 1
print(15 % 5)    # 0   ← perfectly divisible, no remainder
print(7 % 2)     # 1   ← odd numbers % 2 always give 1
```

```python
#Modulus most common use - checking even or odd:
```

```python
number = 8
print(number % 2 == 0)   # True → even
```

```python
#Operator precedence - BODMAS applies:
```

```python
print(2 + 3 * 4)      # 14  ← * before +
print((2 + 3) * 4)    # 20  ← brackets first
print(2 ** 3 ** 2)    # 512 ← ** is right to left: 3**2=9, 2**9=512
```

```python
#Shorthand assignment operators:
```

```python
x = 10
x += 5    # x = x + 5  → 15
x -= 3    # x = x - 3  → 12
x *= 2    # x = x * 2  → 24
x //= 5   # x = x // 5 → 4
x **= 2   # x = x ** 2 → 16
x %= 3    # x = x % 3  → 1
```

```python
#Useful built-in math functions:
```

```python
print(abs(-9))        # 9    ← absolute value
print(round(3.7))     # 4    ← rounds to nearest
print(round(3.456, 2))# 3.46 ← rounds to 2 decimal places
print(pow(2, 8))      # 256  ← same as 2 ** 8
```

```python
#Time converter Given: total_seconds = 5483
```

```python
total_seconds = 5483
hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60
print(hours)
print(minutes)
print(seconds)
print(f"{hours}h {minutes}m {seconds}s")
```
