```python
#A for loop iterates over a sequence — a list, string, range, or anything iterable:
```

```python
fruits = ["apple", "mango", "banana"]
for fruit in fruits:
    print(fruit)
```

```python
#Looping over a string:
```

```python
for char in "Python":
    print(char)
```

```python
#Looping with range():
```

```python
for i in range(5):        # 0,1,2,3,4
    print(i)

for i in range(1, 6):     # 1,2,3,4,5
    print(i)

for i in range(1, 10, 2): # 1,3,5,7,9 (step of 2)
    print(i)

for i in range(10, 0, -1):# 10,9,8,...1 (countdown)
    print(i)
```

```python
#for with else
```

```python
for i in range(3):
    print(i)
else:
    print("Loop done!") 
```

```python
#break and continue in for:
```

```python
for i in range(1, 8):
    if i == 4:
        break         # stops at 4
    print(i)
# 1 2 3

for i in range(1, 8):
    if i % 2 == 0:
        continue      # skips even numbers
    print(i)
# 1 3 5 7
```

```python
#Accumulator with for:
```

```python
numbers = [3, 7, 2, 9, 4]
total = 0
for n in numbers:
    total += n
print(total)   
```

```python
#Nested for loops:
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

```python
#for vs while — when to use which:
#for  → when you know how many times to loop (fixed sequence)
#while → when you loop until a condition changes (unknown count)
```

```python
#Loop through a list
```

```python
sports = ["cricket", "football", "hockey", "tennis"]

for sport in sports:
    print(sport)

print("Total sports:", len(sports))
```

```python
#Odd numbers with range
```

```python
total = 0

for num in range(1, 20, 2):
    print(num)
    total += num

print("Sum of odd numbers 1-19:", total)
```

```python
#Count vowels in a string
```

```python
word = "programming"
count = 0

for ch in word:
    if ch in "aeiou":
        count += 1

print("Vowels in programming:", count)
```

```python
#FizzBuzz
```

```python
for num in range(1, 31):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
#Shopping cart analyser
```

```python
cart = [
    ("Rice", 120, 3),
    ("Oil", 180, 2),
    ("Sugar", 45, 5),
    ("Milk", 60, 4),
    ("Bread", 35, 2)
]

total_bill = 0
max_cost = 0
most_expensive = ""
items_above_200 = 0

for name, price, quantity in cart:
    total = price * quantity

    print(f"{name}: Rs.{price} x {quantity} = Rs.{total}")

    total_bill += total

    if total > max_cost:
        max_cost = total
        most_expensive = name

    if total > 200:
        items_above_200 += 1

print(f"Total bill: Rs.{total_bill}")
print(f"Most expensive item: {most_expensive} (Rs.{max_cost})")
print(f"Items above Rs.200 total: {items_above_200}")
```
