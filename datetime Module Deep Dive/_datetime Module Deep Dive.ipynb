```python
from datetime import date, datetime

def calculate_age(dob_str):
    """
    Returns exact age in years
    """
    dob = datetime.strptime(dob_str, "%Y-%m-%d").date()
    today = date.today()

    age = today.year - dob.year

    # Check if birthday has occurred this year
    if (today.month, today.day) < (dob.month, dob.day):
        age -= 1

    return age


def days_until_birthday(dob_str):
    """
    Returns days until next birthday
    """
    dob = datetime.strptime(dob_str, "%Y-%m-%d").date()
    today = date.today()

    next_birthday = date(today.year, dob.month, dob.day)

    # If birthday already passed this year
    if next_birthday < today:
        next_birthday = date(today.year + 1, dob.month, dob.day)

    return (next_birthday - today).days


def birth_stats(dob_str):
    """
    Returns detailed birth statistics
    """
    dob = datetime.strptime(dob_str, "%Y-%m-%d").date()
    today = date.today()

    # Exact age calculation
    years = today.year - dob.year
    months = today.month - dob.month
    days = today.day - dob.day

    if days < 0:
        # Borrow days from previous month
        previous_month = today.month - 1 or 12
        previous_year = today.year if today.month != 1 else today.year - 1

        days_in_prev_month = (
            date(previous_year, previous_month % 12 + 1, 1)
            - date(previous_year, previous_month, 1)
        ).days

        days += days_in_prev_month
        months -= 1

    if months < 0:
        months += 12
        years -= 1

    # Total days and hours alive
    total_days = (today - dob).days
    total_hours = total_days * 24

    # Day born
    weekday_born = dob.strftime("%A")

    # Days until next birthday
    birthday_days = days_until_birthday(dob_str)

    # Next milestone birthday (multiple of 5)
    next_milestone = ((years // 5) + 1) * 5

    return {
        "age_years": years,
        "age_months": months,
        "age_days": days,
        "age_hours": total_hours,
        "day_of_week_born": weekday_born,
        "days_until_next_birthday": birthday_days,
        "total_days_alive": total_days,
        "next_milestone": next_milestone
    }


# =========================
# TEST
# =========================

dob = "1999-03-15"

# 1. Age
print(f"Age: {calculate_age(dob)} years")

# 2. Days until birthday
print(f"Days until next birthday: {days_until_birthday(dob)}")

# 3. Birth statistics
stats = birth_stats(dob)

print("\nBirth Stats:")
print(f"  Born on: {stats['day_of_week_born']}")
print(
    f"  Age: {stats['age_years']} years, "
    f"{stats['age_months']} months, "
    f"{stats['age_days']} days"
)
print(f"  Total days alive: {stats['total_days_alive']}")
print(f"  Total hours alive: {stats['age_hours']}")
print(f"  Next birthday in: {stats['days_until_next_birthday']} days")
print(f"  Next milestone: {stats['next_milestone']}th birthday")
```

```python
from datetime import date, datetime, timedelta
import calendar


def get_month_calendar(year, month):
    """
    Returns all dates in a month as a list
    """
    num_days = calendar.monthrange(year, month)[1]

    return [
        date(year, month, day)
        for day in range(1, num_days + 1)
    ]


def working_days(start_date, end_date):
    """
    Count Monday-Friday between two dates (inclusive)
    """
    count = 0
    current = start_date

    while current <= end_date:
        if current.weekday() < 5:
            count += 1

        current += timedelta(days=1)

    return count


def list_fridays(year, month):
    """
    Returns all Fridays in a month
    """
    dates = get_month_calendar(year, month)

    return [
        d for d in dates
        if d.weekday() == 4
    ]


def next_n_weekdays(from_date, n, weekday):
    """
    Returns next n occurrences of a weekday
    weekday: 0=Mon ... 6=Sun
    """
    results = []
    current = from_date

    while len(results) < n:
        current += timedelta(days=1)

        if current.weekday() == weekday:
            results.append(current)

    return results


def meeting_scheduler(participants_free_days, duration_days=1):
    """
    Finds first weekday when ALL participants are free
    """
    common_days = set.intersection(
        *[set(days) for days in participants_free_days.values()]
    )

    if not common_days:
        return None

    weekday_order = [
        "Monday",
        "Tuesday",
        "Wednesday",
        "Thursday",
        "Friday",
        "Saturday",
        "Sunday"
    ]

    # Return earliest weekday in normal order
    for day in weekday_order:
        if day in common_days:
            return day

    return None


# ==========================================
# TESTS
# ==========================================

# 1. Jan 2025 calendar stats
jan_dates = get_month_calendar(2025, 1)

jan_working_days = working_days(
    date(2025, 1, 1),
    date(2025, 1, 31)
)

print(f"Jan 2025: {len(jan_dates)} days, {jan_working_days} working days")


# 2. Fridays in Jan 2025
fridays = list_fridays(2025, 1)

friday_numbers = [str(d.day) for d in fridays]

print("Fridays in Jan 2025:", ", ".join(friday_numbers))


# 3. Working days in Q1 2025
q1_working = working_days(
    date(2025, 1, 1),
    date(2025, 3, 31)
)

print(f"Q1 2025 working days: {q1_working}")


# 4. Next 3 Mondays from today
today = date.today()

mondays = next_n_weekdays(today, 3, 0)

formatted = [d.isoformat() for d in mondays]

print("Next 3 Mondays:", ", ".join(formatted))


# 5. Meeting scheduler
participants = {
    "Diya": ["Monday", "Wednesday", "Friday"],
    "Raj": ["Tuesday", "Wednesday", "Thursday"],
    "Amit": ["Wednesday", "Thursday", "Friday"]
}

common_day = meeting_scheduler(participants)

print(f"First common free day (weekday): {common_day}")
```

```python
from datetime import datetime, date

projects = [
    {
        "name": "Website",
        "start": "2025-01-15",
        "deadline": "2025-06-30",
        "priority": "high"
    },
    {
        "name": "Mobile App",
        "start": "2025-03-01",
        "deadline": "2025-09-15",
        "priority": "medium"
    },
    {
        "name": "API",
        "start": "2025-02-01",
        "deadline": "2025-05-31",
        "priority": "high"
    },
    {
        "name": "Dashboard",
        "start": "2025-04-01",
        "deadline": "2025-12-31",
        "priority": "low"
    }
]


def project_report(projects):
    today = date.today()

    report = []

    overdue_count = 0
    urgent_count = 0

    for project in projects:

        start = datetime.strptime(
            project["start"],
            "%Y-%m-%d"
        ).date()

        deadline = datetime.strptime(
            project["deadline"],
            "%Y-%m-%d"
        ).date()

        # Duration
        duration = (deadline - start).days

        # Elapsed days
        elapsed = (today - start).days

        # Remaining days
        remaining = (deadline - today).days

        # Completion %
        if elapsed < 0:
            progress = 0
        else:
            progress = (elapsed / duration) * 100

        # Status
        if remaining < 0:
            status = "OVERDUE"
            overdue_count += 1

        elif remaining < 14:
            status = "URGENT"
            urgent_count += 1

        elif today < start:
            status = "NOT STARTED"

        else:
            status = "ON TRACK"

        report.append({
            "name": project["name"],
            "remaining": remaining,
            "progress": progress,
            "status": status
        })

    # Sort by remaining days
    report.sort(key=lambda x: x["remaining"])

    # Print report
    print("━" * 48)
    print(f"{'Project':15} {'Remain':8} {'Progress':10} Status")
    print("━" * 48)

    for item in report:

        # Progress formatting
        if item["progress"] > 100:
            progress_text = "100%+"
        else:
            progress_text = f"{int(item['progress'])}%"

        remain_text = f"{item['remaining']}d"

        print(
            f"{item['name']:15} "
            f"{remain_text:8} "
            f"{progress_text:10} "
            f"{item['status']}"
        )

    print("━" * 48)

    print(
        f"Total: {len(projects)} | "
        f"Overdue: {overdue_count} | "
        f"Urgent: {urgent_count}"
    )


# =========================
# RUN REPORT
# =========================

project_report(projects)
```

```python
from datetime import date, datetime, timedelta
import calendar


class RecurringEvent:

    def __init__(self, name, start_date, frequency):
        self.name = name
        self.start_date = datetime.strptime(
            start_date,
            "%Y-%m-%d"
        ).date()

        self.frequency = frequency.lower()

    def next_occurrence(self, from_date=None):

        if from_date is None:
            from_date = date.today()

        current = self.start_date

        # Daily recurrence
        if self.frequency == "daily":

            if current >= from_date:
                return current

            days_passed = (from_date - current).days

            return current + timedelta(days=days_passed)

        # Weekly recurrence
        elif self.frequency == "weekly":

            while current < from_date:
                current += timedelta(weeks=1)

            return current

        # Monthly recurrence
        elif self.frequency == "monthly":

            year = from_date.year
            month = from_date.month

            while True:

                try:
                    candidate = date(
                        year,
                        month,
                        self.start_date.day
                    )

                except ValueError:
                    # Handle invalid dates like Feb 30
                    last_day = calendar.monthrange(year, month)[1]

                    candidate = date(year, month, last_day)

                if candidate >= from_date:
                    return candidate

                # Move to next month
                month += 1

                if month > 12:
                    month = 1
                    year += 1

        # Yearly recurrence
        elif self.frequency == "yearly":

            year = from_date.year

            candidate = date(
                year,
                self.start_date.month,
                self.start_date.day
            )

            if candidate < from_date:
                candidate = date(
                    year + 1,
                    self.start_date.month,
                    self.start_date.day
                )

            return candidate

    def occurrences_between(self, start, end):

        occurrences = []

        current = self.next_occurrence(start)

        while current <= end:

            occurrences.append(current)

            # Move to next occurrence
            if self.frequency == "daily":
                current += timedelta(days=1)

            elif self.frequency == "weekly":
                current += timedelta(weeks=1)

            elif self.frequency == "monthly":

                month = current.month + 1
                year = current.year

                if month > 12:
                    month = 1
                    year += 1

                day = self.start_date.day

                last_day = calendar.monthrange(year, month)[1]

                current = date(
                    year,
                    month,
                    min(day, last_day)
                )

            elif self.frequency == "yearly":

                current = date(
                    current.year + 1,
                    self.start_date.month,
                    self.start_date.day
                )

        return occurrences

    def days_until_next(self):

        next_date = self.next_occurrence()

        return (next_date - date.today()).days


class EventCalendar:

    def __init__(self):
        self.events = []

    def add_event(self, recurring_event):
        self.events.append(recurring_event)

    def get_events_on(self, target_date):

        matching = []

        for event in self.events:

            occurrence = event.next_occurrence(target_date)

            if occurrence == target_date:
                matching.append(event.name)

        return matching

    def next_7_days(self):

        today = date.today()

        agenda = {}

        for i in range(7):

            current_day = today + timedelta(days=i)

            agenda[current_day] = self.get_events_on(current_day)

        return agenda

    def conflicts(self):

        agenda = self.next_7_days()

        return {
            day: events
            for day, events in agenda.items()
            if len(events) > 1
        }


# ==========================================
# TESTING
# ==========================================

birthday = RecurringEvent(
    "Birthday",
    "1999-03-15",
    "yearly"
)

salary = RecurringEvent(
    "Salary Day",
    "2025-01-01",
    "monthly"
)

meeting = RecurringEvent(
    "Team Meeting",
    "2025-01-06",   # Monday
    "weekly"
)

standup = RecurringEvent(
    "Daily Standup",
    str(date.today()),
    "daily"
)

# Event calendar
calendar_app = EventCalendar()

calendar_app.add_event(birthday)
calendar_app.add_event(salary)
calendar_app.add_event(meeting)
calendar_app.add_event(standup)

# Next occurrences
next_birthday = birthday.next_occurrence()

print(
    f"Next birthday: {next_birthday} "
    f"({birthday.days_until_next()} days away)"
)

next_salary = salary.next_occurrence()

print(
    f"Next salary day: {next_salary} "
    f"({salary.days_until_next()} days away)"
)

next_meeting = meeting.next_occurrence()

print(
    f"Next Monday meeting: {next_meeting} "
    f"({meeting.days_until_next()} days away)"
)

# Agenda
print("\n--- 7 Day Agenda ---")

agenda = calendar_app.next_7_days()

for day, events in agenda.items():

    if events:
        print(f"{day}: {', '.join(events)}")

# Occurrence counts
birthdays_2025 = birthday.occurrences_between(
    date(2025, 1, 1),
    date(2025, 12, 31)
)

meetings_june = meeting.occurrences_between(
    date(2025, 6, 1),
    date(2025, 6, 30)
)

print(
    f"\nBirthdays in 2025: "
    f"{len(birthdays_2025)} occurrence"
)

print(
    f"Team meetings in June 2025: "
    f"{len(meetings_june)} occurrences"
)
```
